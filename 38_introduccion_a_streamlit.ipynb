## ¿Qué es *Streamlit*?

[*Streamlit*](https://streamlit.io/) es una biblioteca de código abierto que permite crear aplicaciones web interactivas para ciencia de datos y machine learning **sin necesidad de conocimientos de HTML, CSS o JavaScript**.

Con *Streamlit* es posible transformar scripts de *Python* en *dashboards* y aplicaciones web en cuestión de minutos.

### Características principales.

| Característica | Descripción |
| :--- | :--- |
| **Simplicidad** | *API* declarativa y minimalista. |
| **Interactividad** | *Widgets* integrados (*sliders*, entradas, selectores). |
| **Rendimiento** | *Caching* automático de cálculos costosos. |
| **Despliegue** | Integración directa con *GitHub* para *hosting* gratuito. |
| **Integración** | Compatible con *Polars*, *Pandas*, *Matplotlib*, *Plotly*, etc. |

https://docs.streamlit.io/

https://github.com/streamlit/streamlit

## Estructura básica de una aplicación *Streamlit*.

La estructura mínima de una aplicación *Streamlit* es un script de *Python* que se ejecuta desde la terminal con `streamlit run <archivo>.py`.

```python
import streamlit as st
import polars as pl

# Configurar la página
st.set_page_config(page_title='Mi App', layout='wide')

# Título
st.title('Mi Primera App de Streamlit')

# Contenido
st.write('¡Hola Streamlit!')
```

**Para ejecutar desde la terminal:**
```bash
streamlit run app.py
```

## *Widgets* interactivos.

Los *widgets* permiten que el usuario interactúe con la aplicación. Cada vez que el usuario cambia un *widget*, *Streamlit* re-ejecuta el script automáticamente con el nuevo valor.

| *Widget* | Descripción |
| :--- | :--- |
| [`st.button()`](https://docs.streamlit.io/develop/api-reference/widgets/st.button) | Botón de acción. |
| [`st.text_input()`](https://docs.streamlit.io/develop/api-reference/widgets/st.text_input) | Entrada de texto. |
| [`st.number_input()`](https://docs.streamlit.io/develop/api-reference/widgets/st.number_input) | Entrada numérica con mínimo y máximo. |
| [`st.slider()`](https://docs.streamlit.io/develop/api-reference/widgets/st.slider) | Deslizador de rango. |
| [`st.selectbox()`](https://docs.streamlit.io/develop/api-reference/widgets/st.selectbox) | Menú desplegable (una opción). |
| [`st.multiselect()`](https://docs.streamlit.io/develop/api-reference/widgets/st.multiselect) | Menú desplegable (múltiples opciones). |
| [`st.checkbox()`](https://docs.streamlit.io/develop/api-reference/widgets/st.checkbox) | Casilla de verificación. |
| [`st.radio()`](https://docs.streamlit.io/develop/api-reference/widgets/st.radio) | Selección de opción única (botones de radio). |

```python
# Entrada de texto
nombre = st.text_input('¿Cuál es tu nombre?')
st.write(f'Hola, {nombre}')

# Slider
valor = st.slider('Selecciona un valor', 0, 100, 50)
st.write(f'Valor seleccionado: {valor}')

# Selectbox

st.write(f'Región seleccionada: {region}')
```

## *Caching* y rendimiento.

*Streamlit* re-ejecuta el script completo cada vez que el usuario interactúa. El *caching* evita recalcular operaciones costosas en cada re-ejecución.

| Decorador | Uso |
| :--- | :--- |
| [`@st.cache_data`](https://docs.streamlit.io/develop/api-reference/caching-and-state/st.cache_data) | Datos serializables (CSV, *DataFrames*, resultados de consultas). |
| [`@st.cache_resource`](https://docs.streamlit.io/develop/api-reference/caching-and-state/st.cache_resource) | Recursos no serializables (conexiones BD, modelos ML). |

```python
import polars as pl

@st.cache_data
def cargar_datos(url: str) -> pl.DataFrame:
    return pl.read_csv(url)   # Primera llamada: lee el archivo
                              # Llamadas siguientes: devuelve el caché

@st.cache_resource
def conectar_bd():
    return crear_conexion_bd()  # Se conecta solo una vez

df = cargar_datos('ventas.csv')
st.dataframe(df)
```

## Integración con *Polars*.

*Streamlit* acepta *DataFrames* de *Polars* directamente en `st.dataframe()` desde la versión 1.23. Para las funciones de gráficos nativos (`st.line_chart()`, `st.bar_chart()`) se recomienda usar *Pandas* o *Plotly* como intermedio.

**Flujo recomendado:**

```
Polars (carga y transformación rápida)
       ↓  .to_pandas()  (solo para gráficos nativos)
    Pandas / Plotly
       ↓
    Streamlit
```

```python
import streamlit as st
import polars as pl
import plotly.express as px

@st.cache_data
def cargar_y_procesar(url: str) -> pl.DataFrame:
    return (
        pl.read_csv(url)
        .filter(pl.col('ventas') > 0)
        .with_columns(pl.col('fecha').str.to_date())
    )

df = cargar_y_procesar('ventas.csv')

# Polars directo en st.dataframe()
st.dataframe(df)

# Polars para filtros en sidebar
filtro = st.sidebar.slider('Ventas mínimas', 0, 10_000, 1_000)
df_filtrado = df.filter(pl.col('ventas') >= filtro)

# Plotly para gráfico interactivo
fig = px.line(df_filtrado.to_pandas(), x='fecha', y='ventas')
st.plotly_chart(fig, use_container_width=True)
```

## Ejemplo completo: *dashboard* de ventas.

La siguiente celda crea el archivo `streamlit-project/app.py` con un *dashboard* completo listo para ejecutarse. Usa *Polars* para la carga y procesamiento, y *Plotly* para las visualizaciones.

* La siguiente celda usa la *magic* `%%writefile` para crear directamente el archivo `streamlit-project/app.py`. Al ejecutarla, el contenido de la celda se escribe en disco tal como aparece — el código Streamlit es visible y editable sin escapes adicionales.

````python
%%writefile streamlit-project/app.py
import streamlit as st
import polars as pl
import numpy as np
import plotly.express as px
from datetime import date, timedelta

st.set_page_config(
    page_title='Dashboard de Ventas',
    page_icon='📊',
    layout='wide',
)

st.title('📊 Dashboard de Ventas')
st.markdown('---')


@st.cache_data
def generar_datos() -> pl.DataFrame:
    np.random.seed(42)
    n = 200
    start = date(2024, 1, 1)
    return pl.DataFrame({
        'fecha':    [(start + timedelta(days=i)).isoformat() for i in range(n)],
        'region':   (['Norte', 'Sur', 'Este', 'Oeste'] * 50),
        'ventas':   np.random.randint(1_000, 10_000, n).tolist(),
        'clientes': np.random.randint(10, 100, n).tolist(),
    }).with_columns(pl.col('fecha').str.to_date())


df = generar_datos()

# Barra lateral
st.sidebar.title('🎛️ Filtros')
regiones = st.sidebar.multiselect(
    'Región',
    options=df['region'].unique().to_list(),
    default=df['region'].unique().to_list(),
)
df_f = df.filter(pl.col('region').is_in(regiones))

# Métricas
c1, c2, c3, c4 = st.columns(4)
c1.metric('Total ventas',    f"${df_f['ventas'].sum():,.0f}")
c2.metric('Promedio ventas', f"${df_f['ventas'].mean():,.0f}")
c3.metric('Total clientes',  f"{df_f['clientes'].sum():,}")
c4.metric('Registros',       str(len(df_f)))

st.markdown('---')

# Gráficos
col1, col2 = st.columns(2)
df_pd = df_f.to_pandas()

with col1:
    fig = px.line(df_pd, x='fecha', y='ventas', color='region',
                  title='Ventas en el tiempo')
    st.plotly_chart(fig, use_container_width=True)

with col2:
    resumen = (df_f
               .group_by('region')
               .agg(pl.col('ventas').sum())
               .sort('ventas', descending=True))
    fig = px.bar(resumen.to_pandas(), x='region', y='ventas',
                 title='Ventas por región')
    st.plotly_chart(fig, use_container_width=True)

# Tabla y descarga
st.subheader('📋 Datos detallados')
st.dataframe(df_f, use_container_width=True)

csv = df_f.to_pandas().to_csv(index=False)
st.download_button('Descargar CSV', csv, 'ventas.csv', 'text/csv')
````

## Despliegue en *Streamlit Cloud*.

[*Streamlit Community Cloud*](https://share.streamlit.io/) permite desplegar aplicaciones directamente desde un repositorio de *GitHub* de forma gratuita.

**Pasos para desplegar:**

1. Crear un repositorio en *GitHub* con:
   - `app.py` (o el nombre del archivo principal)
   - `requirements.txt` (con las dependencias)
2. Acceder a https://share.streamlit.io/ e iniciar sesión con *GitHub*.
3. Seleccionar el repositorio, la rama y el archivo principal.
4. La app queda publicada en `https://usuario-repo.streamlit.app/`.

**Ejemplo de `requirements.txt`:**
```
streamlit>=1.35
polars>=1.0
plotly>=5.18
numpy>=1.24
```

## Resumen.

*Streamlit* cierra el ciclo del análisis de datos: desde la carga y transformación con *Polars* hasta la presentación interactiva en la web, sin salir del ecosistema *Python*.

| Concepto | Descripción |
| :--- | :--- |
| **`st.write()` / `st.markdown()`** | Elementos de texto y contenido general. |
| **`st.dataframe()`** | Tabla interactiva; acepta *DataFrames* de *Polars* directamente. |
| **`st.plotly_chart()`** | Integración con *Plotly* para gráficos interactivos. |
| ***Widgets*** | `slider`, `selectbox`, `multiselect`, `button`, etc. |
| **`@st.cache_data`** | *Caching* de datos para evitar re-cálculos en cada interacción. |
| **`st.columns()` / `st.sidebar`** | Organización visual del *layout*. |
| **Flujo recomendado** | Polars → Polars/Plotly → Streamlit → *Cloud*. |
