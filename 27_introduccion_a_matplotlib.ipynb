## El objeto ```mpl.pyplot```.

*Matplotlib* contiene una muy extensa biblioteca, la cual tiene como componente principal al objeto ```matplotlib.pyplot```. 

Por convención, ```matplotlib.pyplot``` se importa como ```plt```. Se utilizará esta convención para próximas referencias.

El uso de ```plt``` permite crear objetos capaces de desplegar gráficos y definir múltiples características de éstos.

A lo largo de este capítulo se explorarán algunos de dichos recursos.

https://matplotlib.org/stable/api/pyplot_summary.html

````python
from matplotlib import pyplot as plt
````

### El método ```plt.plot()```.

Esta función permite crear una o más gráficas en 2 dimensiones a partir de arreglos de puntos para los ejes ```x``` y ```y``` que son ingresados como argumentos.

```
plt.plot(<arreglo para x1>, <arreglo para y1>,
 <arreglo para x2>, <arreglo para y2>, ... 
 <arreglo para xn>, <arreglo para yn>, <otros argumentos>)
```

Donde:
- ```<arreglo para xi>``` es un arreglo de puntos para el eje ```x```. 
- ```<arreglo para yi>``` es un arreglo de puntos para el eje ```y```.
- ```<otros argumentos>``` son argumentos opcionales que permiten personalizar la apariencia de las gráficas, como el color, el estilo de línea, el marcador, entre otros.

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

### El método ```plt.show()```.

El método ```plt.show()``` permite desplegar el gráfico creado con ```plt.plot()``` en el entorno desde el que se ejecutan las instrucciones.

**NOTA:** En el caso de las notebooks de *Jupyter*, no es necesario usar ```plt.show()``` para que el gráfico se despliegue al ejecutar una celda con ```plt.plot()```.

**Ejemplos:**

* Las siguientes celdas crearán un arreglo unidimensional de *Numpy* con nombre ```x```, el cual contendrá ```500``` segmentos lineales que van de ```0``` a ```3π``` usando la función ```np.linspace()```.

````python
x = np.linspace(0, 3*np.pi, 500)
````

````python
x
````

* La siguiente celda utilizará ```plt.plot()``` para desplegar dos gráficas que unirán cada punto definido mediante una línea, las cuales están en función del arreglo ```x```:

* ```np.sin(x ** 2)```
* ```np.cos(x ** 2)```

````python
plt.plot(x, np.sin(x**2), x, np.cos(x ** 2))
````

````python
plt.show()
````

* Las celdas anteriores crearon de forma automática un objeto de tipo ```plt.Figure```, el cual fue utilizado para desplegar las gráficas correspondientes.

* La siguiente celda incluye algunas funciones de ```plt``` que definen el título del gráfico (```plt.title()```), y las etiquetas de los ejes (```plt.xlabel``` y ```plt.ylabel```).

````python
plt.plot(x, np.sin(x**2), x, np.cos(x ** 2))
plt.title('Funciones sinusoidales')
plt.xlabel('Eje de las x')
plt.ylabel('f(x)')
plt.show()
````

## Otras funcionalidades de *Matplotlib*.

*Matplotlib* va más allá de los gráficos 2D. Sus módulos adicionales permiten crear visualizaciones tridimensionales y manipular imágenes directamente en el entorno de Python.

### Gráficos 3D.

El módulo `mpl_toolkits.mplot3d` permite crear gráficos tridimensionales: superficies, 
nubes de puntos, curvas espaciales y más. Se activa al crear una figura con `projection='3d'`.

````python
from mpl_toolkits.mplot3d import Axes3D

# Datos: superficie senoidal en 3D
theta = np.linspace(0, 2 * np.pi, 100)
X, Y = np.meshgrid(theta, theta)
Z = np.sin(X) * np.cos(Y)

fig = plt.figure(figsize=(8, 5))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title('Superficie 3D: sin(x)·cos(y)')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.tight_layout()
plt.show()
````

### Procesamiento de imágenes.

Con `matplotlib.image` se pueden cargar imágenes como arreglos NumPy (forma `(alto, ancho, canales)`). 
`plt.imshow()` las despliega y permite aplicar transformaciones como recortes, escala de grises o mapas de color.

````python
import matplotlib.image as mpimg

img = mpimg.imread('img/pythonista.png')  # ndarray shape: (alto, ancho, canales)
print(f'Forma: {img.shape}  |  Dtype: {img.dtype}  |  Min: {img.min():.2f}  Max: {img.max():.2f}')

fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# Original
axes[0].imshow(img)
axes[0].set_title('Imagen original')
axes[0].axis('off')

# Escala de grises (promedio de canales RGB)
img_gris = img[:, :, :3].mean(axis=2)
axes[1].imshow(img_gris, cmap='gray')
axes[1].set_title('Escala de grises')
axes[1].axis('off')

plt.tight_layout()
plt.show()
````
