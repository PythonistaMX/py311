## El atributo ```plot``` de los objetos `DataFrame` y `Series`.

Este atributo contiene una colección de métodos que permiten desplegar gráficos descriptivos básicos basados en *Matplotlib*.

Algunos de los métodos más comunes son:

* ```pd.plot.line()```: para gráficos de líneas.
* ```pd.plot.bar()```: para gráficos de barras.
* ```pd.plot.barh()```: para gráficos de barras horizontales.
* ```pd.plot.pie()```: para gráficos de pastel.
* ```pd.plot.hist()```: para histogramas.
* ```pd.plot.scatter()```: para gráficos de dispersión.
* ```pd.plot.box()```: para diagramas de caja.
* ```pd.plot.area()```: para gráficos de área.

En este capítulo se introducen los métodos básicos. Los capítulos siguientes profundizan en cada tipo de gráfica.

### Creación de gráficos con `plot()`.

Para crear un gráfico con el método ```plot()```, se puede seguir la siguiente sintaxis:

```
<obj>.plot(kind='tipo_de_grafico', x='columna_x', y='columna_y', title='Título del gráfico', xlabel='Etiqueta eje x', ylabel='Etiqueta eje y', **kwargs)
```

Donde:
* ```<obj>```: puede ser un objeto `DataFrame` o `Series` de *Pandas*.
* ```kind```: especifica el tipo de gráfico que se desea crear (por ejemplo, 'line', 'bar', 'hist', 'scatter', 'box').
* ```x```: indica la columna del DataFrame que se utilizará para el eje x.
* ```y```: indica la columna del DataFrame que se utilizará para el eje y.
* ```title```: establece el título del gráfico.
* ```xlabel```: establece la etiqueta del eje x.
* ```ylabel```: establece la etiqueta del eje y.

Además de estos parámetros básicos, el método `plot()` acepta una amplia variedad de argumentos adicionales (`**kwargs`) derivados de *Matplotlib* que permiten personalizar aún más el gráfico, como colores, estilos de línea, tamaños de marcador, leyendas, entre otros.

## Ejemplo de uso.

* El archivo ```data/ITAEE.csv``` contiene una tabla estática de los datos por entidad federativa del Índice Trimestral de la Actividad Económica Estatal (ITAEE) de México, con datos trimestrales y anuales desde 1980 hasta 2025.

* La siguiente celda creará al *dataframe* ```datos``` a partir de cargar el archivo ```data/ITAEE.csv```.

````python
import os
datos = pd.read_csv(os.path.join('data','ITAEE.csv'), encoding='utf-8', index_col=0)
````

* Se puede observar que el *dataframe* tiene un índice que contiene el nombre de cada entidad federativa, y otras columnas que contienen los datos trimestrales y anuales del ITAEE.

````python
datos
````

* Debido a que los archivos *CSV* no soportan tipos de datos complejos ni *multiindex*, el *dataframe* necesita ser procesado antes de poder ser utilizado.

````python
datos.info()
````

### 1. Adecuación de las columnas.

* La siguiente celda extraerá los años que comprenden los datos trimestrales y anuales del ITAEE.

````python
rango = list(set((int(cadena[:4])) for cadena in datos.columns.to_list()))
rango.sort()
rango
````

* La siguiente celda construirá un objeto `MultiIndex` a partir de los años extraídos y los períodos trimestrales y anuales y lo llamará `columnas`.

````python
columnas = pd.MultiIndex.from_product([pd.Index(rango, name="Año"), pd.Index(['T1', 'T2', 'T3', 'T4', 'Anual'], name="periodo")])
columnas
````

* La siguiente celda asignará el contenido de `columnas` al índice de las columnas del *dataframe* ```datos```.

````python
datos.columns = columnas
datos
````

### 2. Delimitación y conversión de datos.

* Se tomarán los datos de la fila `1` en adelante.

````python
datos = datos[1:]
datos
````

* La siguiente celda convertirá los datos del *dataframe* ```datos``` a tipo numérico, para que puedan ser analizados y graficados.

````python
datos = datos.astype(float)
````

````python
datos
````

### 3. Algunas gráficas básicas.

A continuación se muestran algunos ejemplos de gráficas básicas que se pueden crear con el método `plot()` de *Pandas*.

* La siguiente celda mostrará un gráfico de barras con el ITAEE anual por entidad federativa en 2014. El acceso `datos[2014]["Anual"]` aprovecha el *MultiIndex* de las columnas para obtener directamente una *Series* con las entidades como índice.

````python
datos[2014]["Anual"].plot.bar(
    title='ITAEE anual por entidad federativa - 2014',
    xlabel='Entidad',
    ylabel='Índice (base 2018=100)',
    figsize=(14, 5),
    rot=45
)
````

* La siguiente celda mostrará un gráfico de barras horizontales con el mismo corte de datos utilizando el método `plot.barh()`.

````python
datos[2014]["Anual"].plot.barh(
    title='ITAEE anual por entidad federativa - 2014',
    xlabel='Índice (base 2018=100)',
    figsize=(8, 9)
)
````

### 4. Creación de un nuevo *dataframe*.

En este caso se graficará el ITAEE anual por entidad federativa, por lo que es necesario reestructurar el *dataframe* para que cada columna contenga un tipo de información específico como la entidad federativa y el año.

* La siguiente celda creará un nuevo *dataframe* llamado `datos_anuales` a partir del *dataframe* `datos`, tomando únicamente los datos anuales y reestructurando el índice para que contenga la entidad federativa en el eje de las columnas y el año en el índice.

````python
datos_anuales = datos.xs('Anual', level='periodo', axis=1).transpose()
datos_anuales
````

### 5. Visualización con `datos_anuales`.

A continuación se muestran algunos ejemplos de visualización de los datos del *dataframe* `datos_anuales` por medio de los métodos del atributo `plot`.

Los métodos de visualización de datos del atributo `plot` permiten personalizar los gráficos con diferentes opciones, como el título, las etiquetas de los ejes, el tamaño de la figura, entre otras, aprovechando los parámetros de los métodos de *Matplotlib*.

En este caso, se utilizarán los parámetros por defecto para mostrar la simplicidad de uso de los métodos del atributo `plot`.

Los parámetros de *Matplotlib* se estudiarán en capítulos posteriores, pero la documentación oficial está disponible en:

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

* La siguiente celda mostrará un gráfico de líneas con la evolución anual del ITAEE del país utilizando el método ```plot()```.

````python
datos_anuales['Estados Unidos Mexicanos'].plot(
    kind="line",
    title='ITAEE anual por entidad federativa',
    xlabel='Estados Unidos Mexicanos',
    ylabel='Índice (base 2018=100)',
    figsize=(14, 5),
    xticks=datos_anuales.index,
    rot=45,
    grid=True
)
````
