# Gestión de fechas.

El proyecto *Pandas* fue creado originalmente para poder manejar datos financieros, y por lo tanto, el manejo de fechas y series de tiempo es una parte fundamental de su funcionalidad. En esta sección, exploraremos cómo trabajar con fechas en *Pandas*.


En *Pandas*, las series de tipo `datetime` cuentan con un *accessor* especial llamado `.dt`, el cual permite extraer componentes y realizar transformaciones sobre los valores de fechas en una serie.

````python
from datetime import datetime
import pandas as pd
import numpy as np
````

## La función `pd.date_range()`.

La función `pd.date_range()` permite generar un rango de fechas con una frecuencia específica.


```
pd.date_range(start=<inicio>, end=<fin>, freq=<frecuencia>)
```
Donde:
* `<inicio>` es la fecha de inicio del rango.
* `<fin>` es la fecha de fin del rango.
* `<frecuencia>` es la frecuencia con la que se generan las fechas.

Las frecuencias más comunes son:
* `D`: diaria.
* `ME`: mensual (fin de mes).
* `YE`: anual (fin de año).
* `h`: horaria.
* `min`: por minuto.
* `s`: por segundo.

El resultado de `pd.date_range()` es un objeto de tipo `DatetimeIndex`, que es una colección de fechas.

La documentación de `pd.date_range()` se puede encontrar en: 

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.date_range.html

**Ejemplos:**

* La siguiente celda regresará un objeto `DatetimeIndex` con fechas diarias desde el 1 de enero de 2019 hasta el 31 de diciembre de 2019.

````python
pd.date_range('2019-01-01', '2019-12-31', freq='D')
````

* La siguiente celda regresará un objeto `DatetimeIndex` con fechas semanales desde el 1 de enero de 2019 hasta el 31 de diciembre de 2019 usando la zona horaria `"America/Mexico_City"`.

````python
pd.date_range('2019-01-01', '2019-12-31', freq='W', tz='America/Mexico_City')
````

* La siguiente celda regresará un objetos `DatetimeIndex` con un rango horario que va desde las `00:00:00` del 1 de enero de 2019 hasta las `23:00:00` del 7 de enero de 2019.

````python
pd.date_range('2019-01-01', '2019-01-07', freq='h')
````

## *Dataframe* ilustrativo.

Se creará un *dataframe* sintético llamado `ventas` con un índice de fechas que van del 1 de enero de 2019 al 31 de diciembre de 2019 en frecuencia diaria y una columna con valores numéricos asociados.

````python
fechas = pd.date_range(start="2019-01-01", end="2019-12-31", freq="D")
fechas
````

````python
ventas = pd.DataFrame({'fechas': fechas, 'ventas': np.random.randint(1000, 5000, size=len(fechas))}).set_index('fechas')
````

## El atributo `Series.dt`.

Las series de tipo `datetime` cuentan con el atributo `.dt`, el cual permite extraer componentes específicos de las fechas, como el año, mes, día, etc.

Los componentes más comunes que se pueden extraer con `.dt` son:
* `.dt.year`: extrae el año de la fecha.
* `.dt.month`: extrae el mes de la fecha.
* `.dt.day`: extrae el día de la fecha.
* `.dt.hour`: extrae la hora de la fecha.
* `.dt.minute`: extrae los minutos de la fecha.
* `.dt.second`: extrae los segundos de la fecha.
* `.dt.weekday`: extrae el día de la semana de la fecha (0 para lunes, 6 para domingo).
* `.dt.day_name()`: extrae el nombre del día de la semana de la fecha.
* `.dt.month_name()`: extrae el nombre del mes de la fecha.

La documentación de los atributos de `Series.dt` se puede encontrar en:

https://pandas.pydata.org/docs/reference/series.html#api-series-dt

## El método `df.resample()`.

El método `resample()` permite reagrupar una serie temporal por una frecuencia de tiempo y aplicar una función de agregación.

```
df.resample(<frecuencia>).<func>()
```
Donde:
* `<frecuencia>` es la frecuencia con la que se desea reagrupar los datos.
* `<func>` es la función de agregación que se desea aplicar, como `sum()`, `mean()`, `max()`, etc.

Frecuencias comunes:

| Alias | Frecuencia |
| :--- | :--- |
| `'D'` | Diaria |
| `'W'` | Semanal |
| `'ME'` | Fin de mes |
| `'QE'` | Fin de trimestre |
| `'YE'` | Fin de año |

El resultado de `df.resample()` es un nuevo *dataframe* con el índice reagrupado según la frecuencia especificada y los valores agregados según la función aplicada de forma similar a `groupby()`.
