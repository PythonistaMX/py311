## El método ```apply()```.

El método ```apply()``` permite aplicar una función a una serie o dataframe de *Pandas*.

```
<obj>.apply(<func>, axis=<eje>)
```

Donde:

* ```<obj>``` es una serie o un *dataframe* de *Pandas*.
* ```<func>``` es una función de *Python* o de *Numpy*.
* ```<eje>``` puede ser:
 * ```0``` para aplicar la función a los renglones. Este es el valor por defecto.
 * ```1``` para aplicar la función a las columnas.

Este método realiza operaciones de *broadcast* dentro del objeto.

Para fines prácticos se explorará el método ```pd.DataFrame.apply()``` cuya documentación puede ser consultada en:

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.apply.html

### Funciones aceptadas.

* El método ```apply()``` permite ingresar como argumento el nombre de una función o una función *lambda* de *Python*.

**Ejemplos:**

* La siguiente celda definirá a la función ```suma_dos()```.

````python
def suma_dos(x:int) -> int:
    ''''Función que regresa el resultado de sumar 2 unidades a un entero.'''
    return x + 2
````

* La siguiente celda regresará un *dataframe* que contiene el resultado de ejecutar la función ```suma_dos()``` usando a cada elemento del *dataframe* ```población``` como argumento.

````python
poblacion
````

````python
poblacion.apply(suma_dos)
````

* La siguiente celda regresará un *dataframe* que contiene el resultado de ejecutar la función definida como ```lambda x: x + 2``` usando a cada elemento del *dataframe* ```población``` como argumento.

````python
poblacion.apply(lambda x: x + 2)
````

* La siguiente celda regresará una serie que corresponde a ejecutar la función ```suma_dos()``` a cada elemento de la serie que conforma la columna ```poblacion['Norte_2']```. 

````python
poblacion['Norte_2'].apply(suma_dos)
````

### Consideraciones de rendimiento.

Si bien `apply()` es flexible y acepta funciones personalizadas, su rendimiento es **inferior al de las operaciones vectorizadas** de *Pandas* y *NumPy*. Esto se debe a que `apply()` itera elemento por elemento en lugar de aprovechar las optimizaciones internas de *NumPy*.

Como regla general:
- Prefiera operaciones vectorizadas (`df * 2`, `df + valor`) siempre que la lógica lo permita.
- Reserve `apply()` para lógica compleja que no pueda expresarse de forma vectorizada.

* La siguiente celda compara el tiempo de `apply(lambda x: x * 2)` con la operación vectorizada equivalente `poblacion * 2`.

````python
import time

# Operación con apply()
inicio = time.perf_counter()
resultado_apply = poblacion.apply(lambda x: x * 2)
tiempo_apply = time.perf_counter() - inicio

# Operación vectorizada equivalente
inicio = time.perf_counter()
resultado_vectorizado = poblacion * 2
tiempo_vectorizado = time.perf_counter() - inicio

print(f"Tiempo con apply():        {tiempo_apply:.6f} s")
print(f"Tiempo con vectorización:  {tiempo_vectorizado:.6f} s")
print(f"\nAmbos producen el mismo resultado: {resultado_apply.equals(resultado_vectorizado)}")
````

### *Broadcasting*.

* La siguiente celda utilizará las propiedades de *broadcasting* para aplicar una función que suma diversos elementos a cada renglón del *dataframe* ```poblacion``` en el eje ```0```. 
* En vista de que el objeto ```[1, 2, 3, 4, 5, 6]``` tiene ```6``` elementos y el *dataframe* ```poblacion``` es de forma ```(6, 4)```, es posible realizar el *broadcasting*.

````python
poblacion
````

````python
poblacion.apply(lambda x: x + [1, 2, 3, 4, 5, 6])
````

* La siguiente celda utilizará las propiedades de *broadcasting* para aplicar una función que suma diversos elementos a cada renglón del *dataframe* ```poblacion``` en el eje ```1```. 
* En vista de que el objeto ```[1, 2, 3, 4]``` tiene ```4``` elementos y el *dataframe* ```poblacion``` es de forma ```(6, 4)```, es posible realizar el *broadcasting*.

````python
poblacion
````

````python
poblacion.apply(lambda x: x + [1, 2, 3, 4], axis=1)
````

* La siguiente celda aplicará la función con *broadcasting* sobre el eje ```0``` con un objeto de tamaño inadecuado. Se desencadenará una excepción del tipo ```ValueError```.

````python
poblacion.apply(lambda x: x + [1, 2, 3, 4])
````

### Aplicación de funciones de agregación.

*Numpy* cuenta con funciones de agregación capaces de realizar operaciones con la totalidad de los elementos de un arreglo, en vez de con cada uno de ellos. 

El método ```apply()``` es compatible con este tipo de funciones.

**Ejemplo:**

* La siguiente celda realizará una sumatoria de cada elemento en el eje ```0``` (renglones) del *dataframe* ```poblacion```, usando la función ```np.sum()``` y regresará una serie con los resultados.

````python
poblacion.apply(np.sum)
````

* La siguiente celda realizará una sumatoria de cada elemento en el eje ```1``` (columnas) del *dataframe* ```poblacion```, usando la función ```np.sum()``` y regresará una serie con los resultados.

````python
poblacion.apply(np.sum, axis=1)
````

* La siguiente celda realizará un promedio de cada elemento en el eje ```0``` (renglones) del *dataframe* ```poblacion```, usando la función ```np.mean()``` y regresará una serie con los resultados.

````python
poblacion.apply(np.mean)
````

* La siguiente celda realizará un promedio de cada elemento en el eje ```1``` (columnas) del *dataframe* ```poblacion```, usando la función ```np.mean()``` y regresará una serie con los resultados.

````python
poblacion.apply(np.mean, axis=1)
````

### Optimización en función de contexto de los datos.

El método ```pd.DataFrame.apply()``` permite identificar ciertos datos que podrían causar errores o excepciones y es capaz de utilizar funcione de *numpy* análogas que den un resultado en vez de una excepción.

**Ejemplo:**

* La función `np.mean()` regresa un valor `np.nan` cuando encuentra un valor `np.nan` en el arreglo que se le ingresa como argumento.

````python
arreglo = np.array([25, np.nan, 23, 67, 14, 12])
````

````python
arreglo
````

````python
np.mean(arreglo)
````

* La función ```np.nanmean()``` descarta los valores ```np.nan``` que se encuentren en el arreglo que se le ingresa como argumento y calcula el promedio con el resto de los elementos.

````python
np.nanmean(arreglo)
````

* La siguiente celda creará al *dataframe* ```poblacion_nan``` a partir del *dataframe* ```poblacion```, sustituyendo el valor de ```poblacion_nan.loc['jaguar', 'Norte_1']``` por ```np.nan```.

````python
poblacion_nan = poblacion.copy()
poblacion_nan.loc['jaguar', 'Norte_1'] = np.nan
poblacion_nan
````

* La siguiente celda usará la función ```np.mean()``` como argumento del método ```poblacion_nan.apply()```. El comportamiento es idéntico a usar ```np.nanmean()```.
* El resultado para la columna ```Norte_1``` es ```28.2``` en vez de ```np.nan```.

````python
poblacion_nan.apply(np.mean)
````

````python
poblacion_nan.apply(np.nanmean)
````

## El método ```df.transform()```.

Este método permite aplicar una o más funciones a los elementos de un *dataframe*, generando nuevas columnas con los resultados.

```
df.transform(<func_1>, <func_2>, <func_3> ..., axis=<eje>)
```

Donde:

* ```<func>``` es una función de *Python* o de *Numpy*.
* ```<eje>``` puede ser:
  * ```0``` para aplicar la función a los renglones. Este es el valor por defecto.
  * ```1``` para aplicar la función a las columnas.

**NOTA:** Este método no permite realizar operaciones de agregación.


La documentación del método ```df.transform()``` puede ser consultada en:

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.transform.html

**Ejemplo:**

* Se utilizará el *dataframe* ```poblacion``` definido previamente.

````python
poblacion
````

* La siguiente celda aplicará las funciones al *dataframe* ```poblacion```:
 * ```lambda x: x + [1, 2, 3, 4, 5, 6]```.
 * ```np.log```
 * ```np.sin```
* El dataframe resultante tendrá un subnivel debajo de cada columna de ```poblacion``` para en el que e creará una columna con el resultado de aplicar la función tomando a cada elemento como argumento.

````python
poblacion.transform([lambda x: x + [1, 2, 3, 4, 5, 6],
                     np.log,
                     np.sin])
````

* El método ```poblacion.transform()``` no es compatible con la función ```np.mean()```, por lo que se desencadenará una excepción ```ValueError```.

````python
poblacion.transform(np.mean)
````

* Sin embargo, es posible ingresar una función de agregación en otra función que no realice agregación por si misma.

````python
poblacion.transform(lambda x: x - x.mean())
````

## El método `Series.map()`.

El método `map()` aplica una función, un diccionario o una serie sobre cada elemento de una *serie*, devolviendo una nueva serie transformada. A diferencia de `apply()`, `map()` opera **exclusivamente sobre series** y está optimizado para transformaciones elemento a elemento.

```
<serie>.map(<funcion_o_dict_o_serie>, na_action=<None|'ignore'>)
```

Donde:

* `<serie>` es la serie sobre la que se aplicará la transformación.
* `<funcion_o_dict_o_serie>` puede ser una función, un diccionario o una serie.
* `na_action` controla cómo se manejan los valores `np.nan`.
* Si se pasa una **función**, se aplica a cada elemento de la serie.
* Si se pasa un **diccionario**, cada valor se sustituye por su correspondiente en el diccionario; los valores no encontrados se reemplazan por `NaN`.
* `na_action='ignore'` omite los valores `NaN` sin pasarlos a la función.

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.map.html

**Ejemplo:**

* La siguiente celda aplicará una función sobre la columna `'Norte_1'` para clasificar cada valor de población:

````python
poblacion['Norte_1'].map(lambda x: 'crítico' if x < 5 else ('bajo' if x < 15 else 'estable'))
````

* La siguiente celda usará un diccionario para mapear los nombres del índice a su clasificación taxonómica:

````python
taxonomia = {'lobo': 'carnívoro', 'jaguar': 'carnívoro', 'coyote': 'carnívoro',
             'halcón': 'ave rapaz', 'lechuza': 'ave rapaz', 'aguila': 'ave rapaz'}
pd.Series(poblacion.index).map(taxonomia)
````

## El método `df.pipe()`.

El método `pipe()` permite encadenar funciones sobre un *dataframe* o serie, pasando el objeto como primer argumento de la función. Facilita la escritura de **pipelines de transformación** legibles y modulares.

```
df.pipe(<funcion>, *args, **kwargs)
```

Donde:

* `<funcion>` es cualquier función cuyo primer parámetro acepte un *dataframe* o serie.
* `*args` y `**kwargs` son argumentos adicionales que se pasan a la función tras el *dataframe*.

`pipe()` es equivalente a `funcion(df, *args, **kwargs)`, pero permite escribir las operaciones en el orden en que se aplican, de izquierda a derecha.

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.pipe.html

**Ejemplo:**

* La siguiente celda definirá dos funciones de transformación y las encadenará con `pipe()`:

````python
def estandarizar(df):
    """Resta la media y divide entre la desviación estándar (z-score)."""
    return (df - df.mean()) / df.std()

def redondear(df, decimales=2):
    """Redondea a n decimales."""
    return df.round(decimales)

poblacion.pipe(estandarizar).pipe(redondear, decimales=3)
````

* Sin `pipe()`, el código anterior equivale a la siguiente expresión menos legible:

````python
redondear(estandarizar(poblacion), decimales=3)
````
