**Ejemplo:**

```python
import pandas as pd
```

```python
poblaciones = pd.DataFrame({'animal':('jaguar', 
                                    'ocelote', 
                                    'ajolote',
                                    'coyote',
                                    'jaguar', 
                                    'teporingo',
                                    'coyote',
                                    'lobo',
                                    'borrego cimarrón',
                                    'jaguar'),
                            'poblacion':(6,
                                         3,
                                         123,
                                         12,
                                         1023,
                                         6,
                                         20,
                                         200,
                                         2,
                                         15),
                            'region':('Sureste I',
                                     'Sureste I',
                                     'Centro I',
                                     'Centro I',
                                     'Sureste I',
                                     'Centro I',
                                     'Norte I',
                                     'Norte I',
                                     'Norte I',
                                     'Centro I')})
```

```python
poblaciones
```

```python
pd.melt(poblaciones, id_vars=["animal", "region"], value_vars=["poblacion"])
```
