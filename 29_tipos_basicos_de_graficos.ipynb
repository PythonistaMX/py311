# Tipos básicos de gráfico de *Matplotlib*.

En esta notebook se presentan los tipos de gráficos más comunes de *Matplotlib*. Cada tipo de gráfico está diseñado para un propósito específico y la elección adecuada depende de la naturaleza de los datos y del mensaje que se desea comunicar.

La documentación oficial ofrece una galería completa con ejemplos de código:

https://matplotlib.org/stable/plot_types/index.html

https://matplotlib.org/stable/gallery/

* La siguiente celda importa los módulos `numpy` y `matplotlib.pyplot` necesarios para la generación de datos y la creación de gráficos.

````python
import numpy as np
from matplotlib import pyplot as plt
````

## Preliminares.

A continuación se generan los datos aleatorios que se utilizarán como ejemplos a lo largo de esta notebook.

* La siguiente celda inicializa la semilla del generador de números aleatorios de `numpy` en `42` para garantizar la reproducibilidad.

````python
np.random.seed(42)
````

* La siguiente celda genera un arreglo llamado `normal` que contiene 500 muestras de una distribución normal con media 50 y desviación estándar 20.

````python
normal = np.random.normal(50, 20, 500)
````

* La siguiente celda muestra el contenido del arreglo `normal` recién creado.

````python
normal
````

* La siguiente celda genera un segundo arreglo llamado `normal2` con 500 muestras de una distribución normal con media 25 y desviación estándar 12.

````python
normal2 = np.random.normal(25, 12, 500)
````

* La siguiente celda genera un arreglo de 500 números enteros aleatorios entre 5 y 60, llamado `ancho`.

````python
ancho = np.random.randint(5, 60, 500)
````

* La siguiente celda define datos categóricos y valores numéricos para gráficos de barras, un rango lineal para gráficos de líneas y grupos de datos para diagramas de violín.

````python
# Datos categóricos para gráficas de barras y pastel
categorias = ['A', 'B', 'C', 'D', 'E']
valores_bar = np.array([35, 62, 48, 75, 41])

# Eje continuo para gráficas de líneas
x_lineal = np.linspace(0, 4 * np.pi, 200)

# Grupos para diagrama de violín
grupos_violin = [np.random.normal(mu, 10, 150) for mu in [30, 50, 70]]
````

## Histograma.

Un histograma agrupa los datos en intervalos (*bins*) y representa la frecuencia de cada intervalo como una barra vertical. Es el gráfico más habitual para explorar la distribución de una variable numérica continua.

Los parámetros más importantes son:

- ```bins```: número de intervalos o sus límites explícitos.
- ```histtype```: estilo del histograma (```'bar'```, ```'step'```, ```'stepfilled'```).
- ```orientation```: orientación de las barras (```'vertical'``` u ```'horizontal'```).

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html

**Ejemplos:**

* La siguiente celda genera un histograma simple a partir de los datos del arreglo `normal`.

````python
plt.hist(normal)
````

* La siguiente celda crea un histograma utilizando el estilo de línea 'step' en lugar de barras sólidas.

````python
plt.hist(normal, histtype='step')
````

* La siguiente celda genera un histograma con orientación horizontal.

````python
plt.hist(normal, orientation='horizontal')
````

* La siguiente celda crea un histograma especificando el color rojo para las barras.

````python
plt.hist(normal, color='red')
````

* La siguiente celda genera un histograma dividiendo los datos en 13 intervalos (binning).

````python
plt.hist(normal, bins=13)
````

* La siguiente celda muestra un histograma con la opción `stacked=True`.

````python
plt.hist(normal, stacked=True)
````

## Histograma 2D.

El histograma 2D extiende el concepto a dos dimensiones: divide el plano en una cuadrícula de celdas y cuenta cuántos puntos caen en cada una, representando la densidad con un mapa de color. Es útil para visualizar la distribución conjunta de dos variables numéricas continuas.

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist2d.html

**Ejemplo:**

* La siguiente celda genera un histograma bidimensional utilizando `normal` y `normal2` para visualizar la densidad conjunta.

````python
plt.hist2d(normal, normal2)
````

## Boxplot.

El diagrama de caja (*boxplot*) resume la distribución de los datos mediante cinco estadísticos: valor mínimo, primer cuartil (Q1), mediana, tercer cuartil (Q3) y valor máximo. La caja central abarca el rango intercuartílico (IQR) y los bigotes se extienden hasta 1.5 × IQR. Los puntos fuera de este rango se consideran *outliers* y se grafican individualmente.

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.boxplot.html

**Ejemplos:**

* La siguiente celda crea un diagrama de caja (boxplot) para visualizar la distribución estadística del arreglo `normal`.

````python
plt.boxplot(normal)
````

* La siguiente celda genera diagramas de caja para los arreglos `normal` y `normal2` en la misma figura.

````python
plt.boxplot((normal, normal2))
````

## Scatterplot.

El diagrama de dispersión (*scatterplot*) representa cada observación como un punto en el plano cartesiano. Es ideal para explorar la relación entre dos variables numéricas, detectar correlaciones, agrupaciones (*clusters*) o valores atípicos.

Los parámetros ```s``` y ```c``` permiten codificar una tercera variable en el tamaño y el color de cada punto, respectivamente.

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html

**Ejemplos:**

* La siguiente celda crea un gráfico de dispersión (scatterplot) mostrando la relación entre `normal` y `normal2`.

````python
plt.scatter(normal, normal2)
````

* La siguiente celda genera un gráfico de dispersión donde el tamaño y color de los puntos están determinados por el arreglo `ancho`.

````python
plt.scatter(normal, normal2, s=ancho, c=ancho)
````

## Gráfica de líneas.

La gráfica de líneas (*line plot*) conecta puntos de datos con segmentos de línea y es el tipo de gráfico más frecuente para representar series temporales o funciones continuas. El tercer argumento posicional ```fmt``` permite definir de forma compacta el color, el marcador y el estilo de línea con una cadena como ```'b-'``` (azul, línea continua) o ```'r--'``` (rojo, línea discontinua).

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

**Ejemplos:**

* La siguiente celda traza la función seno utilizando los valores de `x_lineal`.

````python
plt.plot(x_lineal, np.sin(x_lineal))
````

* La siguiente celda grafica las funciones seno y coseno con diferentes estilos de línea, colores y una leyenda explicativa.

````python
plt.plot(x_lineal, np.sin(x_lineal), 'b-', linewidth=2, label='seno')
plt.plot(x_lineal, np.cos(x_lineal), 'r--', linewidth=2, label='coseno')
plt.legend()
````

## Gráfica de barras.

La gráfica de barras (*bar chart*) representa variables categóricas con rectángulos cuya longitud es proporcional al valor de cada categoría. ```plt.bar()``` genera barras verticales y ```plt.barh()``` genera barras horizontales, siendo esta última preferible cuando las etiquetas de las categorías son largas.

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html

**Ejemplos:**

* La siguiente celda genera un gráfico de barras vertical utilizando las categorías y valores definidos anteriormente.

````python
plt.bar(categorias, valores_bar)
````

* La siguiente celda crea un gráfico de barras horizontal de color azul acero.

````python
plt.barh(categorias, valores_bar, color='steelblue')
````

## Gráfica de pastel.

La gráfica de pastel (*pie chart*) representa cada categoría como un sector circular proporcional a su valor. El argumento ```autopct``` formatea el porcentaje de cada sector; ```explode``` separa visualmente alguno de ellos; y ```startangle``` controla el ángulo de inicio del primer sector.

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.pie.html

**Ejemplos:**

* La siguiente celda genera un gráfico de pastel mostrando los porcentajes de cada categoría.

````python
plt.pie(valores_bar, labels=categorias, autopct='%1.1f%%')
````

* La siguiente celda crea un gráfico de pastel con una porción separada (exploded), sombra y un ángulo de inicio rotado.

````python
explode = (0.1, 0, 0, 0, 0)
plt.pie(valores_bar, labels=categorias, autopct='%1.1f%%',
        explode=explode, shadow=True, startangle=90)
````

## Diagrama de violín.

El diagrama de violín (*violin plot*) combina el diagrama de caja con una estimación de densidad de kernel (KDE) para mostrar la distribución completa de los datos. La anchura del perfil en cada punto refleja la densidad local: más ancho donde hay más observaciones.

Los parámetros ```showmeans``` y ```showmedians``` permiten añadir marcadores para la media y la mediana sobre el violín.

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.violinplot.html

**Ejemplos:**

* La siguiente celda genera un diagrama de violín para visualizar la distribución de los grupos de datos.

````python
plt.violinplot(grupos_violin)
````

* La siguiente celda muestra un diagrama de violín que incluye indicaciones para la media y la mediana.

````python
plt.violinplot(grupos_violin, showmeans=True, showmedians=True)
````

## Gráfica de área apilada.

La gráfica de área apilada (*stackplot*) rellena el área bajo múltiples series y las superpone visualmente. Es útil para mostrar cómo se distribuye un total entre varias categorías a lo largo de una variable continua, destacando tanto los valores individuales como el acumulado.

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.stackplot.html

**Ejemplos:**

* La siguiente celda genera un gráfico de área apilada (stackplot) con dos series de datos calculadas a partir de funciones trigonométricas.

````python
x_area = np.linspace(0, 10, 100)
y1 = np.abs(np.sin(x_area)) * 20 + 10
y2 = np.abs(np.cos(x_area)) * 15 + 8

plt.stackplot(x_area, y1, y2)
````

* La siguiente celda añade una tercera serie constante al gráfico de área apilada, incluyendo transparencia y una leyenda.

````python
x_area = np.linspace(0, 10, 100)
y1 = np.abs(np.sin(x_area)) * 20 + 10
y2 = np.abs(np.cos(x_area)) * 15 + 8
y3 = np.full_like(x_area, 12)

plt.stackplot(x_area, y1, y2, y3,
              labels=['Serie 1', 'Serie 2', 'Serie 3'], alpha=0.8)
plt.legend(loc='upper left')
````

## Gráfica de tallo.

La gráfica de tallo (*stem plot*) representa datos discretos mediante segmentos verticales que van desde una línea base hasta cada punto de datos, rematados con un marcador. Es el equivalente discreto de la gráfica de líneas y resulta especialmente útil para señales digitales o secuencias de conteo.

Los parámetros más relevantes son:

- `linefmt`: formato de los segmentos verticales (tallos).
- `markerfmt`: formato del marcador en la punta de cada tallo.
- `basefmt`: formato de la línea base horizontal.

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.stem.html

**Ejemplo:**

* La siguiente celda genera una gráfica de tallo de la función seno muestreada en 25 puntos.

````python
x_stem = np.linspace(0, 4 * np.pi, 25)

plt.stem(x_stem, np.sin(x_stem))
````

* La siguiente celda personaliza el color de los tallos con `linefmt`, el marcador con `markerfmt` y la línea base con `basefmt`.

````python
x_stem = np.linspace(0, 4 * np.pi, 25)

plt.stem(x_stem, np.sin(x_stem), linefmt='C1-', markerfmt='C1o', basefmt='k-')
````

## Barras de error.

La gráfica de barras de error (*errorbar plot*) extiende la gráfica de líneas añadiendo segmentos que representan la incertidumbre o variabilidad asociada a cada punto. Los parámetros `xerr` y `yerr` aceptan un escalar (error simétrico) o un arreglo de dos filas (errores asimétricos superior e inferior). El parámetro `capsize` controla el ancho de las tapas de los bigotes.

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.errorbar.html

**Ejemplos:**

* La siguiente celda genera una gráfica de barras de error básica con cinco puntos de medición y sus errores simétricos en el eje $y$.

````python
x_err = np.arange(1, 6)
y_err = np.array([2.1, 3.5, 2.8, 4.2, 3.9])
error = np.array([0.3, 0.5, 0.4, 0.6, 0.2])

plt.errorbar(x_err, y_err, yerr=error)
````

* La siguiente celda agrega formato con `fmt='o-'`, tapas a los bigotes con `capsize` y una leyenda.

````python
x_err = np.arange(1, 6)
y_err = np.array([2.1, 3.5, 2.8, 4.2, 3.9])
error = np.array([0.3, 0.5, 0.4, 0.6, 0.2])

plt.errorbar(x_err, y_err, yerr=error, fmt='o-', capsize=5,
             elinewidth=1.5, color='steelblue', label='medición')
plt.legend()
````

## Gráfica de contorno.

La gráfica de contorno (*contour plot*) visualiza funciones de dos variables $f(x, y)$ dibujando curvas de nivel (isolíneas) donde la función toma un valor constante. `plt.contour()` traza solo las líneas, mientras que `plt.contourf()` las rellena con color generando un mapa de calor suavizado.

Los datos de entrada deben ser matrices 2D generadas con `np.meshgrid()`.

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.contour.html

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.contourf.html

**Ejemplos:**

* La siguiente celda define una cuadrícula 2D con `np.meshgrid()` y traza las isolíneas de la función $f(x,y) = \sin(x)\cos(y)$ con `plt.contour()`.

````python
x_c = np.linspace(-3, 3, 100)
y_c = np.linspace(-3, 3, 100)
X_c, Y_c = np.meshgrid(x_c, y_c)
Z_c = np.sin(X_c) * np.cos(Y_c)

plt.contour(X_c, Y_c, Z_c)
````

* La siguiente celda utiliza `plt.contourf()` para rellenar las regiones entre isolíneas con color, añadiendo una barra de color con `plt.colorbar()`.

````python
plt.contourf(X_c, Y_c, Z_c, levels=20, cmap='RdYlBu')
plt.colorbar()
````
