# Tipos bĂĄsicos de grĂĄfico de *Matplotlib*.

En esta notebook se presentan los tipos de grĂĄficos mĂĄs comunes de *Matplotlib*. Cada tipo de grĂĄfico estĂĄ diseĂąado para un propĂłsito especĂ­fico y la elecciĂłn adecuada depende de la naturaleza de los datos y del mensaje que se desea comunicar.

La documentaciĂłn oficial ofrece una galerĂ­a completa con ejemplos de cĂłdigo:

https://matplotlib.org/stable/plot_types/index.html

https://matplotlib.org/stable/gallery/

La siguiente celda importa los mĂłdulos `numpy` y `matplotlib.pyplot` necesarios para la generaciĂłn de datos y la creaciĂłn de grĂĄficos.

````python
import numpy as np
from matplotlib import pyplot as plt
````

## Preliminares.

A continuaciĂłn se generan los datos aleatorios que se utilizarĂĄn como ejemplos a lo largo de esta notebook.

La siguiente celda inicializa la semilla del generador de nĂşmeros aleatorios de `numpy` en 42 para garantizar la reproducibilidad.

````python
np.random.seed(42)
````

La siguiente celda genera un arreglo llamado `normal` que contiene 500 muestras de una distribuciĂłn normal con media 50 y desviaciĂłn estĂĄndar 20.

````python
normal = np.random.normal(50, 20, 500)
````

La siguiente celda muestra el contenido del arreglo `normal` reciĂŠn creado.

````python
normal
````

La siguiente celda genera un segundo arreglo llamado `normal2` con 500 muestras de una distribuciĂłn normal con media 25 y desviaciĂłn estĂĄndar 12.

````python
normal2 = np.random.normal(25, 12, 500)
````

La siguiente celda genera un arreglo de 500 nĂşmeros enteros aleatorios entre 5 y 60, llamado `ancho`.

````python
ancho = np.random.randint(5, 60, 500)
````

La siguiente celda define datos categĂłricos y valores numĂŠricos para grĂĄficos de barras, un rango lineal para grĂĄficos de lĂ­neas y grupos de datos para diagramas de violĂ­n.

````python
# Datos categĂłricos para grĂĄficas de barras y pastel
categorias = ['A', 'B', 'C', 'D', 'E']
valores_bar = np.array([35, 62, 48, 75, 41])

# Eje continuo para grĂĄficas de lĂ­neas
x_lineal = np.linspace(0, 4 * np.pi, 200)

# Grupos para diagrama de violĂ­n
grupos_violin = [np.random.normal(mu, 10, 150) for mu in [30, 50, 70]]
````

## Histograma.

Un histograma agrupa los datos en intervalos (*bins*) y representa la frecuencia de cada intervalo como una barra vertical. Es el grĂĄfico mĂĄs habitual para explorar la distribuciĂłn de una variable numĂŠrica continua.

Los parĂĄmetros mĂĄs importantes son:

- ```bins```: nĂşmero de intervalos o sus lĂ­mites explĂ­citos.
- ```histtype```: estilo del histograma (```'bar'```, ```'step'```, ```'stepfilled'```).
- ```orientation```: orientaciĂłn de las barras (```'vertical'``` u ```'horizontal'```).

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html

La siguiente celda genera un histograma simple a partir de los datos del arreglo `normal`.

````python
plt.hist(normal)
````

La siguiente celda crea un histograma utilizando el estilo de lĂ­nea 'step' en lugar de barras sĂłlidas.

````python
plt.hist(normal, histtype='step')
````

La siguiente celda genera un histograma con orientaciĂłn horizontal.

````python
plt.hist(normal, orientation='horizontal')
````

La siguiente celda crea un histograma especificando el color rojo para las barras.

````python
plt.hist(normal, color='red')
````

La siguiente celda genera un histograma dividiendo los datos en 13 intervalos (binning).

````python
plt.hist(normal, bins=13)
````

La siguiente celda muestra un histograma con la opciĂłn `stacked=True`.

````python
plt.hist(normal, stacked=True)
````

## Histograma 2D.

El histograma 2D extiende el concepto a dos dimensiones: divide el plano en una cuadrĂ­cula de celdas y cuenta cuĂĄntos puntos caen en cada una, representando la densidad con un mapa de color. Es Ăştil para visualizar la distribuciĂłn conjunta de dos variables numĂŠricas continuas.

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist2d.html

La siguiente celda genera un histograma bidimensional utilizando `normal` y `normal2` para visualizar la densidad conjunta.

````python
plt.hist2d(normal, normal2)
````

## Boxplot.

El diagrama de caja (*boxplot*) resume la distribuciĂłn de los datos mediante cinco estadĂ­sticos: valor mĂ­nimo, primer cuartil (Q1), mediana, tercer cuartil (Q3) y valor mĂĄximo. La caja central abarca el rango intercuartĂ­lico (IQR) y los bigotes se extienden hasta 1.5 Ă IQR. Los puntos fuera de este rango se consideran *outliers* y se grafican individualmente.

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.boxplot.html

La siguiente celda crea un diagrama de caja (boxplot) para visualizar la distribuciĂłn estadĂ­stica del arreglo `normal`.

````python
plt.boxplot(normal)
````

La siguiente celda genera diagramas de caja para los arreglos `normal` y `normal2` en la misma figura.

````python
plt.boxplot((normal, normal2))
````

## Scatterplot.

El diagrama de dispersiĂłn (*scatterplot*) representa cada observaciĂłn como un punto en el plano cartesiano. Es ideal para explorar la relaciĂłn entre dos variables numĂŠricas, detectar correlaciones, agrupaciones (*clusters*) o valores atĂ­picos.

Los parĂĄmetros ```s``` y ```c``` permiten codificar una tercera variable en el tamaĂąo y el color de cada punto, respectivamente.

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html

La siguiente celda crea un grĂĄfico de dispersiĂłn (scatterplot) mostrando la relaciĂłn entre `normal` y `normal2`.

````python
plt.scatter(normal, normal2)
````

La siguiente celda genera un grĂĄfico de dispersiĂłn donde el tamaĂąo y color de los puntos estĂĄn determinados por el arreglo `ancho`.

````python
plt.scatter(normal, normal2, s=ancho, c=ancho)
````

## GrĂĄfica de lĂ­neas.

La grĂĄfica de lĂ­neas (*line plot*) conecta puntos de datos con segmentos de lĂ­nea y es el tipo de grĂĄfico mĂĄs frecuente para representar series temporales o funciones continuas. El tercer argumento posicional ```fmt``` permite definir de forma compacta el color, el marcador y el estilo de lĂ­nea con una cadena como ```'b-'``` (azul, lĂ­nea continua) o ```'r--'``` (rojo, lĂ­nea discontinua).

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

La siguiente celda traza la funciĂłn seno utilizando los valores de `x_lineal`.

````python
plt.plot(x_lineal, np.sin(x_lineal))
````

La siguiente celda grafica las funciones seno y coseno con diferentes estilos de lĂ­nea, colores y una leyenda explicativa.

````python
plt.plot(x_lineal, np.sin(x_lineal), 'b-', linewidth=2, label='seno')
plt.plot(x_lineal, np.cos(x_lineal), 'r--', linewidth=2, label='coseno')
plt.legend()
````

## GrĂĄfica de barras.

La grĂĄfica de barras (*bar chart*) representa variables categĂłricas con rectĂĄngulos cuya longitud es proporcional al valor de cada categorĂ­a. ```plt.bar()``` genera barras verticales y ```plt.barh()``` genera barras horizontales, siendo esta Ăşltima preferible cuando las etiquetas de las categorĂ­as son largas.

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html

La siguiente celda genera un grĂĄfico de barras vertical utilizando las categorĂ­as y valores definidos anteriormente.

````python
plt.bar(categorias, valores_bar)
````

La siguiente celda crea un grĂĄfico de barras horizontal de color azul acero.

````python
plt.barh(categorias, valores_bar, color='steelblue')
````

## GrĂĄfica de pastel.

La grĂĄfica de pastel (*pie chart*) representa cada categorĂ­a como un sector circular proporcional a su valor. El argumento ```autopct``` formatea el porcentaje de cada sector; ```explode``` separa visualmente alguno de ellos; y ```startangle``` controla el ĂĄngulo de inicio del primer sector.

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.pie.html

La siguiente celda genera un grĂĄfico de pastel mostrando los porcentajes de cada categorĂ­a.

````python
plt.pie(valores_bar, labels=categorias, autopct='%1.1f%%')
````

La siguiente celda crea un grĂĄfico de pastel con una porciĂłn separada (exploded), sombra y un ĂĄngulo de inicio rotado.

````python
explode = (0.1, 0, 0, 0, 0)
plt.pie(valores_bar, labels=categorias, autopct='%1.1f%%',
        explode=explode, shadow=True, startangle=90)
````

## Diagrama de violĂ­n.

El diagrama de violĂ­n (*violin plot*) combina el diagrama de caja con una estimaciĂłn de densidad de kernel (KDE) para mostrar la distribuciĂłn completa de los datos. La anchura del perfil en cada punto refleja la densidad local: mĂĄs ancho donde hay mĂĄs observaciones.

Los parĂĄmetros ```showmeans``` y ```showmedians``` permiten aĂąadir marcadores para la media y la mediana sobre el violĂ­n.

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.violinplot.html

La siguiente celda genera un diagrama de violĂ­n para visualizar la distribuciĂłn de los grupos de datos.

````python
plt.violinplot(grupos_violin)
````

La siguiente celda muestra un diagrama de violĂ­n que incluye indicaciones para la media y la mediana.

````python
plt.violinplot(grupos_violin, showmeans=True, showmedians=True)
````

## GrĂĄfica de ĂĄrea apilada.

La grĂĄfica de ĂĄrea apilada (*stackplot*) rellena el ĂĄrea bajo mĂşltiples series y las superpone visualmente. Es Ăştil para mostrar cĂłmo se distribuye un total entre varias categorĂ­as a lo largo de una variable continua, destacando tanto los valores individuales como el acumulado.

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.stackplot.html

La siguiente celda genera un grĂĄfico de ĂĄrea apilada (stackplot) con dos series de datos calculadas a partir de funciones trigonomĂŠtricas.

````python
x_area = np.linspace(0, 10, 100)
y1 = np.abs(np.sin(x_area)) * 20 + 10
y2 = np.abs(np.cos(x_area)) * 15 + 8

plt.stackplot(x_area, y1, y2)
````

La siguiente celda aĂąade una tercera serie constante al grĂĄfico de ĂĄrea apilada, incluyendo transparencia y una leyenda.

````python
x_area = np.linspace(0, 10, 100)
y1 = np.abs(np.sin(x_area)) * 20 + 10
y2 = np.abs(np.cos(x_area)) * 15 + 8
y3 = np.full_like(x_area, 12)

plt.stackplot(x_area, y1, y2, y3,
              labels=['Serie 1', 'Serie 2', 'Serie 3'], alpha=0.8)
plt.legend(loc='upper left')
````
