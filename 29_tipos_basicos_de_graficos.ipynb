# Tipos básicos de gráfico de *Matplotlib*.

En esta notebook se presentan los tipos de gráficos más comunes de *Matplotlib*. Cada tipo de gráfico está diseñado para un propósito específico y la elección adecuada depende de la naturaleza de los datos y del mensaje que se desea comunicar.

La documentación oficial ofrece una galería completa con ejemplos de código:

https://matplotlib.org/stable/plot_types/index.html

https://matplotlib.org/stable/gallery/

````python
import numpy as np
from matplotlib import pyplot as plt
````

## Preliminares.

A continuación se generan los datos aleatorios que se utilizarán como ejemplos a lo largo de esta notebook.

````python
np.random.seed(42)
````

````python
normal = np.random.normal(50, 20, 500)
````

````python
normal
````

````python
normal2 = np.random.normal(25, 12, 500)
````

````python
ancho = np.random.randint(5, 60, 500)
````

````python
# Datos categóricos para gráficas de barras y pastel
categorias = ['A', 'B', 'C', 'D', 'E']
valores_bar = np.array([35, 62, 48, 75, 41])

# Eje continuo para gráficas de líneas
x_lineal = np.linspace(0, 4 * np.pi, 200)

# Grupos para diagrama de violín
grupos_violin = [np.random.normal(mu, 10, 150) for mu in [30, 50, 70]]
````

## Histograma.

Un histograma agrupa los datos en intervalos (*bins*) y representa la frecuencia de cada intervalo como una barra vertical. Es el gráfico más habitual para explorar la distribución de una variable numérica continua.

Los parámetros más importantes son:

- ```bins```: número de intervalos o sus límites explícitos.
- ```histtype```: estilo del histograma (```'bar'```, ```'step'```, ```'stepfilled'```).
- ```orientation```: orientación de las barras (```'vertical'``` u ```'horizontal'```).

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html

````python
plt.hist(normal)
````

````python
plt.hist(normal, histtype='step')
````

````python
plt.hist(normal, orientation='horizontal')
````

````python
plt.hist(normal, color='red')
````

````python
plt.hist(normal, bins=13)
````

````python
plt.hist(normal, stacked=True)
````

## Histograma 2D.

El histograma 2D extiende el concepto a dos dimensiones: divide el plano en una cuadrícula de celdas y cuenta cuántos puntos caen en cada una, representando la densidad con un mapa de color. Es útil para visualizar la distribución conjunta de dos variables numéricas continuas.

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist2d.html

````python
plt.hist2d(normal, normal2)
````

## Boxplot.

El diagrama de caja (*boxplot*) resume la distribución de los datos mediante cinco estadísticos: valor mínimo, primer cuartil (Q1), mediana, tercer cuartil (Q3) y valor máximo. La caja central abarca el rango intercuartílico (IQR) y los bigotes se extienden hasta 1.5 × IQR. Los puntos fuera de este rango se consideran *outliers* y se grafican individualmente.

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.boxplot.html

````python
plt.boxplot(normal)
````

````python
plt.boxplot((normal, normal2))
````

## Scatterplot.

El diagrama de dispersión (*scatterplot*) representa cada observación como un punto en el plano cartesiano. Es ideal para explorar la relación entre dos variables numéricas, detectar correlaciones, agrupaciones (*clusters*) o valores atípicos.

Los parámetros ```s``` y ```c``` permiten codificar una tercera variable en el tamaño y el color de cada punto, respectivamente.

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html

````python
plt.scatter(normal, normal2)
````

````python
plt.scatter(normal, normal2, s=ancho, c=ancho)
````

## Gráfica de líneas.

La gráfica de líneas (*line plot*) conecta puntos de datos con segmentos de línea y es el tipo de gráfico más frecuente para representar series temporales o funciones continuas. El tercer argumento posicional ```fmt``` permite definir de forma compacta el color, el marcador y el estilo de línea con una cadena como ```'b-'``` (azul, línea continua) o ```'r--'``` (rojo, línea discontinua).

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

````python
plt.plot(x_lineal, np.sin(x_lineal))
````

````python
plt.plot(x_lineal, np.sin(x_lineal), 'b-', linewidth=2, label='seno')
plt.plot(x_lineal, np.cos(x_lineal), 'r--', linewidth=2, label='coseno')
plt.legend()
````

## Gráfica de barras.

La gráfica de barras (*bar chart*) representa variables categóricas con rectángulos cuya longitud es proporcional al valor de cada categoría. ```plt.bar()``` genera barras verticales y ```plt.barh()``` genera barras horizontales, siendo esta última preferible cuando las etiquetas de las categorías son largas.

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html

````python
plt.bar(categorias, valores_bar)
````

````python
plt.barh(categorias, valores_bar, color='steelblue')
````

## Gráfica de pastel.

La gráfica de pastel (*pie chart*) representa cada categoría como un sector circular proporcional a su valor. El argumento ```autopct``` formatea el porcentaje de cada sector; ```explode``` separa visualmente alguno de ellos; y ```startangle``` controla el ángulo de inicio del primer sector.

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.pie.html

````python
plt.pie(valores_bar, labels=categorias, autopct='%1.1f%%')
````

````python
explode = (0.1, 0, 0, 0, 0)
plt.pie(valores_bar, labels=categorias, autopct='%1.1f%%',
        explode=explode, shadow=True, startangle=90)
````

## Diagrama de violín.

El diagrama de violín (*violin plot*) combina el diagrama de caja con una estimación de densidad de kernel (KDE) para mostrar la distribución completa de los datos. La anchura del perfil en cada punto refleja la densidad local: más ancho donde hay más observaciones.

Los parámetros ```showmeans``` y ```showmedians``` permiten añadir marcadores para la media y la mediana sobre el violín.

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.violinplot.html

````python
plt.violinplot(grupos_violin)
````

````python
plt.violinplot(grupos_violin, showmeans=True, showmedians=True)
````

## Gráfica de área apilada.

La gráfica de área apilada (*stackplot*) rellena el área bajo múltiples series y las superpone visualmente. Es útil para mostrar cómo se distribuye un total entre varias categorías a lo largo de una variable continua, destacando tanto los valores individuales como el acumulado.

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.stackplot.html

````python
x_area = np.linspace(0, 10, 100)
y1 = np.abs(np.sin(x_area)) * 20 + 10
y2 = np.abs(np.cos(x_area)) * 15 + 8

plt.stackplot(x_area, y1, y2)
````

````python
x_area = np.linspace(0, 10, 100)
y1 = np.abs(np.sin(x_area)) * 20 + 10
y2 = np.abs(np.cos(x_area)) * 15 + 8
y3 = np.full_like(x_area, 12)

plt.stackplot(x_area, y1, y2, y3,
              labels=['Serie 1', 'Serie 2', 'Serie 3'], alpha=0.8)
plt.legend(loc='upper left')
````
