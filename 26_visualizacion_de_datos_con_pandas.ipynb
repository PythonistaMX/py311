# Visualización de datos con *Pandas*.

````python
import pandas as pd
````

## El atributo ```plot``` de los objetos `DataFrame` y `Series`.

Este atributo contiene una colección de métodos que permiten deplegar gráficos descriptivos básicos basados en *Matplotlib*.

Algunos de los métodos más comunes son:

* ```pd.plot.line()```: para gráficos de líneas.
* ```pd.plot.bar()```: para gráficos de barras.
* ```pd.plot.hist()```: para histogramas.
* ```pd.plot.scatter()```: para gráficos de dispersión.
* ```pd.plot.box()```: para diagramas de caja.
* ```pd.plot.pie()```: para gráficos de pastel.

### Creación de graficos con `plot()`.

Para crear un gráfico con el método ```plot()```, se puede seguir la siguiente sintaxis:

```
<obj>.plot(kind='tipo_de_grafico', x='columna_x', y='columna_y', title='Título del gráfico', xlabel='Etiqueta eje x', ylabel='Etiqueta eje y')
```

Donde:
* ```<obj>```: puede ser un objeto `DataFrame` o `Series` de *Pandas*.
* ```kind```: especifica el tipo de gráfico que se desea crear (por ejemplo, 'line', 'bar', 'hist', 'scatter', 'box').
* ```x```: indica la columna del DataFrame que se utilizará para el eje x.
* ```y```: indica la columna del DataFrame que se utilizará para el eje y.
* ```title```: establece el título del gráfico.
* ```xlabel```: establece la etiqueta del eje x.
* ```ylabel```: establece la etiqueta del eje y.

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.html


## Ejemplo de uso.

* El archivo ```data/datos_inegi.csv``` contiene los datos de crecimiento económico trimestral y anual de las Entidades Federativas de la República Mexicana desde *2003* hasta *2024*.

* La siguiente celda creará al *dataframe* ```datos``` a partir de cargar el archivo ```data/datos_inegi.csv```.

````python
datos = pd.read_csv('data/datos_inegi.csv')
````

````python
datos
````

````python
datos.info()
````

### 1. Adecuación del dataframe.

* La siguiente celda asignará a la columna `'Año'` como el indice del *dataframe* ```datos``` y le asignará un nombre `'Entidad'` al índice.

````python
datos.set_index('Año', inplace=True)
datos.index.name = 'Entidad'
datos
````

* La siguiente celda eliminará las filas con los índices  `'Entidades'` y `'Periodo'` del *dataframe* ```datos```.


````python
datos.drop(['Entidades', 'Periodo'], inplace=True)
datos
````

* Se creará un nuevo multiíndice  de nombre `'columnas'`.
    * El primer nivel del multiíndice se llamará `'Año'` y contendrá los años de 2003 a 2024.
    * El segundo nivel del multiíndice se llamará `'Período'` y contendrá los períodos trimestrales, de los primeros 6 meses, de los primeros 9 meses y anuales.

````python
columnas = pd.MultiIndex.from_product([(x for x in range(2003, 2025)),
                                       ('T1', 'T2', 'T3', 'T4', 'Seis meses', 'Nueve meses', 'Anual')],
                                      names=['Año', 'Período'])

columnas
````

* La siguiente celda asignará el contenido de `columnas` al índice de las columnas del *dataframe* ```datos```.

````python
datos.columns = columnas
datos
````

* La siguiente celda convertirá los datos del *dataframe* ```datos``` a tipo numérico, para que puedan ser analizados y graficados.

````python
datos = datos.astype(float)
````

### 2. Selección de los datos.

Ahora es posible poder manipular y visualizar los datos del *dataframe* ```datos```.

Para este caso se utilizarán los datos anuales de 2004 en todas las Entidades Federativas.

* Los datos que nos interesan se encuentran en la columna con el multiíndice `(2004, 'Anual')`.

````python
datos[2004]['Anual']
````

* La siguiente celda creará la serie ```historico_2004``` a partir de la columna con el multiíndice `(2004, 'Anual')` del *dataframe* ```datos```.

````python
historico_2004 = datos[2004]['Anual'][1:]
````

````python
historico_2004
````

### 3. Visualización de los datos.

A continuación se muestran algunos ejemplos de visualización de los datos de la serie ```historico_2004``` por medio de los métodos del atributo ```plot```.

Los métodos de visualización de datos del atributo ```plot``` permiten personalizar los gráficos con diferentes opciones, como el título, las etiquetas de los ejes, el tamaño de la figura, entre otras, aprovechando los parámetros de los métodos de *Matplotlib*.

En este caso, se utilizarán los parámetros por defecto para mostrar la simplicidad de uso de los métodos del atributo ```plot```.

Los parámetros de *Matplotlib* se estudiarán en capítulos posteriores, pero la documentación oficial está disponible en:

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

* La siguiente celda mostrará un gráfico de líneas de la serie ```historico_2004``` utilizando el método ```plot.line()```, el cual es el método por defecto para crear gráficos de líneas en *Pandas*.

````python
historico_2004.plot()
````

* La misma gráfica de líneas se puede crear utilizando el método ```plot.line()``` sin especificar el tipo de gráfico, ya que el método por defecto es el de líneas.

````python
historico_2004.plot.line()
````

* La siguiente celda desplegará un histograma a partir del método ```plot.hist()``` de la serie ```historico_2004```.

````python
historico_2004.plot.hist()
````

* La siguiente celda desplegará un gráfico de pastel (*pie chart*) a partir del método ```plot.pie()``` de la serie ```historico_2004```.

````python
historico_2004.plot.pie()
````

* La siguiente celda desplegará un gráfico de área a partir del método ```plot.area()``` de la serie ```historico_2004```.

````python
historico_2004.plot.area()
````

* La siguiente celda desplegará un gráfico de barras a partir del método ```plot.bar()``` de la serie ```historico_2004```.

````python
historico_2004.plot.bar()
````

* La siguiente celda desplegará un gráfico de barras horizontales a partir del método ```plot.barh()``` de la serie ```historico_2004```.

````python
historico_2004.plot.barh()
````

* La siguiente celda utilizará más de un método de visualización para mostrar la serie ```historico_2004``` en un gráfico de líneas y en un gráfico de barras, utilizando el método ```plot()``` con el parámetro ```kind``` para especificar el tipo de gráfico.
* Además se ingresarán algunos parámetros de *Matplotlib* para personalizar el gráfico, como el título, las etiquetas de los ejes y el tamaño de la figura.

````python
historico_2004.plot(kind='line',
                    x='Entidad', 
                    y='Anual',
                    xlabel='Entidad Federativa',
                    ylabel='Ponderación anual con base en 2018',
                    title='Valor anual por entidad',
                    legend=True,
                    grid=True,
                    color='darkblue',
                    figsize=(10, 6))
historico_2004.plot(kind='bar',
                    x='Entidad', 
                    y='Anual',
                    color='orange')
````

* La siguiente celda extraerá la evolución del crecimiento desestacionalizado de todo el país, correspondiente al índice `'Estados Unidos Mexicanos'`. Dicha serie se almacenará en la variable ```evolucion```.

````python
datos.filter(like='Anual')
````

````python
datos.filter(like='Anual').droplevel(1, axis=1)
````

````python
datos.filter(like='Anual').droplevel(1, axis=1).transpose()
````

````python
datos.filter(like='Anual').droplevel(1, axis=1).transpose()["Estados Unidos Mexicanos"]
````

````python
evolucion = (datos.
 filter(like='Anual').
 droplevel(1, axis=1).
transpose())['Estados Unidos Mexicanos']
evolucion
````

* La siguiente celda desplegará un gráfico de líneas de la serie ```evolucion``` utilizando el método ```plot()```.

````python
evolucion.plot(kind='line',
                xlabel='Año',
                ylabel='Valor',
                title='Evolución anual de México',
                legend=False,
                grid=True,
                color='green',
                figsize=(10, 6))
````
