# Introducción a *Plotnine*.

*Plotnine* es un proyecto que pretende implementar en *Python* las funcionalidades de [*ggplot2*](https://ggplot2.tidyverse.org/), la popular herramienta de visualización de datos para *R* y la [gramática de gráficas por capas](http://vita.had.co.nz/papers/layered-grammar.html); ambas creadas por [*Hadley Wickham*](https://hadley.nz/) y basadas en la gramática de gráficas de [*Leland Wilkinson*](https://en.wikipedia.org/wiki/Leland_Wilkinson).

Aún cuando muchas de las funcionalidades de *ggplot2* ya han sido portadas, aún quedan muchas que se encuentran pendientes.

## Gramática de capas de un gráfico.

La gramática de capas define una estructura de componentes independientes que, combinados, conforman un gráfico. La implementación de referencia es [**ggplot2**](https://ggplot2.tidyverse.org/) para *R*, considerada la biblioteca de visualización más avanzada y expresiva disponible actualmente. *Plotnine* es su equivalente directo en *Python* y comparte la misma sintaxis.

| Componente | Función en *Plotnine* | Descripción |
|---|---|---|
| **Datos** | `ggplot(data=…)` | Conjunto de datos que se desea visualizar |
| **Mapeo estético** | `aes(…)` | Relación entre variables y propiedades visuales: ejes, color, tamaño, forma… |
| **Objetos geométricos** | `geom_*()` | Representación visual: puntos, líneas, barras, áreas… |
| **Transformaciones estadísticas** | `stat_*()` | Cálculos o resúmenes aplicados a los datos antes de graficarlos |
| **Escalas** | `scale_*()` | Ajuste de la representación de cada variable (continua, discreta, logarítmica…) |
| **Facetas** | `facet_*()` | División del gráfico en subpaneles según una o más variables |
| **Sistema de coordenadas** | `coord_*()` | Espacio en el que se dibuja el gráfico (cartesiano, polar, etc.) |
| **Temas** | `theme_*()` | Apariencia general del gráfico: fuente, fondo, márgenes… |

## Composición por capas.

En *Plotnine*, un gráfico se construye sumando componentes con el operador `+`. Cada `+` añade una nueva capa o modifica la presentación del gráfico anterior:

```
ggplot(data=<datos>, mapping=aes(<estéticas>)) \
    + <geom_*>() \
    + <scale_*>() \
    + <facet_*>() \
    + <theme_*>()
```

El orden de las capas geométricas es significativo: las capas posteriores se dibujan sobre las anteriores. Las escalas, facetas y temas pueden colocarse en cualquier posición.

## La función `aes()`.

La función `aes()` (*aesthetics*) define el **mapeo estético**: la relación entre las columnas del *dataframe* y las propiedades visuales del gráfico. Se puede especificar a nivel global en `ggplot()` o a nivel local en cada `geom_*()`, en cuyo caso solo afecta a esa capa.

Los mapeos estéticos más comunes son:

| Argumento | Descripción |
|---|---|
| `x` | Variable asignada al eje horizontal |
| `y` | Variable asignada al eje vertical |
| `color` | Color del borde o de la línea |
| `fill` | Color de relleno |
| `size` | Tamaño del elemento |
| `shape` | Forma del punto |
| `alpha` | Transparencia (0 = transparente, 1 = opaco) |
| `linetype` | Tipo de línea (`"solid"`, `"dashed"`, `"dotted"`…) |
| `group` | Agrupación sin mapeo visual explícito |

### La función `plotnine.mapping.aes()`.

```
aes(x=<col_x>, y=<col_y>, color=<col_color>, fill=<col_fill>, <kwargs>)
```

Donde cada argumento es el nombre de una columna del *dataframe* expresado como cadena de texto, o una expresión estadística. Los mapeos definidos en `ggplot()` se heredan por todas las geometrías; los definidos en `geom_*()` son exclusivos de esa capa.

https://plotnine.readthedocs.io/en/stable/generated/plotnine.mapping.aes.html

## Funciones de geometría.

Las funciones `geom_*()` definen cómo se representan visualmente los datos. Cada geometría requiere ciertos mapeos estéticos obligatorios y admite otros opcionales. Se añaden al objeto `ggplot()` mediante el operador `+`.

| Función | Estéticas principales | Uso típico |
|---|---|---|
| `geom_point()` | `x`, `y` | Diagrama de dispersión |
| `geom_line()` | `x`, `y` | Series temporales, tendencias |
| `geom_smooth()` | `x`, `y` | Curva de ajuste con banda de confianza |
| `geom_bar()` | `x` | Conteo de categorías |
| `geom_col()` | `x`, `y` | Barras con valores preagregados |
| `geom_histogram()` | `x` | Distribución de una variable continua |
| `geom_density()` | `x` | Estimación de densidad de kernel |
| `geom_boxplot()` | `x`, `y` | Distribución por categoría |
| `geom_violin()` | `x`, `y` | Distribución por categoría con forma de densidad |
| `geom_area()` | `x`, `y` | Área bajo una curva |
| `geom_tile()` | `x`, `y`, `fill` | Mapa de calor |
| `geom_text()` | `x`, `y`, `label` | Anotaciones de texto |

https://plotnine.readthedocs.io/en/stable/api.html#geoms

Todas las funciones `geom_*()` aceptan el parámetro `stat=` que permite aplicar una transformación estadística sobre los datos antes de graficarlos. El valor por defecto es `stat="identity"` (sin transformación). También aceptan `position=` para controlar el posicionamiento de los elementos (`"identity"`, `"dodge"`, `"stack"`, `"fill"`).

## Funciones de temas.

Los temas controlan la apariencia general del gráfico: fondo, tipografía, cuadrícula, márgenes y demás elementos no relacionados con los datos. Se aplican con el operador `+`. *Plotnine* incluye los mismos temas que *ggplot2*:

| Función | Descripción |
|---|---|
| `theme_gray()` | Tema predeterminado: fondo gris con cuadrícula blanca |
| `theme_bw()` | Fondo blanco con cuadrícula gris |
| `theme_minimal()` | Mínimos elementos decorativos |
| `theme_classic()` | Sin cuadrícula, con líneas de eje marcadas |
| `theme_void()` | Sin ejes ni fondo (útil para mapas) |
| `theme_dark()` | Fondo oscuro |
| `theme_538()` | Estilo del portal *FiveThirtyEight* |
| `theme_tufte()` | Estilo minimalista de *Edward Tufte* |
| `theme_xkcd()` | Estilo de trazo a mano (*xkcd*) |

Además, la función `theme()` permite sobreescribir elementos individuales como la rotación de etiquetas de ejes (`axis_text_x=element_text(rotation=45)`) o el tamaño de la leyenda.

https://plotnine.readthedocs.io/en/stable/api.html#themes

## Ejemplos.

### Dataframe ilustrativo.

Los ejemplos de histograma y columnas usarán datos generados aleatoriamente con *NumPy*. Para garantizar reproducibilidad, la siguiente celda fijará la semilla del generador aleatorio.

````python
np.random.seed(45)
````

* La siguiente celda creará el *dataframe* `arreglo_base` con 1000 observaciones de una distribución normal (media=12, desviación estándar=25), que servirá como datos de ejemplo para los gráficos de histograma.

````python
arreglo_base = pd.DataFrame(np.random.normal(12, 25, 1000),
                            columns=pd.Index(['observaciones']))
arreglo_base
````

### Ejemplo de histograma.

* La siguiente celda creará un objeto `ggplot()` vacío a partir de `arreglo_base`. Sin capas geométricas, el resultado es un lienzo en blanco que muestra el sistema de coordenadas pero sin datos dibujados.

````python
ggplot(data=arreglo_base)
````

* La siguiente celda agregará la geometría `geom_histogram()` al objeto `ggplot()` para construir el histograma completo con los parámetros por defecto.

````python
ggplot(data=arreglo_base, mapping=aes(x='observaciones')) + geom_histogram()
````

* La siguiente celda personalizará el histograma ajustando el número de `bins` y los colores de relleno y borde con los parámetros `fill` y `color`.

````python
(ggplot(data=arreglo_base, mapping=aes(x='observaciones')) + 
geom_histogram(bins=10, fill='yellow', color="orange"))
````

* La siguiente celda calculará las frecuencias del histograma con `np.histogram()` y construirá el *dataframe* `histograma` con columnas `frecuencias` y `rangos`, que se usará en el ejemplo de columnas.

````python
histograma = pd.DataFrame(np.histogram(arreglo_base, bins=13)).T
histograma.columns = pd.Index(['frecuencias','rangos'])
histograma
````

### Ejemplo de columnas.

* La siguiente celda crea un gráfico de columnas con `geom_col()` a partir del *dataframe* `histograma`, asignando el color de relleno al valor del rango de cada bin.

````python
ggplot(histograma, aes(x='rangos', y='frecuencias', fill='rangos')) + geom_col()
````

### Ejemplo de líneas con datos de ventas mensuales.

* La siguiente celda generará el *dataframe* `ventas_mensuales` con ventas ficticias mensuales para las sucursales `CDMX`, `MTY` y `GDL` durante 48 meses (2020–2023), más el total `Nacional`.

````python
np.random.seed(42)
fechas = pd.date_range('2020-01-01', periods=48, freq='ME')
ventas_mensuales = pd.DataFrame({
    'CDMX': np.random.normal(80_000, 15_000, 48).clip(20_000).astype(int),
    'MTY':  np.random.normal(55_000, 10_000, 48).clip(10_000).astype(int),
    'GDL':  np.random.normal(45_000,  8_000, 48).clip(10_000).astype(int),
}, index=fechas)
ventas_mensuales['Nacional'] = ventas_mensuales.sum(axis=1)
ventas_mensuales.index.name = 'fecha'
ventas_mensuales
````

* La siguiente celda traza la evolución de las ventas nacionales con `geom_line()`.

````python
(ggplot(ventas_mensuales, aes(x=ventas_mensuales.index, y='Nacional'))
+ geom_line())
````

* La siguiente celda añade una curva de ajuste suavizado (`geom_smooth()`) sobre la línea de ventas con los parámetros por defecto.

````python
(ggplot(ventas_mensuales, aes(x=ventas_mensuales.index, y='Nacional'))
+ geom_line()
+ geom_smooth())
````

* La siguiente celda ajusta el parámetro `span` de `geom_smooth()` para obtener una curva de suavizado más local y sensible a las variaciones de corto plazo.

````python
(ggplot(ventas_mensuales, aes(x=ventas_mensuales.index, y='Nacional'))
+ geom_line()
+ geom_smooth(span=0.3, color='red'))
````

* La siguiente celda aplica el tema `theme_xkcd()` para un estilo de trazo a mano y rota las etiquetas del eje $x$ 90° para mayor legibilidad.

````python
(ggplot(ventas_mensuales, aes(x=ventas_mensuales.index, y='Nacional'))
 + geom_line()
 + geom_smooth(span=0.3, color='blue')
 + theme_xkcd()
 + theme(axis_text_x=element_text(rotation=90, hjust=0.5)))
````

* La siguiente celda aplica el tema `theme_tufte()` para un estilo minimalista, con la línea de ventas en rojo y el suavizado en azul.

````python
(ggplot(ventas_mensuales, aes(x=ventas_mensuales.index, y='Nacional'))
 + geom_line(color='red')
 + geom_smooth(span=0.3, color='blue')
 + theme_tufte()
 + theme(axis_text_x=element_text(rotation=45, hjust=1)))
````

### Comparación de ventas por sucursal en un mes.

* La siguiente celda seleccionará las ventas del mes de enero de 2021 por sucursal para construir el *dataframe* `data`, que se utilizará como base del gráfico de columnas.

````python
data = ventas_mensuales.drop('Nacional', axis=1).loc['2021-01-31'].to_frame()
data.columns = pd.Index(['Ventas'])
data.index.name = 'Sucursal'
data
````

* La siguiente celda crea un gráfico de columnas con `geom_col()` de las ventas por sucursal en enero de 2021, codificando el importe en el color de relleno.

````python
(ggplot(data, aes(x=data.index, y='Ventas', fill='Ventas'))
 + geom_col()
 + theme(axis_text_x=element_text(rotation=45, hjust=1)))
````
