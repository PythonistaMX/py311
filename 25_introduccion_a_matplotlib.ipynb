## El objeto ```mpl.pyplot```.

*Matplotlib* contiene una muy extensa biblioteca, la cual tiene como componente principal al objeto ```matplotlib.pyplot```. Por convención, ```matplotlib.pyplot``` se importa como ```plt```. Se utilizará esta convención para próximas referencias.

El uso de ```plt``` permite crear objetos capaces de desplegar gráficos y definir múltiples características de éstos.

A lo largo de este capítulo se explorarán algunos de dichos recursos.

https://matplotlib.org/3.5.3/api/_as_gen/matplotlib.pyplot.html

````python
from matplotlib import pyplot as plt
````

### El método ```plt.plot()```.

Esta función permite crear una o más gráficas en 2 dimensiones a partir de arreglos de puntos para los ejes ```x``` y ```y``` que son ingresados como argumentos.

```
plt.plot(<arreglo para x1>, <arreglo para y1>,
 <arreglo para x2>, <arreglo para y2>, ... 
 <arreglo para xn>, <arreglo para yn>, <otros argumentos>)
```

### El método ```plt.show()```.

El método ```plt. show()``` permite desplegar el gráfico creado con ```plt.plot()``` en el entorno desde el que se ejecutan las instrucciones.

**NOTA:** En el caso de las notebooks de *Jupyter*, no es necesario usar ```plt. show()``` para que el gráfico se despliegue al ejecutar una celda con ```plt.plot()```.

**Ejemplos:**

* Las siguientes celdas crearán un arreglo unidimensional de *Numpy* con nombre ```x```, el cual contendrá ```500``` segmentos lineales que van de ```0``` a ```3π``` usando la función ```np.linspace()```.

````python
x = np.linspace(0, 3*np.pi, 500)
````

````python
x
````

* La siguiente celda utilizará ```plt.plot()``` para desplegar dos gráficas que unirán cada punto definido mediante una línea, las cuales están en función del arreglo ```x```:

* ```np.sin(x ** 2)```
* ```np.cos(x ** 2)```

````python
plt.plot(x, np.sin(x**2), x, np.cos(x ** 2))
````

````python
plt.show()
````

* Las celdas anteriores crearon de forma automática un objeto de tipo ```plt.Figure```, el cual fue utilizado para desplegar las gráficas correspondientes.

* La siguiente celda incluye algunas funciones de ```plt``` que definen el título del gráfico (```plt.title()```), y las estiqueta de los ejes ( ```plt.xlabel```y ```plt.ylabel```).

````python
plt.plot(x, np.sin(x**2), x, np.cos(x ** 2))
plt.title('Funciones sinusoidales')
plt.xlabel('Eje de las x')
plt.ylabel('f(x)')
plt.show()
````

## Capacidades avanzadas de Matplotlib.

Además de los gráficos básicos y elementos fundamentales, *Matplotlib* ofrece una amplia gama de funcionalidades avanzadas que permiten crear visualizaciones complejas y sofisticadas:

### Disposición avanzada de gráficos

* **Subplots con `GridSpec`**: Permite crear distribuciones irregulares de múltiples gráficos en una sola figura, con control granular sobre el tamaño y posición de cada uno.

* **`constrained_layout`**: Un sistema de posicionamiento automático que ajusta espacios entre elementos (títulos, etiquetas, leyendas) para evitar solapamientos sin interferencia manual.

### Gráficos en tres dimensiones (3D)

* **Gráficos de superficie** (`plot_surface`): Visualización de funciones $z = f(x, y)$ como superficies en el espacio tridimensional.

* **Gráficos de dispersión 3D** (`scatter3D`): Puntos en un espacio tridimensional con opciones de coloreo y tamaño variable.

* **Gráficos de contorno en 3D** (`contour3D`): Proyecciones de superficies mediante líneas de nivel en tres dimensiones.

### Gráficos vectoriales y campos

* **Gráficos de vectores** (`quiver`): Representación de campos vectoriales mediante flechas que indican dirección y magnitud.

* **Gráficos de flujo** (`streamplot`): Líneas de corriente que visualizan el flujo de un campo vectorial de forma continua y elegante.

### Gráficos en coordenadas polares

* **Gráficos polares** (`projection='polar'`): Visualización de datos en coordenadas polares $(r, \theta)$, útiles para datos periódicos, direccionales o cíclicos.

### Ejes secundarios y múltiples

* **Ejes duales**: Permite tener dos o más ejes $y$ diferentes en la misma figura, útil para comparar variables en diferentes escalas.

* **Ejes incrustados**: Pequeños gráficos dentro de gráficos principales para mostrar detalles o contexto adicional.

### Animaciones

* **Animaciones con `FuncAnimation`**: Creación de animaciones que evolucionan en el tiempo, actualizando datos y redibujando la figura en cada fotograma.

* **Exportación de animaciones**: Guardado de animaciones en formatos como MP4 o GIF para compartir o presentar.

### Configuración global y estilos

* **Parámetros de configuración** (`rcParams`): Ajuste de opciones globales como tamaño de fuente, espesor de línea, colores por defecto y temas de figura.

* **Estilos predefinidos**: *Matplotlib* incluye estilos preconfiguranos (como `'seaborn'`, `'ggplot'`, `'dark_background'`) que permiten cambiar la apariencia completa de los gráficos instantáneamente.
