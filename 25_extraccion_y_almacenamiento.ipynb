## Funciones de lectura de Pandas.

* [`pd.read_clipboard()`](https://pandas.pydata.org/docs/reference/api/pandas.read_clipboard.html) Permite leer datos que se encuentran en el portapapeles del sistema operativo.
* [`pd.read_csv()`](https://pandas.pydata.org/docs/reference/api/pandas.read_csv.html) Permite leer datos desde un archivo CSV (*Comma-Separated Values*).
* [`pd.read_excel()`](https://pandas.pydata.org/docs/reference/api/pandas.read_excel.html) Permite leer datos desde un archivo de *Excel* (`.xlsx` con `openpyxl`).
* [`pd.read_feather()`](https://pandas.pydata.org/docs/reference/api/pandas.read_feather.html) Permite leer datos desde el formato [*Feather*](https://github.com/wesm/feather), diseñado para intercambio rápido de dataframes.
* [`pd.read_fwf()`](https://pandas.pydata.org/docs/reference/api/pandas.read_fwf.html) Permite leer archivos de texto con campos de ancho fijo (*Fixed-Width Format*).
* [`pd.read_gbq()`](https://pandas.pydata.org/docs/reference/api/pandas.read_gbq.html) Para leer datos de [*Google BigQuery*](https://pandas-gbq.readthedocs.io/en/latest/).
* [`pd.read_hdf()`](https://pandas.pydata.org/docs/reference/api/pandas.read_hdf.html) Para leer datos en formato [HDF5](https://www.hdfgroup.org/solutions/hdf5), orientado al almacenamiento de grandes volúmenes de datos científicos.
* [`pd.read_html()`](https://pandas.pydata.org/docs/reference/api/pandas.read_html.html) Extrae todas las tablas `<table>` de un documento HTML.
* [`pd.read_json()`](https://pandas.pydata.org/docs/reference/api/pandas.read_json.html) Permite leer datos en formato JSON (*JavaScript Object Notation*).
* [`pd.read_orc()`](https://pandas.pydata.org/docs/reference/api/pandas.read_orc.html) Permite leer datos en formato [*Apache ORC*](https://orc.apache.org/), formato columnar optimizado para *big data*. Disponible desde Pandas 1.0.
* [`pd.read_parquet()`](https://pandas.pydata.org/docs/reference/api/pandas.read_parquet.html) Permite leer datos en formato [*Parquet*](https://databricks.com/glossary/what-is-parquet), formato columnar de alto rendimiento ampliamente usado en ecosistemas de datos.
* [`pd.read_pickle()`](https://pandas.pydata.org/docs/reference/api/pandas.read_pickle.html) Permite leer objetos serializados en formato *Pickle* de Python.
* [`pd.read_sas()`](https://pandas.pydata.org/docs/reference/api/pandas.read_sas.html) Permite leer archivos de datos en formato *SAS* (`.sas7bdat`, `.xport`).
* [`pd.read_sql()`](https://pandas.pydata.org/docs/reference/api/pandas.read_sql.html) Permite leer datos a partir de una consulta SQL o nombre de tabla en una base de datos.
* [`pd.read_sql_query()`](https://pandas.pydata.org/docs/reference/api/pandas.read_sql_query.html) Ejecuta una consulta SQL y retorna los resultados como un dataframe.
* [`pd.read_sql_table()`](https://pandas.pydata.org/docs/reference/api/pandas.read_sql_table.html) Lee una tabla completa de una base de datos relacional como un dataframe.
* [`pd.read_stata()`](https://pandas.pydata.org/docs/reference/api/pandas.read_stata.html) Permite leer archivos de datos en formato *Stata* (`.dta`).
* [`pd.read_table()`](https://pandas.pydata.org/docs/reference/api/pandas.read_table.html) Similar a `pd.read_csv()`, pero con separador de tabulación (`\t`) por defecto.
* [`pd.read_xml()`](https://pandas.pydata.org/docs/reference/api/pandas.read_xml.html) Permite leer datos en formato XML (*eXtensible Markup Language*). Disponible desde Pandas 1.3.

**Nota:** En la mayor parte de los casos, los datos extraídos son almacenados en un *dataframe*.

## Métodos de persistencia y almacenamiento de los dataframes de *Pandas*.

* [`pd.DataFrame.to_clipboard()`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.to_clipboard.html) Copia el contenido del dataframe al portapapeles del sistema.
* [`pd.DataFrame.to_csv()`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.to_csv.html) Exporta el dataframe a un archivo CSV.
* [`pd.DataFrame.to_dict()`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.to_dict.html) Convierte el dataframe a un diccionario de Python.
* [`pd.DataFrame.to_excel()`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.to_excel.html) Exporta el dataframe a un archivo de *Excel*.
* [`pd.DataFrame.to_feather()`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.to_feather.html) Exporta el dataframe al formato *Feather*.
* [`pd.DataFrame.to_gbq()`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.to_gbq.html) Exporta el dataframe a *Google BigQuery*.
* [`pd.DataFrame.to_hdf()`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.to_hdf.html) Exporta el dataframe al formato HDF5.
* [`pd.DataFrame.to_html()`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.to_html.html) Convierte el dataframe a una tabla HTML.
* [`pd.DataFrame.to_json()`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.to_json.html) Exporta el dataframe al formato JSON.
* [`pd.DataFrame.to_latex()`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.to_latex.html) Convierte el dataframe a una tabla en formato *LaTeX*.
* [`pd.DataFrame.to_markdown()`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.to_markdown.html) Convierte el dataframe a una tabla en formato *Markdown*. Disponible desde Pandas 1.0.
* [`pd.DataFrame.to_numpy()`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.to_numpy.html) Convierte el dataframe a un arreglo de *NumPy*.
* [`pd.DataFrame.to_orc()`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.to_orc.html) Exporta el dataframe al formato *Apache ORC*. Disponible desde Pandas 1.4.2.
* [`pd.DataFrame.to_parquet()`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.to_parquet.html) Exporta el dataframe al formato *Parquet*.
* [`pd.DataFrame.to_pickle()`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.to_pickle.html) Serializa el dataframe en formato *Pickle* de Python.
* [`pd.DataFrame.to_records()`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.to_records.html) Convierte el dataframe a un arreglo de registros de *NumPy*.
* [`pd.DataFrame.to_sql()`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.to_sql.html) Escribe el dataframe en una tabla de una base de datos SQL.
* [`pd.DataFrame.to_stata()`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.to_stata.html) Exporta el dataframe al formato *Stata*.
* [`pd.DataFrame.to_xml()`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.to_xml.html) Exporta el dataframe al formato XML. Disponible desde Pandas 1.3.

## Obtención de datos a partir de datos abiertos publicados por el INEGI.

Los datos abiertos del Indicador Trimestral de la Actividad Económica Estatal (ITAEE) del INEGI ofrecen información desestacionalizada y original de la coyuntura económica de las 32 entidades federativas de México.

https://www.inegi.org.mx/programas/itaee/2018/#tabulados

En este caso se descargará la hoja de cálculo publicada en el sitio del INEGI, que contiene los datos trimestrales del ITAEE para el total de la economía, localizada en la siguiente *URL*:

https://www.inegi.org.mx/contenidos/programas/itaee/2018/tabulados/ori/ITAEE_r_3.xlsx

**ADVERTENCIA:** El enlace anterior es el que se encontraba activo al momento de escribir este capítulo. Es posible que el INEGI actualice o cambie la ubicación de los archivos, por lo que se recomienda verificar la página oficial y el contenido del programa ITAEE para obtener la información más reciente.


### Obtención del archivo usando `urllib`.

El módulo de la biblioteca estándar `urllib.request.urlretrieve` permite descargar el archivo directamente desde la *URL* proporcionada por el INEGI. Se especifica un nombre de archivo local para guardar el contenido descargado.

```
urllib.request.urlretrieve(<URL>, <archivo_local>)
```
Donde:
- `<URL>` es la dirección web del archivo que se desea descargar.
- `<archivo_local>` es el nombre del archivo local donde se guardará el contenido descargado

* La siguiente celda importará la función `urllib.request` para realizar solicitudes HTTP y descargar archivos desde la web.

````python
import urllib.request
````

* Se le asignará la *URL* a una variable llamada `url`.
* Se construirá la ruta local para guardar el archivo descargado utilizando `os.path.join()`, combinando el directorio `data` con el nombre del archivo `ITAEE.xlsx`.

````python
url = "https://www.inegi.org.mx/contenidos/programas/itaee/2018/tabulados/ori/ITAEE_r_3.xlsx"
ruta_local = os.path.join("data", "ITAEE.xlsx")
````

* Se realizará la descarga del archivo utilizando `urllib.request.urlretrieve()` en la *URL* especificada en la variable `url`, guardando el achivo descargado en la ruta local definida en `ruta_local`.

````python
urllib.request.urlretrieve(url, ruta_local)
````

Se verifica que el archivo `data/ITAEE.xlsx` fue generado correctamente en el directorio de trabajo.

````python
os.path.exists(ruta_local)
````

### Carga del archivo con `pd.read_excel()`.

Se carga el archivo desde la ruta local `ruta_local` con `pd.read_excel()` y se almacena en el dataframe `original`.

````python
original = pd.read_excel(ruta_local)
````

Se muestra el dataframe completo para explorar la estructura general del archivo antes de procesarlo.

````python
original
````

* La siguiente celda eliminará las primeras filas, que corresponden a anotaciones.

````python
datos = original.drop(index=range(0, 6))
datos
````

* La columna que contiene los nombres de las entidades federativas será el índice de *dataframe* `datos`.

````python
datos.set_index("Instituto Nacional de Estadística y Geografía (INEGI).", inplace=True)
````

* La columna que contienen datos de las entidades federativas se renombrará a `Entidad`.

````python
datos.index.name = "Entidad"
````

````python
datos
````

### Obtención de los datos relevantes.

El dataframe contiene tres bloques de 33 filas (1 total nacional + 32 entidades federativas): índice de volumen físico, variación porcentual anual y variación acumulada. Las primeras 33 filas (`iloc[:33]`) corresponden al **índice de volumen físico base 2018=100**, que es el dato de interés.

* Se selecciona el primer bloque de filas, que contiene el índice de volumen físico para las 33 entidades.

````python
datos = datos.iloc[:33]
````

* Se muestra el subconjunto con los datos del índice de volumen físico.

````python
datos
````

### Creación de un índice de columnas adecuado.
Las columnas del dataframe no tienen nombres adecuados, por lo que se asignarán nombres descriptivos a cada columna.

Se construye el `MultiIndex` usando `pd.MultiIndex.from_product()`, que genera un índice a partir del producto cartesiano de los años (1980-2025) y los periodos (T1, T2, T3, T4, Anual).

````python
periodos = pd.MultiIndex.from_product((range(1980, 2026), ["T1", "T2", "T3", "T4", "Anual"]))
````

* Se muestra el `MultiIndex` creado para verificar su estructura.

````python
periodos
````

Se asigna el `MultiIndex` construido como índice de columnas del dataframe `datos`.

````python
datos.columns = periodos
````

Se muestra el dataframe con el índice jerárquico de columnas ya aplicado.

````python
datos
````

Gracias al `MultiIndex`, se puede acceder a todos los periodos de un año con una sola clave entera. Se obtienen todos los periodos del año 2005.

````python
datos[2005]
````

Se accede al primer trimestre (`T1`) del año 2005 para todas las entidades.

````python
datos[2005]['T1']
````

La primera fila (índice 0) corresponde al total nacional. Con `[1:]` se excluye para obtener únicamente los valores de las 32 entidades federativas.

````python
datos[2005]['T1'][1:]
````

Se almacena la selección en la variable `periodo` para facilitar su análisis.

````python
periodo = datos[2005]['T1'][1:]
````

Se calcula el promedio del índice de volumen físico del primer trimestre de 2005 entre todas las entidades federativas.

````python
periodo.mean()
````

Se exporta la serie `periodo` a un archivo Excel con `to_excel()`, demostrando el uso de `openpyxl` para escritura.

````python
periodo.to_excel('ITAEE_corregido.xlsx')
````

## Extracción y escritura en formato CSV.

El formato CSV (*Comma-Separated Values*) es uno de los más utilizados para el intercambio de datos tabulares por su simplicidad y compatibilidad universal. *Pandas* permite exportar dataframes con el método `to_csv()` e importarlos con `pd.read_csv()`.

Se crea el directorio `data/` si no existe. El dataframe `datos` se exporta a CSV con `to_csv()` después de convertir el índice de entidades a una columna mediante `reset_index()`.

````python
datos.reset_index().to_csv('data/ITAEE.csv', index=True)
````

Se lee el archivo CSV recién creado con `pd.read_csv()` y se almacena en el dataframe `nuevos_datos`.

````python
nuevos_datos = pd.read_csv('data/ITAEE.csv')
````

Se muestra el dataframe `nuevos_datos` para verificar que el contenido fue almacenado y recuperado correctamente.

````python
nuevos_datos
````

````python
nuevos_datos.info()
````

### El formato *parquet*.

El formato *Parquet* es un formato de almacenamiento columnar optimizado para el rendimiento y la compresión, ampliamente utilizado en ecosistemas de big data. *Pandas* permite exportar dataframes a *Parquet* con `to_parquet()` e importarlos con `pd.read_parquet()`.

* Se almacenarán los datos del *dataframe* `datos` en un archivo `parquet`.

````python
datos.to_parquet('data/ITAEE.parquet')
````

* Se cargará el archivo `parquet` recién creado con `pd.read_parquet()` y se almacenará en el dataframe `datos_parquet`.

````python
datos_parquet = pd.read_parquet('data/ITAEE.parquet')
datos_parquet
````
