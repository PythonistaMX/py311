## Funciones de lectura de Pandas.

* [`pd.read_clipboard()`](https://pandas.pydata.org/docs/reference/api/pandas.read_clipboard.html) Permite leer datos que se encuentran en el portapapeles del sistema operativo.
* [`pd.read_csv()`](https://pandas.pydata.org/docs/reference/api/pandas.read_csv.html) Permite leer datos desde un archivo CSV (*Comma-Separated Values*).
* [`pd.read_excel()`](https://pandas.pydata.org/docs/reference/api/pandas.read_excel.html) Permite leer datos desde un archivo de *Excel* (`.xlsx` con `openpyxl`).
* [`pd.read_feather()`](https://pandas.pydata.org/docs/reference/api/pandas.read_feather.html) Permite leer datos desde el formato [*Feather*](https://github.com/wesm/feather), diseñado para intercambio rápido de dataframes.
* [`pd.read_fwf()`](https://pandas.pydata.org/docs/reference/api/pandas.read_fwf.html) Permite leer archivos de texto con campos de ancho fijo (*Fixed-Width Format*).
* [`pd.read_gbq()`](https://pandas.pydata.org/docs/reference/api/pandas.read_gbq.html) Para leer datos de [*Google BigQuery*](https://pandas-gbq.readthedocs.io/en/latest/).
* [`pd.read_hdf()`](https://pandas.pydata.org/docs/reference/api/pandas.read_hdf.html) Para leer datos en formato [HDF5](https://www.hdfgroup.org/solutions/hdf5), orientado al almacenamiento de grandes volúmenes de datos científicos.
* [`pd.read_html()`](https://pandas.pydata.org/docs/reference/api/pandas.read_html.html) Extrae todas las tablas `<table>` de un documento HTML.
* [`pd.read_json()`](https://pandas.pydata.org/docs/reference/api/pandas.read_json.html) Permite leer datos en formato JSON (*JavaScript Object Notation*).
* [`pd.read_orc()`](https://pandas.pydata.org/docs/reference/api/pandas.read_orc.html) Permite leer datos en formato [*Apache ORC*](https://orc.apache.org/), formato columnar optimizado para *big data*. Disponible desde Pandas 1.0.
* [`pd.read_parquet()`](https://pandas.pydata.org/docs/reference/api/pandas.read_parquet.html) Permite leer datos en formato [*Parquet*](https://databricks.com/glossary/what-is-parquet), formato columnar de alto rendimiento ampliamente usado en ecosistemas de datos.
* [`pd.read_pickle()`](https://pandas.pydata.org/docs/reference/api/pandas.read_pickle.html) Permite leer objetos serializados en formato *Pickle* de Python.
* [`pd.read_sas()`](https://pandas.pydata.org/docs/reference/api/pandas.read_sas.html) Permite leer archivos de datos en formato *SAS* (`.sas7bdat`, `.xport`).
* [`pd.read_sql()`](https://pandas.pydata.org/docs/reference/api/pandas.read_sql.html) Permite leer datos a partir de una consulta SQL o nombre de tabla en una base de datos.
* [`pd.read_sql_query()`](https://pandas.pydata.org/docs/reference/api/pandas.read_sql_query.html) Ejecuta una consulta SQL y retorna los resultados como un dataframe.
* [`pd.read_sql_table()`](https://pandas.pydata.org/docs/reference/api/pandas.read_sql_table.html) Lee una tabla completa de una base de datos relacional como un dataframe.
* [`pd.read_stata()`](https://pandas.pydata.org/docs/reference/api/pandas.read_stata.html) Permite leer archivos de datos en formato *Stata* (`.dta`).
* [`pd.read_table()`](https://pandas.pydata.org/docs/reference/api/pandas.read_table.html) Similar a `pd.read_csv()`, pero con separador de tabulación (`\t`) por defecto.
* [`pd.read_xml()`](https://pandas.pydata.org/docs/reference/api/pandas.read_xml.html) Permite leer datos en formato XML (*eXtensible Markup Language*). Disponible desde Pandas 1.3.

**Nota:** En la mayor parte de los casos, los datos extraídos son almacenados en un *dataframe*.

## Obtención de datos a partir de datos abiertos publicados por el INEGI.

El Indicador Trimestral de la Actividad Económica Estatal (ITAEE) del INEGI es un indicador de coyuntura que permite analizar la evolución de la actividad económica por entidad federativa y por tipo de actividad económica.

En este capítulo se trabajará con la serie en números índice de volumen físico con **año base 2018=100**. Esto significa que el nivel de referencia del índice se fija en 2018 para facilitar comparaciones en el tiempo; no implica que todos los valores trimestrales de 2018 sean exactamente 100.

La página oficial del programa ITAEE se encuentra en:

https://www.inegi.org.mx/programas/itaee/2018/#tabulados

En este caso se descargará la hoja de cálculo publicada por INEGI para el total de la economía, disponible en:

https://www.inegi.org.mx/contenidos/programas/itaee/2018/tabulados/ori/ITAEE_r_3.xlsx

**ADVERTENCIA:** La ubicación de los archivos puede cambiar con nuevas publicaciones del INEGI. Se recomienda verificar siempre la página oficial del programa ITAEE para obtener la versión más reciente.

### Obtención del archivo usando `urllib`.

El módulo de la biblioteca estándar `urllib.request.urlretrieve` permite descargar el archivo directamente desde la *URL* proporcionada por el INEGI. Se especifica un nombre de archivo local para guardar el contenido descargado.

```
urllib.request.urlretrieve(<URL>, <archivo_local>)
```
Donde:
- `<URL>` es la dirección web del archivo que se desea descargar.
- `<archivo_local>` es el nombre del archivo local donde se guardará el contenido descargado

* La siguiente celda importará la función `urllib.request` para realizar solicitudes HTTP y descargar archivos desde la web.

````python
import urllib.request
````

* Se le asignará la *URL* a una variable llamada `url`.
* Se construirá la ruta local para guardar el archivo descargado utilizando `os.path.join()`, combinando el directorio `data_tmp` con el nombre del archivo `ITAEE.xlsx`.

````python
url = "https://www.inegi.org.mx/contenidos/programas/itaee/2018/tabulados/ori/ITAEE_r_3.xlsx"
ruta_local = os.path.join("data_tmp", "ITAEE.xlsx")
````

* Se realizará la descarga del archivo utilizando `urllib.request.urlretrieve()` en la *URL* especificada en la variable `url`, guardando el archivo descargado en la ruta local definida en `ruta_local`.

````python
urllib.request.urlretrieve(url, ruta_local)
````

* Se verifica que el archivo fue generado correctamente en el directorio de trabajo.

````python
os.path.exists(ruta_local)
````

### Creación de un índice de columnas adecuado.
Las columnas del dataframe no tienen nombres adecuados, por lo que se asignarán nombres descriptivos a cada columna.

## Operaciones con archivos de formato *CSV*.

El formato *CSV (*Comma-Separated Values*) es uno de los más utilizados para el intercambio de datos tabulares por su simplicidad y compatibilidad universal. 

*Pandas* permite exportar dataframes con el método `df.to_csv()` e importarlos con `pd.read_csv()`.

Aún cuando el formato *CSV* es ampliamente compatible, tiene limitaciones como la falta de soporte para tipos de datos complejos, problemas con caracteres especiales y la ausencia de metadatos. Por ello, es importante considerar estas limitaciones al elegir el formato de almacenamiento para los datos. 

Todos los datos extraídos de un archivo *CSV* mediante `pd.read_csv()` son considerados como cadenas de caracteres, por lo que es necesario convertirlos a los tipos de datos adecuados para su análisis.

* La documentación de `pd.read_csv()` está disponible en:
https://pandas.pydata.org/docs/reference/api/pandas.read_csv.html

* La documentación de `df.to_csv()` está disponible en:
https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.to_csv.html

* Se creará la variable `ruta_csv` con la ruta local para guardar el archivo en formato *CSV*, utilizando `os.path.join()` para combinar el directorio `data` con el nombre del archivo `ITAEE.csv`.

````python
ruta_csv = os.path.join("data_tmp", "ITAEE.csv")
````

* El dataframe `datos` se exportará en formato *CSV* con el método `to_csv()` a la ruta especificada en la variable `ruta_csv`. 
    * El argumento `header=True` se utiliza para incluir los nombres de las columnas en la primera fila del archivo *CSV*.
    * El argumento `encoding='utf-8'` se utiliza para especificar la codificación de caracteres del archivo *CSV*, asegurando que se manejen correctamente caracteres especiales y acentos.

````python
datos.reset_index().to_csv(ruta_csv, header=True, encoding='utf-8')
````

* Se leerá el archivo *CSV* recién creado con `pd.read_csv()` y se almacena en el dataframe `nuevos_datos`.
   * El argumento `index_col=0` se utiliza para indicar que la primera columna del archivo *CSV* debe ser utilizada como índice del dataframe resultante.
   * El argumento `header=True` se utiliza para indicar que la primera fila del archivo *CSV* contiene los nombres de las columnas.
   * El argumento `encoding='utf-8'` se utiliza para especificar la codificación de caracteres del archivo *CSV* al leerlo, asegurando que se manejen correctamente caracteres especiales y acentos.

   **Nota:** El formato *CSV* no soporta índices de columnas jerárquicos (`MultiIndex`), por lo que al exportar el dataframe con un `MultiIndex` se perderá esta estructura y las columnas serán tratadas como cadenas de caracteres. Al leer el archivo *CSV* nuevamente, las columnas se interpretarán como texto sin la jerarquía original.

````python
nuevos_datos = pd.read_csv(ruta_csv, header=0, encoding='utf-8')
````

* Se muestra el dataframe `nuevos_datos` para verificar que el contenido fue almacenado y recuperado correctamente.

````python
nuevos_datos
````

* La siguiente celda muestra que el dataframe `nuevos_datos` está compuesto por datos de tipo `object` (cadenas de caracteres) debido a la naturaleza del formato *CSV*.

````python
nuevos_datos.info()
````

## Operaciones con archivos de formato *parquet*.

El formato *Parquet* es un formato de almacenamiento columnar optimizado para el rendimiento y la compresión, ampliamente utilizado en ecosistemas de big data. *Pandas* permite exportar dataframes a *Parquet* con el método `df.to_parquet()` e importarlos con `pd.read_parquet()`.

* La documentación de `pd.read_parquet()` está disponible en: https://pandas.pydata.org/docs/reference/api/pandas.read_parquet.html
* La documentación de `df.to_parquet()` está disponible en: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.to_parquet.html

* Se definirá la variable `ruta_parquet` con la ruta local para guardar el archivo en formato *Parquet*, utilizando `os.path.join()` para combinar el directorio `data` con el nombre del archivo `ITAEE.parquet`.

````python
ruta_parquet = os.path.join("data_tmp", "ITAEE.parquet")
````

* Se almacenarán los datos del *dataframe* `datos` en un archivo en formato *parquet* en la ruta `ruta_parquet`.
   * El argumento `index=True` se utiliza para incluir el índice del dataframe como parte del archivo *Parquet*.
   * El argumento `engine='pyarrow'` se utiliza para especificar el motor de escritura a utilizar, en este caso `pyarrow`, que es una biblioteca de Python para trabajar con el formato *Parquet*. La biblioteca `pyarrow` debe estar instalada para utilizar este motor y se estudiará en capítulos posteriores.

````python
datos.to_parquet(ruta_parquet, index=True, engine='pyarrow')
````

* Se cargará el archivo `parquet` recién creado con `pd.read_parquet()` y se almacenará en el dataframe `datos_parquet`.

````python
datos_parquet = pd.read_parquet(ruta_parquet,engine='pyarrow')
datos_parquet
````

* La siguiente celda mostrará que el dataframe `datos_parquet` conserva los tipos de datos originales, a diferencia del formato *CSV* que los convierte a cadenas de caracteres. Esto se debe a que el formato *Parquet* es capaz de almacenar metadatos sobre los tipos de datos, índices y columnas, lo que permite una recuperación más precisa de la información al leer el archivo.

````python
datos_parquet.info()
````
