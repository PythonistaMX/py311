# El accesor *str* de *Pandas*.

El accesor [`Series.str`](https://pandas.pydata.org/docs/reference/series.html#string-handling) de *Pandas* permite realizar operaciones vectorizadas sobre cadenas de caracteres en una serie. A diferencia de iterar manualmente sobre cada elemento, el accesor `.str` aplica transformaciones y búsquedas de forma eficiente sobre toda la serie simultáneamente.

De manera similar al accesor `.dt` para fechas, el accesor `.str` expone métodos equivalentes a los de las cadenas de caracteres nativas de *Python*, con la ventaja de operar sobre colecciones completas y manejar correctamente los valores nulos (`pd.NA`, `np.nan`).

Para utilizar el accesor, la serie debe contener datos de tipo `object` o `StringDtype`.

````python
import pandas as pd
````

## *DataFrame* ilustrativo.

Se creará un *dataframe* llamado `empleados` con datos sintéticos que incluyen nombres con espacios e inconsistencias de capitalización, correos electrónicos, números de teléfono y ciudades. Este *dataframe* se utilizará a lo largo de la notebook para ilustrar los métodos del accesor `.str`.

````python
datos = {
    'nombre': ['  Ana García  ', 'CARLOS LÓPEZ', 'maría martínez', 'José Rodríguez  ', '  LAURA SÁNCHEZ'],
    'email': ['ana.garcia@empresa.com', 'carlos.lopez@empresa.com', 'maria.martinez@empresa.com',
               'jose.rodriguez@empresa.com', 'laura.sanchez@empresa.com'],
    'telefono': ['55-1234-5678', '55-9876-5432', '55-2345-6789', '55-3456-7890', '55-4567-8901'],
    'ciudad': ['Ciudad de México', 'Guadalajara', 'Monterrey', 'Ciudad de México', 'Puebla']
}
empleados = pd.DataFrame(datos)
empleados
````

## El accesor `Series.str`.

El accesor `.str` expone un conjunto de métodos agrupados por funcionalidad. La siguiente tabla resume las categorías principales:

| Categoría | Métodos principales |
| :--- | :--- |
| Transformación de capitalización | `.str.lower()`, `.str.upper()`, `.str.title()`, `.str.capitalize()` |
| Eliminación de espacios | `.str.strip()`, `.str.lstrip()`, `.str.rstrip()` |
| Verificación de contenido | `.str.contains()`, `.str.startswith()`, `.str.endswith()`, `.str.match()` |
| Reemplazo | `.str.replace()` |
| División y acceso | `.str.split()`, `.str.get()`, `.str.slice()` |
| Extracción con regex | `.str.extract()`, `.str.findall()` |
| Longitud y conteo | `.str.len()`, `.str.count()` |
| Relleno y alineación | `.str.pad()`, `.str.center()`, `.str.ljust()`, `.str.rjust()`, `.str.zfill()` |

La documentación completa del accesor `Series.str` se puede consultar en:

https://pandas.pydata.org/docs/reference/series.html#string-handling

### Transformaciones de mayúsculas y minúsculas.

Los métodos de capitalización permiten estandarizar el formato del texto en una serie. Los métodos disponibles son:

* [`Series.str.lower()`](https://pandas.pydata.org/docs/reference/api/pandas.Series.str.lower.html): convierte todos los caracteres a minúsculas.
* [`Series.str.upper()`](https://pandas.pydata.org/docs/reference/api/pandas.Series.str.upper.html): convierte todos los caracteres a mayúsculas.
* [`Series.str.title()`](https://pandas.pydata.org/docs/reference/api/pandas.Series.str.title.html): convierte la primera letra de cada palabra a mayúscula.
* [`Series.str.capitalize()`](https://pandas.pydata.org/docs/reference/api/pandas.Series.str.capitalize.html): convierte la primera letra de la cadena a mayúscula y el resto a minúsculas.

La sintaxis genérica es:

```
serie.str.lower()
serie.str.upper()
serie.str.title()
serie.str.capitalize()
```

**Ejemplos:**

* La siguiente celda convertirá los valores de la columna `'nombre'` a minúsculas con [`.str.lower()`](https://pandas.pydata.org/docs/reference/api/pandas.Series.str.lower.html).

````python
empleados['nombre'].str.lower()
````

* La siguiente celda convertirá los valores de la columna `'nombre'` a mayúsculas con [`.str.upper()`](https://pandas.pydata.org/docs/reference/api/pandas.Series.str.upper.html).

````python
empleados['nombre'].str.upper()
````

* La siguiente celda convertirá los valores de la columna `'nombre'` a formato de título con [`.str.title()`](https://pandas.pydata.org/docs/reference/api/pandas.Series.str.title.html).

````python
empleados['nombre'].str.title()
````

* La siguiente celda convertirá los valores de la columna `'nombre'` con [`.str.capitalize()`](https://pandas.pydata.org/docs/reference/api/pandas.Series.str.capitalize.html), poniendo en mayúscula solo la primera letra de cada cadena.

````python
empleados['nombre'].str.capitalize()
````

### Eliminación de espacios en blanco.

Los métodos de eliminación de espacios permiten limpiar caracteres de espacio al inicio, al final, o en ambos extremos de una cadena:

* [`Series.str.strip()`](https://pandas.pydata.org/docs/reference/api/pandas.Series.str.strip.html): elimina espacios al inicio y al final.
* [`Series.str.lstrip()`](https://pandas.pydata.org/docs/reference/api/pandas.Series.str.lstrip.html): elimina espacios solo al inicio (*left strip*).
* [`Series.str.rstrip()`](https://pandas.pydata.org/docs/reference/api/pandas.Series.str.rstrip.html): elimina espacios solo al final (*right strip*).

La sintaxis genérica es:

```
serie.str.strip()
serie.str.lstrip()
serie.str.rstrip()
```

**Nota:** Estos métodos también aceptan un argumento `<caracteres>` para eliminar caracteres específicos en lugar de espacios.

**Ejemplos:**

* La siguiente celda eliminará los espacios en blanco al inicio y al final de los valores de la columna `'nombre'` con [`.str.strip()`](https://pandas.pydata.org/docs/reference/api/pandas.Series.str.strip.html).

````python
empleados['nombre'].str.strip()
````

* La siguiente celda encadenará [`.str.strip()`](https://pandas.pydata.org/docs/reference/api/pandas.Series.str.strip.html) y [`.str.title()`](https://pandas.pydata.org/docs/reference/api/pandas.Series.str.title.html) para normalizar completamente la columna `'nombre'`, eliminando espacios y estandarizando la capitalización.

````python
empleados['nombre'].str.strip().str.title()
````

### Verificación de contenido.

Los métodos de verificación regresan una serie booleana que indica si cada elemento cumple con la condición especificada. Son especialmente útiles para filtrar registros:

* [`Series.str.contains()`](https://pandas.pydata.org/docs/reference/api/pandas.Series.str.contains.html): verifica si cada elemento contiene un patrón o subcadena.
* [`Series.str.startswith()`](https://pandas.pydata.org/docs/reference/api/pandas.Series.str.startswith.html): verifica si cada elemento comienza con una subcadena.
* [`Series.str.endswith()`](https://pandas.pydata.org/docs/reference/api/pandas.Series.str.endswith.html): verifica si cada elemento termina con una subcadena.
* [`Series.str.match()`](https://pandas.pydata.org/docs/reference/api/pandas.Series.str.match.html): verifica si cada elemento coincide con una expresión regular desde el inicio de la cadena.

La sintaxis genérica es:

```
serie.str.contains(<patrón>, case=<sensible_a_mayúsculas>, na=<valor_para_nulos>)
serie.str.startswith(<prefijo>)
serie.str.endswith(<sufijo>)
```

**Ejemplos:**

* La siguiente celda verificará qué registros de la columna `'ciudad'` contienen la subcadena `'México'` con [`.str.contains()`](https://pandas.pydata.org/docs/reference/api/pandas.Series.str.contains.html).

````python
empleados['ciudad'].str.contains('México')
````

* La siguiente celda utilizará la serie booleana anterior para filtrar el *dataframe* `empleados` y mostrar solo los registros cuya ciudad contiene `'México'`.

````python
empleados[empleados['ciudad'].str.contains('México')]
````

* La siguiente celda verificará qué correos electrónicos comienzan con `'ana'` con [`.str.startswith()`](https://pandas.pydata.org/docs/reference/api/pandas.Series.str.startswith.html).

````python
empleados['email'].str.startswith('ana')
````

* La siguiente celda verificará qué correos electrónicos terminan con `'.com'` con [`.str.endswith()`](https://pandas.pydata.org/docs/reference/api/pandas.Series.str.endswith.html).

````python
empleados['email'].str.endswith('.com')
````

### Reemplazo de cadenas.

El método [`Series.str.replace()`](https://pandas.pydata.org/docs/reference/api/pandas.Series.str.replace.html) reemplaza ocurrencias de un patrón por una cadena de sustitución. A diferencia del método `str.replace()` de *Python*, soporta expresiones regulares de forma nativa.

La sintaxis genérica es:

```
serie.str.replace(<patrón>, <reemplazo>, regex=<booleano>)
```

Donde:
* `<patrón>` es la cadena o expresión regular a buscar.
* `<reemplazo>` es la cadena con la que se sustituirán las coincidencias.
* `regex=True` indica que el patrón es una expresión regular (valor predeterminado).

**Ejemplos:**

* La siguiente celda reemplazará los guiones `'-'` de la columna `'telefono'` por espacios con [`.str.replace()`](https://pandas.pydata.org/docs/reference/api/pandas.Series.str.replace.html).

````python
empleados['telefono'].str.replace('-', ' ', regex=False)
````

* La siguiente celda reemplazará el dominio `@empresa.com` por `@corp.mx` en la columna `'email'` utilizando una expresión regular.

````python
empleados['email'].str.replace(r'@empresa\.com$', '@corp.mx', regex=True)
````

### División y acceso a elementos.

Los métodos de división permiten separar cadenas en partes y acceder a elementos específicos:

* [`Series.str.split()`](https://pandas.pydata.org/docs/reference/api/pandas.Series.str.split.html): divide cada cadena por un separador y devuelve una serie de listas.
* [`Series.str.get()`](https://pandas.pydata.org/docs/reference/api/pandas.Series.str.get.html): accede al elemento en la posición indicada dentro de cada lista o cadena.
* [`Series.str.slice()`](https://pandas.pydata.org/docs/reference/api/pandas.Series.str.slice.html): extrae una subcadena mediante índices de inicio y fin, equivalente a la notación `[inicio:fin]`.

La sintaxis genérica es:

```
serie.str.split(<separador>, expand=<booleano>)
serie.str.get(<posición>)
serie.str.slice(<inicio>, <fin>)
```

**Ejemplos:**

* La siguiente celda dividirá los valores de la columna `'email'` por el carácter `'@'` con [`.str.split()`](https://pandas.pydata.org/docs/reference/api/pandas.Series.str.split.html).

````python
empleados['email'].str.split('@')
````

* La siguiente celda dividirá la columna `'email'` por `'@'` y accederá al primer elemento (el nombre de usuario) con [`.str.get()`](https://pandas.pydata.org/docs/reference/api/pandas.Series.str.get.html).

````python
empleados['email'].str.split('@').str.get(0)
````

* La siguiente celda utilizará `expand=True` en [`.str.split()`](https://pandas.pydata.org/docs/reference/api/pandas.Series.str.split.html) para expandir la división de `'email'` en dos columnas: usuario y dominio.

````python
empleados['email'].str.split('@', expand=True).rename(columns={0: 'usuario', 1: 'dominio'})
````

* La siguiente celda extraerá los primeros dos dígitos del número de teléfono (el código de área) con [`.str.slice()`](https://pandas.pydata.org/docs/reference/api/pandas.Series.str.slice.html).

````python
empleados['telefono'].str.slice(0, 2)
````

### Extracción con expresiones regulares.

Los métodos de extracción permiten capturar patrones específicos dentro de las cadenas usando expresiones regulares:

* [`Series.str.extract()`](https://pandas.pydata.org/docs/reference/api/pandas.Series.str.extract.html): extrae grupos de captura de una expresión regular y los devuelve como columnas de un *dataframe*.
* [`Series.str.findall()`](https://pandas.pydata.org/docs/reference/api/pandas.Series.str.findall.html): encuentra todas las ocurrencias de un patrón y las devuelve como una lista por cada elemento.

La sintaxis genérica es:

```
serie.str.extract(<expresión_regular>)
serie.str.findall(<expresión_regular>)
```

**Nota:** En `Series.str.extract()`, los grupos de captura `()` de la expresión regular se convierten en columnas del *dataframe* resultante.

**Ejemplos:**

* La siguiente celda extraerá las tres partes del número de teléfono (código, número local y extensión) usando grupos de captura con [`.str.extract()`](https://pandas.pydata.org/docs/reference/api/pandas.Series.str.extract.html).

````python
empleados['telefono'].str.extract(r'(\d{2})-(\d{4})-(\d{4})').rename(
    columns={0: 'codigo', 1: 'numero', 2: 'extension'}
)
````

* La siguiente celda encontrará todos los grupos de dígitos consecutivos en la columna `'telefono'` con [`.str.findall()`](https://pandas.pydata.org/docs/reference/api/pandas.Series.str.findall.html).

````python
empleados['telefono'].str.findall(r'\d+')
````

### Longitud y conteo.

Los métodos de longitud y conteo permiten obtener métricas sobre el contenido de las cadenas:

* [`Series.str.len()`](https://pandas.pydata.org/docs/reference/api/pandas.Series.str.len.html): devuelve el número de caracteres de cada elemento.
* [`Series.str.count()`](https://pandas.pydata.org/docs/reference/api/pandas.Series.str.count.html): cuenta las ocurrencias de un patrón dentro de cada elemento.

La sintaxis genérica es:

```
serie.str.len()
serie.str.count(<patrón>)
```

**Ejemplos:**

* La siguiente celda calculará la longitud de cada valor en la columna `'nombre'` con [`.str.len()`](https://pandas.pydata.org/docs/reference/api/pandas.Series.str.len.html), incluyendo los espacios en blanco.

````python
empleados['nombre'].str.len()
````

* La siguiente celda contará el número de vocales (incluyendo acentuadas) en cada nombre con [`.str.count()`](https://pandas.pydata.org/docs/reference/api/pandas.Series.str.count.html).

````python
empleados['nombre'].str.count(r'[aeiouáéíóúAEIOUÁÉÍÓÚ]')
````

### Relleno y alineación.

Los métodos de relleno permiten estandarizar la longitud de las cadenas añadiendo caracteres de relleno en distintas posiciones:

* [`Series.str.pad()`](https://pandas.pydata.org/docs/reference/api/pandas.Series.str.pad.html): rellena hasta un ancho dado con un carácter específico.
* [`Series.str.center()`](https://pandas.pydata.org/docs/reference/api/pandas.Series.str.center.html): centra la cadena dentro de un campo de ancho dado.
* [`Series.str.ljust()`](https://pandas.pydata.org/docs/reference/api/pandas.Series.str.ljust.html): alinea la cadena a la izquierda (*left justify*).
* [`Series.str.rjust()`](https://pandas.pydata.org/docs/reference/api/pandas.Series.str.rjust.html): alinea la cadena a la derecha (*right justify*).
* [`Series.str.zfill()`](https://pandas.pydata.org/docs/reference/api/pandas.Series.str.zfill.html): rellena con ceros a la izquierda hasta un ancho dado.

La sintaxis genérica es:

```
serie.str.pad(width=<ancho>, side=<'left'|'right'|'both'>, fillchar=<carácter>)
serie.str.center(width=<ancho>, fillchar=<carácter>)
serie.str.zfill(width=<ancho>)
```

**Ejemplos:**

* La siguiente celda rellenará con puntos a la derecha los valores de la columna `'ciudad'` hasta un ancho de 20 caracteres con [`.str.pad()`](https://pandas.pydata.org/docs/reference/api/pandas.Series.str.pad.html).

````python
empleados['ciudad'].str.pad(width=20, side='right', fillchar='.')
````

* La siguiente celda centrará los valores de la columna `'ciudad'` dentro de un campo de 20 caracteres usando guiones como relleno con [`.str.center()`](https://pandas.pydata.org/docs/reference/api/pandas.Series.str.center.html).

````python
empleados['ciudad'].str.center(20, fillchar='-')
````

* La siguiente celda rellenará con ceros a la izquierda el código de área extraído de la columna `'telefono'` hasta un ancho de 5 caracteres con [`.str.zfill()`](https://pandas.pydata.org/docs/reference/api/pandas.Series.str.zfill.html).

````python
empleados['telefono'].str.slice(0, 2).str.zfill(5)
````
