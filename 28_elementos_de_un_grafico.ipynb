## El objeto `plt`.

El objeto `plt` es una instancia de la clase `matplotlib.pyplot` y es el módulo principal de *Matplotlib* para trazar gráficos y visualizaciones.

Este objeto permite crear diversos objetos de visualización, como figuras, ejes, líneas, puntos, etc., y proporciona una interfaz de alto nivel para personalizar y configurar estos objetos.

Existen múltiples maneras de crear gráficos con *Matplotlib*, pero la forma más común es utilizando el objeto `plt` para crear gráficos de líneas, barras, dispersión, etc., de manera rápida y sencilla, sin necesidad de interactuar directamente con los objetos *figure* y *axes*, los cuales son gestionados automáticamente por `plt`.

## Los objetos *figure*.

Los elementos base para el trazado de gráficos con *Matplotlib* son los objetos *figure*, los cuales son instancias de la clase ```matplotlib.figure.Figure```.

Todos los elementos de un gráfico de *Matplotlib* son parte de un objeto *figure*, el cual es el contenedor principal para todos los elementos gráficos. Un objeto *figure* puede contener uno o más objetos *axes*, los cuales son los que realmente contienen los gráficos y las visualizaciones.

https://matplotlib.org/stable/api/figure_api.html#matplotlib.figure.Figure

### La función ```plt.figure()```.
   
Para poder crear un objeto *figure* se utiliza el método ```plt.figure```.

Los objetos *figure* no son creados de forma automática, por lo que es necesario crear un objeto *figure* antes de poder crear cualquier otro elemento gráfico. Sin embargo, si se intenta crear un elemento gráfico sin haber creado un objeto *figure* previamente, *Matplotlib* creará automáticamente un objeto *figure* con los parámetros por defecto. Para evitar esto, es recomendable crear un objeto *figure* de forma explícita antes de crear cualquier otro elemento gráfico.

Los objetos *figure* no se despliegan por sí mismos, sino que son contenedores para otros elementos gráficos. Para desplegar un objeto *figure* es necesario crear un objeto *axes* dentro del objeto *figure* y luego trazar los gráficos dentro del objeto *axes*.

```
plt.figure(num=<id>, label=<etiqueta>, figsize=<tamaño>, dpi=<resolución>)
```

Donde:
 
* ```<id>``` es un identificador único del objeto *figure*. En caso de no definirse, el identificador será un número que iniciará en ```1``` e irá incrementándose de 1 en 1.
* ```<etiqueta>``` es una cadena de texto que sirve como etiqueta para la figura.
* ```<tamaño>``` es una tupla de dos elementos cuyo primer elemento define el ancho y el segundo la altura de la figura.
* ```<resolución>``` que corresponde al número de *dpi* (puntos por pulgada) de la figura.

Por convención se le asigna el nombre ```fig``` al objeto *figure*.

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.figure.html

### Objetos *figure* en celdas de *IPython/Jupyter*.

Los objetos *figure* de *Matplotlib* en celdas de *IPython/Jupyter* sólo existen mientras la celda que las creó esté activa. Si se ejecuta otra celda, la figura creada en la celda anterior se cerrará automáticamente. Para evitar esto, es necesario crear un objeto *figure* de forma explícita y luego trazar los gráficos dentro del objeto *axes* asociado a ese objeto *figure*. De esta manera, la figura permanecerá activa incluso después de ejecutar otras celdas.

**Ejemplo**:

* La siguiente celda creará un objeto *figure* con un tamaño de 15x7 pulgadas y una resolución de 50 dpi. El objeto *figure* se almacenará en la variable ```fig```.

````python
fig = plt.figure(dpi=50, figsize=(15, 7))
````

### Gestión de los objetos *figure*.

Aún cuando pueden existir varios objetos *figure* al mismo tiempo, sólo uno de ellos puede estar activo en un momento dado. El objeto *figure* activo es el que se utilizará para trazar los gráficos y las visualizaciones. Para cambiar el objeto *figure* activo, se puede utilizar el método ```plt.figure()``` con el identificador del objeto *figure* que se desea activar.

```
plt.figure(<id>)
```
Donde ```<id>``` es el identificador del objeto *figure* que se desea activar. Si el objeto *figure*   con el identificador especificado no existe, se creará un nuevo objeto *figure* con ese identificador.

Existen múltiples métodos para gestionar los objetos *figure* en *Matplotlib*, como por ejemplo:

* [`plt.gcf()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.gcf.html) que devuelve el objeto *figure* activo.
* [`plt.get_fignums()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.get_fignums.html) que devuelve una lista con los identificadores de todos los objetos *figure* existentes.
* [`plt.get_figlabels()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.get_figlabels.html) que devuelve una lista con las etiquetas de todos los objetos *figure* existentes.
* [`plt.close()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.close.html) que cierra el objeto *figure* activo o el objeto *figure* con el identificador especificado.


### Gestión sugerida de los objetos *figure*.

Es importante tener en cuenta que el uso de estos métodos puede variar dependiendo del entorno de ejecución y la configuración de *Matplotlib*. Para evitar problemas de gestión de los objetos *figure*, es recomendable crear sólo un objeto *figure* de forma explícita y luego trazar los gráficos dentro del objeto *axes* asociado a ese objeto *figure*.

## Los objetos *axes*.

Los objetos *axes* son instancias de la clase ```matplotlib.axes.Axes``` y son los que realmente contienen los gráficos y las visualizaciones. Un objeto *figure* puede contener uno o más objetos *axes*, pero cada objeto *axes* solo puede pertenecer a un objeto *figure*.

Los objetos *axes* son los que realmente contienen los gráficos y las visualizaciones, por lo que es necesario crear un objeto *axes* dentro de un objeto *figure* para poder trazar cualquier gráfico o visualización.

Por convención se le asigna el nombre ```ax``` al objeto *axes*.

### Los objetos *artist*.

Los objetos *artist* son instancias de la clase ```matplotlib.artist.Artist``` y son los que realmente representan los elementos gráficos como líneas, puntos, texto, etc. 

Por lo general no es necesario interactuar directamente con los objetos *artist*, ya que la mayoría de las funciones de *Matplotlib* se encargan de crear y gestionar estos objetos de forma automática. 

https://matplotlib.org/stable/api/artist_api.html#matplotlib.artist.Artist

### Creación de objetos *axes*.

Existen varias formas de crear objetos *axes* en *Matplotlib*, pero la forma más común es utilizando los métodos:
* ```fig.add_subplot()```, el cual permite crear un objeto *axes* dentro de un objeto *figure* especificando la posición del objeto *axes* dentro de una cuadrícula.
* ```fig.subplots()```, el cual permite crear una cuadrícula de objetos *axes* dentro de un objeto *figure* especificando el número de filas y columnas de la cuadrícula.

#### Método ```fig.add_subplot()```.
El método ```fig.add_subplot()``` permite crear un objeto *axes* dentro de un objeto *figure* especificando la posición del objeto *axes* dentro de una cuadrícula. La sintaxis es la siguiente:

```
fig.add_subplot(<nfilas>, <ncols>, <index>, **kwargs)
```
Donde:
* `<nfilas>` es el número de filas en la cuadrícula.
* `<ncols>` es el número de columnas en la cuadrícula.
* `<index>` es el índice del objeto *axes* dentro de la cuadrícula, comenzando desde 1 y contando de izquierda a derecha y de arriba hacia abajo.
* `**kwargs` son argumentos adicionales que se pasan al objeto *axes*.

**Ejemplos:**

* La siguiente celda:
    * Creará al objeto `fig`.
    * Creará al objeto `ax` utilizando el método `fig.add_subplot()`.
    * Trazará un gráfico de líneas utilizando el método `ax.plot()` con los arreglos `x` y `y` definidos previamente para los ejes $x$ e $y$ respectivamente.
    * Se definirá el título del gráfico utilizando el método `ax.set_title()`.
    * Se definirá la etiqueta del eje $x$ utilizando el método `ax.set_xlabel()`
    * Se definirá la etiqueta del eje $y$ utilizando el método `ax.set_ylabel()`.
    * Se definirá un recuadro de leyenda utilizando el método `ax.legend()`.

````python
fig = plt.figure(figsize=(8, 6), dpi=100)
ax = fig.add_subplot(1, 1, 1) # 1 fila, 1 columna, primer gráfico
ax.plot(x, y)
ax.set_title('Parábola')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend(['y = x^2 - 2x'])
````

* La siguiente celda creará una figura con 2 *subplots* apilados verticalmente.

````python
fig = plt.figure(figsize=(6, 8), dpi=100)
ax1 = fig.add_subplot(2, 1, 1) # 2 filas, 1 columna, primer subplot
ax1.plot(x, y)
ax1.set_title('Parabola')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.legend(['y = x^2 - 2x'])
ax2 = fig.add_subplot(2, 1, 2) # 2 filas, 1 columna, segundo subplot
ax2.plot(x, z)
ax2.set_title('Parabola con seno')
ax2.set_xlabel('x')
ax2.set_ylabel('z')
ax2.legend(['z = x^2 * (1 + sin(x))'])
````

* La siguiente celda desplegará 3 objetos `axes`. Dos de ellos estaán en dos columnas en la parte superior y uno de ellos ocupará todo el espacio inferior. 

````python
fig = plt.figure(figsize=(8, 6), dpi=100)
ax1 = fig.add_subplot(2, 2, 1)
ax1.plot(x, y)
ax1.set_title('Parabola')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.legend(['y = x^2 - 2x'])
ax2 = fig.add_subplot(2, 2, 2) 
ax2.plot(x, z)
ax2.set_title('Parabola with Sine')
ax2.set_xlabel('x')
ax2.set_ylabel('z')
ax2.legend(['z = x^2 * (1 + sin(x))'])
ax3=fig.add_subplot(2, 1, 2)
ax3.plot(x, y, x, z)
ax[1, 0].set_title('Parabola and Parabola with Sine')
ax[1, 0].set_xlabel('x')
ax[1, 0].set_ylabel('y and z')
ax[1, 0].legend(['y = x^2 - 2x', 'z = x^2 * (1 + sin(x))'])
````

#### El método ```fig.subplots()```.
El método ```fig.subplots()``` permite crear una cuadrícula de objetos *axes* dentro de un objeto *figure* especificando el número de filas y columnas de la cuadrícula. La sintaxis es la siguiente:
```
fig.subplots(nrows, ncols)
```
Donde:
* `nrows` es el número de filas en la cuadrícula.
* `ncols` es el número de columnas en la cuadrícula.
El método `fig.subplots()` devuelve una tupla que contiene los objetos *axes* creados. Si se crea una cuadrícula de un solo objeto *axes*, el método devolverá directamente el objeto *axes* en lugar de una tupla.

Cada objeto *axes* creado con el método `fig.subplots()` se puede personalizar de forma independiente utilizando los métodos de los objetos *axes* descritos anteriormente.

#### Posición de los subplots.

El objeto *ax* creado con el método subplot contiene un arreglo de objetos *axes* que corresponden a cada uno de los subplots creados. Para acceder a cada uno de los objetos *axes* individuales, se puede utilizar la indexación del arreglo.

```
ax[<fila>, <columna>]
```
Donde `<fila>` es el índice de la fila del subplot y `<columna>` es el índice de la columna del subplot. La indexación comienza desde 0, por lo que el primer subplot se encuentra en la posición `[0, 0]`, el segundo subplot en la posición `[0, 1]`, y así sucesivamente.

* En caso de que la cuadrícula sea de una sola fila o una sola columna, el método `fig.subplots()` devolverá un arreglo unidimensional de objetos *axes* en lugar de una tupla, por lo que la indexación se realizará utilizando un solo índice.

* En caso de que la cuadrícula sea de un solo subplot, el método `fig.subplots()` devolverá directamente el objeto *axes* en lugar de una tupla, por lo que no será necesario utilizar la indexación para acceder al objeto *axes*.


**Ejemplos**:

* La siguiente celda creará un objeto `fig` y una cuadrícula de objetos `ax` utilizando el método `fig.subplots()` con una columna de dos filas.
* Trazará un gráfico de líneas en cada objeto `ax` utilizando los arreglos `x`, `y` y `z` definidos previamente para los ejes $x$ e $y$ respectivamente. 
* Finalmente, se definirán los títulos, etiquetas de los ejes y leyendas para cada gráfico utilizando los métodos correspondientes de los objetos `ax`.

````python
fig = plt.figure(figsize=(8, 6), dpi=100)
ax = fig.subplots(2, 1)
ax[0].plot(x, y)
ax[0].set_title('Parábola')
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
ax[0].legend(['y = x^2 - 2x'])
ax[1].plot(x, z)
ax[1].set_title('Parábola con Seno')
ax[1].set_xlabel('x')
ax[1].set_ylabel('z')
ax[1].legend(['z = x^2 * (1 + sin(x))'])
````

* La siguiente celda creará una figura con 2 columnas y una fila de subplots.

````python
fig = plt.figure(figsize=(10, 4), dpi=100)
ax = fig.subplots(1, 2)
ax[0].plot(x, y)
ax[0].set_title('Parabola')
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
ax[0].legend(['y = x^2 - 2x'])  
ax[1].plot(x, z)
ax[1].set_title('Parabola with Sine')
ax[1].set_xlabel('x')
ax[1].set_ylabel('z')
ax[1].legend(['z = x^2 * (1 + sin(x))'])
````

* La siguiente celda creará una figura que contiene 2 filas y 2 columnas de subplots.

````python
fig = plt.figure(figsize=(8, 6), dpi=100)
ax = fig.subplots(2, 2)
ax[0, 0].plot(x, y)
ax[0, 0].set_title('Parabola')
ax[0, 0].set_xlabel('x')
ax[0, 0].set_ylabel('y')
ax[0, 0].legend(['y = x^2 - 2x'])  
ax[0, 1].plot(x, z)
ax[0, 1].set_title('Parabola with Sine')
ax[0, 1].set_xlabel('x')
ax[0, 1].set_ylabel('z')
ax[0, 1].legend(['z = x^2 * (1 + sin(x))'])
ax[1, 0].plot(x, y, x, z)
ax[1, 0].set_title('Parabola and Parabola with Sine')
ax[1, 0].set_xlabel('x')
ax[1, 0].set_ylabel('y and z')
ax[1, 0].legend(['y = x^2 - 2x', 'z = x^2 * (1 + sin(x))'])
````

### El método ```plt.plot()```.

El método `plt.plot()` es uno de los métodos más utilizados en *Matplotlib* para trazar gráficos de líneas, y ofrece una gran variedad de opciones de personalización a través de los argumentos adicionales.

El método `plt.plot()` se basa en el método `ax.plot()` de los objetos *axes*, pero ofrece una interfaz de alto nivel para trazar gráficos de líneas de forma rápida y sencilla, sin necesidad de interactuar directamente con los objetos *figure* y *axes*.

Este método se inspira en la sintaxis de la función `plot()` de *MATLAB*, lo que facilita su uso para aquellos que están familiarizados con esa biblioteca.

```
plt.plot(x, y, scalex=<bool>, scaley=<bool>, data=<datos>, **kwargs)
```
Donde:
* `x` es el conjunto de datos para el eje $x$.
* `y` es el conjunto de datos para el eje $y$.
* `scalex` es un valor booleano que indica si se deben escalar los límites del eje $x$ para ajustarse a los datos proporcionados. El valor por defecto es `True`.
* `scaley` es un valor booleano que indica si se deben escalar los límites del eje $y$ para ajustarse a los datos proporcionados. El valor por defecto es `True`.
* `data` es un conjunto de datos opcional que se puede utilizar para proporcionar los datos para el eje $x$ y el eje $y$ utilizando nombres de columnas en lugar de pasar los datos directamente como argumentos `x` e `y`.
* `**kwargs` son argumentos adicionales que se pasan a la función de trazado, como por ejemplo el color, el estilo de línea, el ancho de línea, etc.

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

### Personalización de los gráficos de líneas.
El método `plt.plot()` ofrece una gran variedad de opciones de personalización a través de los argumentos adicionales, como por ejemplo:
* `color` para especificar el color de la línea.
* `linestyle` para especificar el estilo de línea (sólida, discontin
ua, punteada, etc.).
* `linewidth` para especificar el ancho de la línea.
* `marker` para especificar el marcador utilizado para los puntos de datos (círculo, cuadrado, triángulo, etc.).
* `markersize` para especificar el tamaño del marcador.
* `label` para especificar la etiqueta de la línea, que se utilizará en la leyenda del gráfico.
* `alpha` para especificar la transparencia de la línea.

Por motivos de espacio, sólo se han mencionado algunos de los argumentos adicionales disponibles para personalizar los gráficos de líneas con el método `plt.plot()`. Se recomienda consultar la documentación oficial de *Matplotlib* para obtener una lista completa de los argumentos adicionales disponibles y su descripción detallada.

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot


**Ejemplos:**

* La siguiente celda trazará un gráfico mediante el método ```plt.plot()``` con los argumentos básicos.
   * EL trazo se realizará utilizando los arreglos `x` y `y` definidos previamente para los ejes $x$ e $y$ respectivamente.
   * Se utilizarán los parámetros por defecto para el color, estilo de línea, ancho de línea, etc.

````python
plt.figure(dpi=80)
plt.title("Pythonista")
plt.plot(x, y)
````

* La siguiente celda trazará una línea con diversos argumentos para personalizar el gráfico, como el color, estilo de línea, ancho de línea, etc. 

````python
plt.figure(dpi=80)
plt.title("Pythonista")
plt.plot(x, y,
         marker='x', # Tipo de marcador
         markersize=3, # Tamaño del marcador
         markeredgecolor='red', # Color del borde del marcador
         markerfacecolor='blue', # Color de relleno del marcador
         linestyle='--', # Estilo de línea (guiones)
         linewidth=2, # Ancho de la línea 
         color='green', # Color de la línea
         alpha=0.5, # Transparencia de la línea
         label='y = x^2 - 2x')
````

## Solapamientos entre los elementos del gráfico.
En ocasiones, los elementos del gráfico pueden solaparse entre sí, lo que puede dificultar la visualización de los datos y la interpretación de los valores en los ejes. Para evitar este problema, se puede utilizar el método ```plt.tight_layout()```, el cual ajusta automáticamente el diseño de la figura para evitar solapamientos entre los elementos del gráfico.

```plt.tight_layout(pad=<margen>, w_pad=<margen_horizontal>, h_pad=<margen_vertical>, **kwargs)```

Donde:
* `<margen>` es el margen general que se desea establecer entre los elementos del gráfico, el cual se puede especificar como un número entero o un número de punto flotante. El valor por defecto es `1.08`, lo que significa que se establecerá un margen del 8% del tamaño de la figura entre los elementos del gráfico.
* `<margen_horizontal>` es el margen horizontal que se desea establecer entre los elementos del gráfico, el cual se puede especificar como un número entero o un número de punto flotante. El valor por defecto es `0.2`, lo que significa que se establecerá un margen del 20% del tamaño de la figura entre los elementos del gráfico en la dirección horizontal.
* `<margen_vertical>` es el margen vertical que se desea establecer entre los elementos del gráfico, el cual se puede especificar como un número entero o un número de punto flotante. El valor por defecto es `0.2`, lo que significa que se establecerá un margen del 20% del tamaño de la figura entre los elementos del gráfico en la dirección vertical.

**Ejemplos:**

* La siguiente celda definirá un espacio entre los elementos del gráfico utilizando el método ```plt.tight_layout()``` con los argumentos:
   * ```pad=0.7```
   * ```w_pad=0.7```
   * ```h_pad=0.7```.

````python
plt.figure(dpi=80)
plt.title("Pythonista")
plt.plot(x, y, label='función 1')
plt.plot(x, z, label='función 2')
plt.legend(loc="lower right")
plt.xlabel('Eje de las x')
plt.ylabel('f(x)')
plt.tight_layout(pad=0.7, w_pad=0.7, h_pad=0.7) # Ajusta automáticamente el diseño de la figura para evitar solapamientos entre los elementos del gráfico
````

## Estilos.
Los estilos en *Matplotlib* permiten personalizar la apariencia de los gráficos de forma rápida y sencilla, sin necesidad de configurar manualmente cada uno de los elementos del gráfico. *Matplotlib* ofrece una variedad de estilos predefinidos que se pueden utilizar para cambiar el aspecto de los gráficos con un solo comando.

Para utilizar un estilo predefinido en *Matplotlib*, se puede utilizar el método `plt.style.use()`, el cual permite aplicar un estilo específico a todos los gráficos que se tracen después de llamar a este método.

Una vez que se ha aplicado un estilo utilizando `plt.style.use()`, todos los gráficos que se tracen a continuación utilizarán ese estilo de forma predeterminada, lo que facilita la creación de gráficos con una apariencia consistente y atractiva.

```
plt.style.use(<estilo>)
```
Donde:
* `<estilo>` es una cadena de texto que representa el nombre del estilo que se desea aplicar a los gráficos, el cual se puede especificar utilizando los siguientes valores:
    * `'default'` para aplicar el estilo predeterminado de *Matplotlib*.
    * `'classic'` para aplicar el estilo clásico de *Matplotlib*.
    * `'ggplot'` para aplicar el estilo inspirado en la biblioteca *ggplot2* de R.
    * `'seaborn'` para aplicar el estilo inspirado en la biblioteca *Seaborn* de Python.
    * `'bmh'` para aplicar el estilo inspirado en la biblioteca *Bayesian Methods for Hackers*.
    * `'dark_background'` para aplicar un estilo con fondo oscuro.
    * `'fivethirtyeight'` para aplicar el estilo inspirado en el sitio web *FiveThirtyEight*.
    * `'grayscale'` para aplicar un estilo en escala de grises.
    * `'Solarize_Light2'` para aplicar un estilo inspirado en el tema Solarized Light.
    * `'tableau-colorblind10'` para aplicar un estilo inspirado en la paleta de colores para daltónicos de Tableau.
    * `'seaborn-darkgrid'` para aplicar un estilo inspirado en la biblioteca *Seaborn* con fondo oscuro y rejilla.
    * `'seaborn-notebook'` para aplicar un estilo inspirado en la biblioteca *Seaborn* con un diseño optimizado para cuadernos de Jupyter.
    * `'seaborn-poster'` para aplicar un estilo inspirado en la biblioteca *Seaborn* con un diseño optimizado para presentaciones.

https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html
