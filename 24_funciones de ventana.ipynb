# Window Functions (Funciones de Ventana).

````python
import pandas as pd
import numpy as np
from datetime import datetime
````

Las funciones de ventana (window functions) permiten realizar cálculos sobre **rangos dinámicos de filas** en un *dataframe*, manteniendo el contexto original.

Estas funciones son útiles para realizar operaciones como:
- Cálculo de medias móviles.
- Cálculo de sumas acumulativas.
- Cálculo de diferencias entre filas.
- Cálculo de estadísticas móviles.

https://pandas.pydata.org/docs/user_guide/window.html

# Métodos de funciones de ventana.

Existen varios métodos para aplicar funciones de ventana en *Pandas*, entre los más comunes se encuentran:
- **rolling()**: Permite calcular estadísticas móviles sobre una ventana de tamaño fijo.
- **expanding()**: Permite calcular estadísticas acumulativas desde el inicio del *dataframe* hasta la fila actual.
- **ewm()**: Permite calcular estadísticas móviles exponencialmente ponderadas, donde los valores más recientes tienen un mayor peso. Este método es útil para suavizar datos y detectar tendencias a lo largo del tiempo.

### El método ```df.rolling()```.

El método ```df.rolling()``` permite realizar cálculos sobre una ventana móvil de tamaño fijo. Por ejemplo, si deseas calcular la media móvil de los montos de las facturas con una ventana de 3 filas, puedes usar el siguiente código:

```
df.rolling(window=<n>, on=<col>).<método>()
```
Donde:
* `<n>` es el tamaño de la ventana.
* `<col>` es la columna sobre la cual deseas aplicar la función de ventana.
* `<método>` es el método de agregación que deseas aplicar (por ejemplo, `mean`, `sum`, `max`, etc.).

La documenttación de `df.rolling()` se encuentra en: 

https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.rolling.html


**Ejemplo:**

* La siguiente celda calculará la media móvil de los montos de las facturas con una ventana de 3 filas y la agregará como una nueva columna llamada ```'monto_rolling_mean'``` al dataframe ```facturas```.

````python
facturas["rolling_mean"] =facturas.rolling(window=3)["monto"].mean()
facturas
````

### El método `df.expanding()`.

El método `df.expanding()` permite realizar cálculos acumulativos desde el inicio hasta la fila actual. Por ejemplo, si deseas calcular la suma acumulativa de los montos de las facturas, puedes usar el siguiente código:

```
df.expanding()[<col>].<método>()
```
Donde:
* `<col>` es la columna sobre la cual deseas aplicar la función de ventana.
* `<método>` es el método de agregación que deseas aplicar (por ejemplo, `mean`, `sum`, `max`, etc.).   

https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.expanding.html

**Ejemplo:**

* La siguiente celda calculará la suma acumulativa de los montos de las facturas y la agregará como una nueva columna llamada ```'monto_expanding_sum'``` al dataframe ```facturas```.

````python
facturas["monto_expanding_sum"] = facturas.expanding()["monto"].sum()
facturas
````

### El método `df.ewm()`.

El método `df.ewm()` permite calcular estadísticas móviles exponencialmente ponderadas, donde los valores más recientes tienen un mayor peso. Este método es útil para suavizar datos y detectar tendencias a lo largo del tiempo. Por ejemplo, si deseas calcular la media móvil exponencialmente ponderada de los montos de las facturas con un factor de suavizado de 0.5, puedes usar el siguiente código:

```
df.ewm(alpha=<alpha>, on=<col>).<método>()
```
Donde:
* `<alpha>` es el factor de suavizado (en este caso, 0.5).
* `<col>` es la columna sobre la cual deseas aplicar la función de ventana.
* `<método>` es el método de agregación que deseas aplicar (por ejemplo, `mean`, `sum`, `max`, etc.).

https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.ewm.html

**Ejemplo:**

* La siguiente celda regresará un *dataframe* con la media móvil de 3 períodos exponencialmente ponderada de los montos de las facturas con un factor de suavizado de 0.5, agregada como una nueva columna llamada ```'monto_ewm_mean'``` al dataframe ```facturas```.

````python
facturas["monto_ewm"] = facturas.ewm(span=3, adjust=False)["monto"].mean()
facturas
````

## El método `df.shift()`.

El método `df.shift()` permite desplazar las filas de un dataframe hacia arriba o hacia abajo. Esto es útil para comparar valores entre filas adyacentes o para calcular diferencias entre filas.

```
df.shift(periods=<n>, fill_value=<valor>)
```
Donde:
* `<n>` es el número de períodos que deseas desplazar (positivo para desplazar hacia abajo, negativo para desplazar hacia arriba).
* `<valor>` es el valor que se utilizará para llenar las filas vacías que resultan del desplazamiento.

**Ejemplo:**

* Es posible combinar `df.rolling()` y `df.shift()` para realizar operaciones complejas, como calcular la diferencia entre el monto de cada factura y la media móvil de los montos de las facturas anteriores.

````python
facturas['monto_diff'] = facturas['monto'] - facturas['rolling_mean'].shift(1)
facturas
````

## Combinación de funciones de ventana con agrupaciones.

Las funciones de ventana también se pueden combinar con operaciones de agrupación utilizando el método `groupby()`. Esto permite aplicar funciones de ventana dentro de cada grupo de datos.

### Tabla sintética para funciones de ventana y `groupby()`.
.

Se creará el dataframe `ventas_diarias` con la siguiente estructura:
* Una columna que corresponde a las fechas de ventas que van del 1  de enero de 2025 al 31 de diciembre de 2025.
* Columnas que representan las ventas diarias para las sucursales: 'CDMX01', 'CDMX02', 'MTY01', 'GDL01'.
* Valores aleatorios de ventas diarias para cada sucursal.
* Los domingos venden por lo general 50% mas que los días de semana, por lo que se aplicará un factor de 1.5 a las ventas de los domingos.
* Los sábados venden por lo general 20% mas que los días de semana, por lo que se aplicará un factor de 1.2 a las ventas de los sábados.
* Los lunes venden por lo general 20% menos que los días de semana, por lo que se aplicará un factor de 0.8 a las ventas de los lunes.

````python
fechas = pd.date_range('2025-01-01', '2025-12-31', freq='D', name='fecha')
columnas_ventas = ['CDMX01', 'CDMX02', 'MTY01', 'GDL01']
ventas = np.random.randint(100, 1000, size=(len(fechas), len(columnas_ventas)))
ventas[4::7] = ventas[4::7] * 1.5 # Los domingos tienen un aumento del 50% en las ventas
ventas[3::7] = ventas[3::7] * 1.2 # Los sábados tienen un aumento del 20% en las ventas
ventas[5::7] = ventas[5::7] * 0.8 # Los lunes tienen una disminución del 20% en las ventas
ventas_diarias = pd.DataFrame(ventas, columns=columnas_ventas)
ventas_diarias.insert(0, 'fecha', fechas)
ventas_diarias.head()
````

* Se añadirá la columna 'semana' que indica el número de semana del año para cada fecha.

````python
ventas_diarias["semana"] = ventas_diarias.fecha.dt.isocalendar().week
ventas_diarias.tail()
````

* La última semana del año se considerará como la semana 53, aunque en realidad solo tiene 3 días (29, 30 y 31 de diciembre).

````python
ventas_diarias["semana"] = (ventas_diarias["semana"].
                            mask((ventas_diarias["semana"] == 1) & (ventas_diarias["fecha"] >= "2025-12-01"), 53))
ventas_diarias.tail()
````

* Se agregará una columna adicional llamada 'día_semana' que indicará el día de la semana correspondiente a cada fecha (0 para lunes, 1 para martes, ..., 6 para domingo).

````python
ventas_diarias["dia_semana"] = ventas_diarias["fecha"].dt.day_of_week
ventas_diarias.tail()
````

* A continuación se mostrará la estructura del dataframe `ventas_diarias`.

````python
ventas_diarias.info()
````

### Operaciones de `df.rolling()` usando criterio de días.

El método `rolling()` también permite realizar cálculos sobre una ventana móvil basada en un criterio de tiempo, como días, semanas o meses. Esto es especialmente útil para analizar series temporales.

```
df.rolling(window='<n>D', on='<col>').<método>()
```
Donde:
* `<n>` es el número de días que deseas incluir en la ventana (por ejemplo, '7D' para una ventana de 7 días).
* `<col>` es la columna de fecha sobre la cual deseas aplicar la función de ventana.
* `<método>` es el método de agregación que deseas aplicar (por ejemplo, `mean`, `sum`, `max`, etc.).


**Ejemplo:**

* La siguiente celda calculará la media móvil de las ventas diarias para cada sucursal utilizando una ventana de 7 días, y el *dataframe* resultante se llamará `media_movil_7d`.

````python
media_movil_7d = ventas_diarias.rolling(window="7D", on='fecha')[['CDMX01', 'CDMX02', 'MTY01', 'GDL01']].mean()
media_movil_7d.head(15)
````

* La siguiente celda calculará las ventas acumulativas de todo el año para cada sucursal usando el método `expanding()` del *dataframe* `ventas_diarias`.

````python
ventas_diarias.expanding()[columnas_ventas].sum()
````

* La siguiente celda calculará el promedio de ventas semanales usando el método `resample()` en la columna `'fecha'`.

````python
ventas_diarias.resample("W", on="fecha")[columnas_ventas].mean()
````

* La siguiente celda también calculará el promedio de ventas semanales usando el método `df.groupby()` en la columna `'semana'`.

````python
ventas_diarias.groupby("semana")[columnas_ventas].mean()
````

````python
ventas_diarias.groupby("semana").rolling(window=2)[columnas_ventas].mean()
````

* La siguente celda graficará las ventas totales por día de la semana para cada sucursal usando el método `df.groupby()` en la columna `'día_semana'`.

````python
ventas_por_dia_semana = ventas_diarias.groupby("dia_semana")[columnas_ventas]
ventas_por_dia_semana.sum().plot(kind="bar", figsize=(10, 6), title="Ventas por día de la semana")
````
