# Objetos de *Seaborn*.

Aún cuando las funciones *Seaborn* son muy populares y fáciles de desarrollar, tienen ciertas desventajas con respecto a otras soluciones basadas en gramáticas de gráficas. A partir de la versión 0.12, *Seaborn* cuenta con una biblioteca de objetos que se apega a dicha gramática.


<img src="img/grammar_of_graphics.png">

Por convención, la biblioteca de objetos de seaborn ```seaborn.objects``` es importada como ```so```. En adelente, se seguirá dicha convención.

https://seaborn.pydata.org/tutorial/objects_interface.html

**NOTA:** Debido a que los objetos de *Seaborn* son una adición muy reciente, aún tiene funcionalidades limitadas en comparación con las funciones.

````python
import seaborn as sns
from seaborn import objects as so
````

## La clase ```so.Plot ```.

Los componentes principales de este tipo de visualizaciones son los objetos instanciados de la clase ```so.Plot```.

```
so.Plot(data=<datos>, x=<x>, y=<y>, <args>)
```

Donde:

* ```<datos>``` es un *dataset* compatible con un *dataframe* de *Pandas*.
* ```<x>``` es el identificador de la columna del ```<dataset>``` que se utilizará para el eje de las $x$ en caso de que se requiera.
* ```<y>``` es el identificador de la columna del ```<dataset>``` que se utilizará para el eje de las $y$ en caso de que se requiera.


Los objetos instanciados de ```so.Plot``` cuentan con varios métodos y atributos que permiten cumplir con la funcionalidades de las capas:

* Datos.
* Estética.
* Escala.
* Facetas.

https://seaborn.pydata.org/generated/seaborn.objects.Plot.html

### Métodos en cascada.

Los métodos de loss objetos instanciados de ```so.Plot``` también regresan objetos instanciados de ```so.Plot```, por lo que es posible aplicar métodos en cascada.

### El método ```so.Plot.add() ```.

Este método permite añadir funcionalidades que extienden a los objeto de tipo ```so.Plot.add()``` en las capas de objetos geométricos, y estadísitca, principalmente.

```
so.Plot.add(<func_1>(), <func_2>(), ... <func_n>())
```

Donde:

* Cada ```<func_i>```es una función de ```seaborn.objects```.

**Ejemplo:**

````python
dataset = dataset = sns.load_dataset("iris")
dataset
````

````python
so.Plot(data=dataset, x="sepal_length", y="sepal_width")
````

````python
(so.Plot(data=dataset, x="sepal_length", y="sepal_width")
 .add(so.Dots()))
````

````python
(so.Plot(data=dataset, x="sepal_length", y="sepal_width", color="species")
 .add(so.Dots()))
````

````python
(so.Plot(data=dataset, x='sepal_length',
         y='sepal_width')
 .facet('species'))
````

````python
(so.Plot(data=dataset, x='sepal_length', color='sepal_length').
 facet('species').
 add(so.Bar(), so.Hist()))
````

````python
(so.Plot(data=dataset, x='sepal_length', color='sepal_length').
 facet('species').
 add(so.Bar(), so.Hist()).scale(x=1))
````

````python
(so.Plot(data=dataset, x='sepal_length', color='sepal_length').
 add(so.Bar(), so.S()))
````

## Limitaciones actuales.

A pesar de sus ventajas, la API de objetos aún tiene limitaciones en comparación con las funciones tradicionales de *Seaborn*:

- **Funcionalidad limitada**: No todos los tipos de gráficos disponibles en funciones están disponibles en la API de objetos.
- **Parámetros reducidos**: Algunos parámetros de personalización avanzada aún no están completamente soportados.
- **Documentación en evolución**: Siendo una característica relativamente nueva, la documentación y ejemplos aún están en desarrollo.

Se recomienda conocer ambas APIs y elegir la más apropiada según las necesidades específicas del proyecto.

````python
# Comparación de todas las variables por especie
(so.Plot(data=dataset.melt(id_vars='species', 
                          var_name='medida', 
                          value_name='valor'),
         x='medida', y='valor', color='species')
 .add(so.Dots(alpha=0.4))
 .add(so.Line(), so.Est())
 .label(y='Medida (cm)', x='Variable', 
        color='Especie', title='Análisis comparativo por especie'))
````

## Caso de uso: Análisis exploratorio comparativo.

A continuación se muestra un ejemplo completo que demuestra las capacidades de la API de objetos para un análisis exploratorio completo.

````python
# Guardar un gráfico
plot_final = (so.Plot(data=dataset, x='sepal_length', y='sepal_width', color='species')
              .add(so.Dots(alpha=0.7))
              .label(x='Longitud del sépalo (cm)', y='Ancho del sépalo (cm)',
                     color='Especie', title='Características del iris'))

# Para guardar (no ejecutar en notebook):
# plot_final.save("iris_plot.png", dpi=300)

# Mostrar el gráfico
plot_final
````

## Guardado y exportación.

Los gráficos creados con objetos de *Seaborn* pueden ser guardados como archivos de imagen utilizando el método ```save()```.

````python
# Ejemplo con operaciones de pandas
dataset['sepal_ratio'] = dataset['sepal_length'] / dataset['sepal_width']

(so.Plot(data=dataset, x='sepal_ratio', y='petal_length', color='species', 
         pointsize='petal_width')
 .add(so.Dots(alpha=0.7))
 .scale(pointsize=(30, 200), color='Set2')
 .label(x='Proporción sepal (L/W)', y='Longitud del pétalo', 
        color='Especie', pointsize='Ancho del pétalo'))
````

## Integración con datasets de Pandas.

Los objetos de *Seaborn* están diseñados para trabajar directamente con dataframes de *Pandas*, permitiendo referencias simples a columnas.

### Consistencia en la API

Todos los métodos siguen patrones similares, facilitando el aprendizaje y la predicción del comportamiento.

### Gramática de gráficas

Sigue el paradigma "Grammar of Graphics" que simplifica el razonamiento sobre visualizaciones de datos complejas.

````python
# Composición paso a paso
plot = (so.Plot(data=dataset, x='sepal_length', y='sepal_width')
        .add(so.Dots(alpha=0.6, color='steelblue')))

# Extender el gráfico con nuevas capas
plot_con_regresion = (plot
                      .add(so.Line(color='red'), so.PolyFit(order=1))
                      .label(x='Longitud del sépalo', y='Ancho del sépalo'))

plot_con_regresion
````

## Ventajas de la API orientada a objetos.

La API de objetos de *Seaborn* presenta ventajas significativas respecto a las funciones tradicionales:

### Composición flexible

Permite encadenar métodos de forma clara y legible para construir visualizaciones complejas paso a paso.

````python
# Limitar el rango de los ejes
(so.Plot(data=dataset, x='sepal_length', y='sepal_width', color='species')
 .add(so.Dots())
 .limit(x=(4, 8), y=(2, 4.5))
 .label(x='Longitud del sépalo', y='Ancho del sépalo'))
````

## Limitación de ejes y coordenadas.

El método ```limit()``` permite definir rangos específicos para los ejes.

````python
# Línea de regresión
import numpy as np

(so.Plot(data=dataset, x='sepal_length', y='sepal_width', color='species')
 .add(so.Dots(alpha=0.5))
 .add(so.Line(), so.PolyFit(order=1))
 .label(x='Longitud del sépalo', y='Ancho del sépalo',
        title='Relación lineal por especie'))
````

## Aplicación de estadísticas.

El método ```add()``` puede combinar funciones geométricas con funciones estadísticas para calcular resumenes de datos.

````python
# Histograma por especie
(so.Plot(data=dataset, x='sepal_length', color='species')
 .add(so.Hist(bins=15, alpha=0.5))
 .label(x='Longitud del sépalo', title='Distribución por especie'))
````

### Histogramas y áreas.

- **`so.Hist()`**: Histogramas de distribuciones.
- **`so.Area()`**: Áreas bajo curvas.

````python
# Gráfico de barras con agregación
dataset_agg = dataset.groupby('species')[['sepal_length', 'sepal_width']].mean()
(so.Plot(data=dataset_agg.reset_index(), x='species', y='sepal_length')
 .add(so.Bar())
 .label(x='Especie', y='Longitud promedio del sépalo'))
````

### Barras.

- **`so.Bar()`**: Barras para comparación de valores.

````python
# Gráfico de líneas
(so.Plot(data=dataset.sort_values('sepal_length'), 
         x='sepal_length', y='sepal_width', color='species')
 .add(so.Line()))
````

### Puntos y líneas.

- **`so.Dots()`**: Puntos de dispersión.
- **`so.Line()`**: Líneas que conectan puntos.
- **`so.Path()`**: Caminos o trayectorias conectadas.

## Funciones geométricas disponibles.

*Seaborn* proporciona varias funciones geométricas para representar datos de diferentes formas:

````python
# Facetas por filas y columnas
(so.Plot(data=dataset, x='sepal_length', y='sepal_width')
 .add(so.Dots(color='steelblue'))
 .facet(col='species', row=dataset['petal_width'] > dataset['petal_width'].median())
 .label(x='Longitud del sépalo', y='Ancho del sépalo'))
````

````python
# Facetas por columnas
(so.Plot(data=dataset, x='sepal_length', y='sepal_width', color='species')
 .add(so.Dots())
 .facet(col='species')
 .label(x='Longitud del sépalo', y='Ancho del sépalo'))
````

## Facetas avanzadas.

El método ```facet()``` permite crear múltiples gráficos separados por categorías de variable(s), similar a los subplots en matplotlib pero de forma más intuitiva.

````python
(so.Plot(data=dataset, x='sepal_length', y='sepal_width', 
         color='petal_length', pointsize='petal_width')
 .add(so.Dots())
 .scale(color='viridis', pointsize=(50, 300)))
````

## Métodos de escala.

El método ```scale()``` permite ajustar escalas de variables visuales como tamaño, color, opacidad, etc.

````python
(so.Plot(data=dataset, x='sepal_length', y='sepal_width', color='species')
 .add(so.Dots())
 .label(x='Longitud del sépalo (cm)', 
        y='Ancho del sépalo (cm)',
        color='Especie',
        title='Relación entre dimensiones del sépalo'))
````

## Personalización de títulos y etiquetas.

El método ```label()``` permite definir títulos personalizados para los ejes y la figura, así como para las leyendas.
