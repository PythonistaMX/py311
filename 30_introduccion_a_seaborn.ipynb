# Introducción a *Seaborn*.

[*Seaborn*](https://seaborn.pydata.org/) es una biblioteca de funciones especializadas en visualizaciones estadísticas de datos. A partir de la versión 0.12, *Seaborn* cuenta con una biblioteca de objetos.

**Nota:** Por convención, el paquete ```seaborn``` es nombrado como ```sns```.

https://seaborn.pydata.org/api.html

https://seaborn.pydata.org/tutorial.html

````python
%pip install seaborn
````

La siguiente celda importa la librería `seaborn` con el alias convencional `sns`.

````python
import seaborn as sns
````

## Acceso a *datasets* de muestra de *Seaborn*.

*Seaborn* cuenta con un repositorio de *datasets* muy conocidos y de uso común para fines didácticos. El repositorio de *Github* de dichos *datasets* es el siguiente:

https://github.com/mwaskom/seaborn-data

### La función ```seaborn.load_dataset()```.

Para acceder a algún *dataset* de *Seaborn* se utiliza la función ```seaborn.load_dataset()```.

```
sns.load_dataset(<dataset>)
```

Donde:

* ```<dataset>``` es una cadena de caracteres con el nombre del *dataset* al que se quiere acceder.

**Ejemplo:**

* La siguiente celda carga el conjunto de datos de ejemplo "iris" utilizando la función `load_dataset` de `seaborn`.

````python
sns.load_dataset("iris")
````

## El *dataset* "iris".

El conjunto de datos "iris" es un conjunto de datos clásico en el campo del aprendizaje automático y la estadística, que contiene información sobre las características de diferentes especies de flores de iris.

El conjunto de datos "iris" consta de 150 filas y 5 columnas. Cada fila representa una flor de iris, y las columnas contienen las siguientes características:
1. **sepal_length**: La longitud del sépalo de la flor (en centímetros).
2. **sepal_width**: El ancho del sépalo de la flor (en centímetros).
3. **petal_length**: La longitud del pétalo de la flor (en centímetros).
4. **petal_width**: El ancho del pétalo de la flor (en centímetros).
5. **species**: La especie de la flor de iris (setosa, versicolor o virginica).

A lo largo de este capítulo, utilizaremos el conjunto de datos "iris" para ilustrar cómo se pueden crear visualizaciones estadísticas utilizando la biblioteca *Seaborn*.

* La siguiente celda carga el conjunto de datos "iris" utilizando la función `load_dataset` de `seaborn` y lo asigna a la variable `dataset`.

````python
dataset = sns.load_dataset("iris")
````

## Funciones de *Seaborn*.

Las funciones de visualización de *Seaborn* reciben como primer argumento un conjunto de datos (*dataset*), con una estructura de *dataframe* y los siguientes argumentos definen la columnas que serán usadas para trazar la visualización.

```
sns.<func>(data=<dataset>, x=<x>, y=<y>, <kwargs>)
```

Donde:

* ```<func>``` es una función de *Seaborn*.
* ```<dataset>```es un conjunto de datos en formato compatible con un *dataframe* de *Pandas*.
* ```<x>``` es el identificador de la columna del ```<dataset>``` que se utilizará para el eje de las $x$ en caso de que se requiera.
* ```<y>``` es el identificador de la columna del ```<dataset>``` que se utilizará para el eje de las $y$ en caso de que se requiera.

https://seaborn.pydata.org/tutorial/function_overview.html

### Estructuras tabulares optimizadas para *Seaborn*.

*Seaborn* es una biblioteca altamente especializada que está optimizada para cierto tipo de estructuras de tabla (*dataframes*) en las que las columnas corresponden a cierta variable o característica que se desea visualizar y los renglones de la columna son los datos que se utilizarán para graficar.

https://seaborn.pydata.org/tutorial/data_structure.html#long-form-vs-wide-form-data

## Tipos de gráficos en Seaborn.

### Gráficos relacionales.

Los gráficos relacionales permiten visualizar la relación estadística entre dos o más variables numéricas. Las funciones principales son:

* [`scatterplot`](https://seaborn.pydata.org/generated/seaborn.scatterplot.html) la cual traza un gráfico de dispersión entre dos variables numéricas.
* [`lineplot`](https://seaborn.pydata.org/generated/seaborn.lineplot.html) la cual traza un gráfico de líneas entre dos variables numéricas.

**Ejemplos:**

* La siguiente celda genera un gráfico de dispersión (`scatterplot`) que muestra la relación entre la longitud del sépalo y la longitud del pétalo, diferenciando las especies por color.

````python
sns.scatterplot(data=dataset, x="sepal_length", y="petal_length", hue="species")
````

* La siguiente celda carga el conjunto de datos "flights" y genera un gráfico de líneas (`lineplot`) para visualizar la evolución del número de pasajeros a lo largo de los años.

````python
flights = sns.load_dataset("flights")
flights
````

````python
sns.lineplot(data=flights, x="year", y="passengers")
````

### Gráficos categóricos.

Seaborn ofrece funciones especializadas para visualizar distribuciones de variables con respecto a categorías. Las funciones principales son:
* [`boxplot`](https://seaborn.pydata.org/generated/seaborn.boxplot.html) la cual traza un diagrama de caja para visualizar la distribución de una variable numérica con respecto a una variable categórica.
* [`violinplot`](https://seaborn.pydata.org/generated/seaborn.violinplot.html) la cual traza un diagrama de violín para visualizar la distribución de una variable numérica con respecto a una variable categórica.
* [`barplot`](https://seaborn.pydata.org/generated/seaborn.barplot.html) la cual traza un gráfico de barras para visualizar la media de una variable numérica con respecto a una variable categórica.

**Ejemplos:**

* La siguiente celda genera un diagrama de barras (`barplot`) mostrando el promedio de la longitud del sépalo por especie.

````python
sns.barplot(data=dataset, x="species", y="sepal_length")
````

* La siguiente celda crea un diagrama de caja (`boxplot`) para visualizar la distribución de la longitud del sépalo por especie.

````python
sns.boxplot(data=dataset, x="species", y="sepal_length")
````

* La siguiente celda genera un diagrama de violín (`violinplot`) que combina la caja con una estimación de densidad.

````python
sns.violinplot(data=dataset, x="species", y="sepal_length")
````

### Gráficos de distribución.

Estos gráficos permiten examinar distribuciones univariadas o bivariadas. Incluye histogramas y gráficos de densidad (KDE).

* [`histplot`](https://seaborn.pydata.org/generated/seaborn.histplot.html) la cual traza un histograma para visualizar la distribución de una variable numérica.
* [`kdeplot`](https://seaborn.pydata.org/generated/seaborn.kdeplot.html) la cual traza un gráfico de densidad para visualizar la distribución de una variable numérica.

**Ejemplos:**

* La siguiente celda crea un histograma apilado (`histplot`) de la longitud del sépalo, coloreado por especie.

````python
sns.histplot(data=dataset, x="sepal_length", hue="species", multiple="stack")
````

* La siguiente celda genera un gráfico de densidad de kernel (`kdeplot`) para la longitud del sépalo, diferenciando por especie.

````python
sns.kdeplot(data=dataset, x="sepal_length", hue="species", fill=True)
````

### Gráficos matriciales.

Permiten visualizar datos en forma de matriz de color. La función principal es [```heatmap```](https://seaborn.pydata.org/generated/seaborn.heatmap.html).

**Ejemplo:**

* La siguiente celda crea un mapa de calor (`heatmap`) de la matriz de correlación de las variables numéricas del conjunto de datos iris.

````python
corr = dataset.select_dtypes('number').corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
````

## Niveles de visualización.

* **Visualizaciones a nivel de ejes**, las cuales son trazadas dentro de un objeto [```matplotlib.pyplot.Axes```](https://matplotlib.org/stable/api/axes_api.html).
* **Visualizaciones a nivel de figuras**, las cuales permiten trazar múltiple elementos dentro de un objeto [```matplotlib.pyplot.figure```](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.figure.html) mediante el objetos ```seaborn.FacetGrid```.

https://seaborn.pydata.org/tutorial/function_overview.html#figure-level-vs-axes-level-functions

### Los objetos ```seaborn.FacetGrid```.

Los objetos instanciados de ```seaborn.FacetGrid``` permiten crear *subplots* de forma sencilla.

```
sns.FacetGrid(<dataset>, col=<col>, row=<row>)
```

Donde:

* ```<dataset>```es un conjunto de datos en formato compatible con un *dataframe* de *Pandas*.
* ```<col>``` es el identificador de la columna del ```<dataset>``` que se utilizará para clasificar las columnas de los *subplots*.
* ```<row>``` es el identificador de la columna del ```<dataset>``` que se utilizará para clasificar las fila de los *subplots*.

https://seaborn.pydata.org/generated/seaborn.FacetGrid.html#seaborn-facetgrid

#### El método ```seaborn.FacetGrid.map()```.

El método ```seaborn.FacetGrid.map()``` permite añadir una función de *Seaborn* a los *subplot* de las instancias de ```seaborn.FacetGrid```.


```
<fg>.map(<func>, <col_1>, <col_2>, .. <col_n>)
```

Donde:

* ```<fg>```es una instancia de ```seaborn.FacetGrid```.
* ```<func>```es una función de *Seaborn*.
* Cada ```<col>``` es el identificador de columna que se quiera usar con ```<func>```.

https://seaborn.pydata.org/generated/seaborn.FacetGrid.map.html#seaborn.FacetGrid.map

**Ejemplo:**

* La siguiente celda muestra nuevamente el contenido del *dataframe* `dataset` para referencia.

````python
dataset
````

* La siguiente celda crea una rejilla de gráficos (`FacetGrid`) vacía, clasificando los datos por la columna `"species"`.

````python
sns.FacetGrid(dataset, col="species")
````

* La siguiente celda utiliza `map` sobre el objeto `FacetGrid` para trazar un histograma (`histplot`) de la longitud del pétalo frente a la longitud del sépalo para cada especie.

````python
g = sns.FacetGrid(dataset, col="species")
g.map(sns.histplot, "petal_length", "sepal_length")
````

### Los objetos ```seaborn.PairGrid```.

Son subclases de ```seaborn.FacetGrid```, que permiten mapear columnas de un *dataset* por pares.


```
sns.PairGrid(<dataset>)
```

Donde:

* ```<dataset>```es un conjunto de datos en formato compatible con un *dataframe*.

https://seaborn.pydata.org/generated/seaborn.PairGrid.html

#### Métodos de mapeo para objetos ```seaborn.PairGrid```.

Lo objetos instanciados de ```seaborn.PairGrid``` pueden tener los siguientes métodos de mapeo de funciones.

* [```seaborn.PairGrid.map()```](https://seaborn.pydata.org/generated/seaborn.PairGrid.map.html), el cual mapea una función de *Seaborn* a todos los *subplots*.
* [```seaborn.PairGrid.map_diag()```](https://seaborn.pydata.org/generated/seaborn.PairGrid.map_diag.html), el cual mapea una función de *Seaborn* a los *subplots* diagonales.
* [```seaborn.PairGrid.map_offdiag()```](https://seaborn.pydata.org/generated/seaborn.PairGrid.map_offdiag.html), el cual mapea una función de *Seaborn* a los *subplots* que no sean diagonales.
* [```seaborn.PairGrid.map_upper()```](https://seaborn.pydata.org/generated/seaborn.PairGrid.map_upper.html), el cual mapea una función de *Seaborn* a los *subplots* por encima de la diagonal.
* [```seaborn.PairGrid.map_lower()```](https://seaborn.pydata.org/generated/seaborn.PairGrid.map_lower.html), el cual mapea una función de *Seaborn* a los *subplots* por debajo de la diagonal.


```
<pg>.<método>(<func>, <col_1>, <col_2>, .. <col_n>)
```

Donde:

* ```<fg>```es una instancia de ```seaborn.PairGrid```.
* ```<método>``` es alguno de los métodos enumerados.
* ```<func>```es una función de *Seaborn*.

**Ejemplos:**

* La siguiente celda muestra el contenido del *dataframe* `dataset`.

````python
dataset
````

* La siguiente celda inicializa un objeto `PairGrid` con el conjunto de datos, preparando una rejilla para comparar pares de variables.

````python
sns.PairGrid(dataset)
````

* La siguiente celda mapea un gráfico de dispersión (`scatterplot`) a todos los subgráficos de la rejilla `PairGrid`.

````python
g = sns.PairGrid(dataset)
g.map(sns.scatterplot)
````

* La siguiente celda personaliza el `PairGrid` utilizando gráficos de densidad (`kdeplot`) en la diagonal y gráficos de dispersión (`scatterplot`) fuera de ella.

````python
g = sns.PairGrid(dataset)
g.map_offdiag(sns.scatterplot)
g.map_diag(sns.kdeplot)
````

* La siguiente celda crea una visualización compleja con `PairGrid`: dispersión en el triángulo superior, histogramas en el inferior y densidad en la diagonal.

````python
g = sns.PairGrid(dataset)
g.map_upper(sns.scatterplot)
g.map_lower(sns.histplot)
g.map_diag(sns.kdeplot)
````

### La función ```seaborn.jointplot()```

Esta función permite crear un gráfico combinado que permite visualizar una función de *Seaborn* con 2 argumentos y cada eje mostrará dicha función de *Seaborn* para cada argumento.


```
sns.jointplot(data=<dataset>, x=<eje_x>, y=<eje_y>, kind=<func>, hue=<categoría>)
```
Donde:

* ```<dataset>```es un conjunto de datos en formato compatible con *dataframe*.
* ```<func>``` es una cadena de caracteres que define a la función de *Seaborn* que se aplicará.
* ```<x>``` es el identificador de la columna del ```<dataset>``` que se utilizará para el eje de las $x$ en caso de que se requiera.
* ```<y>``` es el identificador de la columna del ```<dataset>``` que se utilizará para el eje de las $y$ en caso de que se requiera.
* ```<categoría>``` es el identificador de la columna del ```<dataset>``` que se utilizará para categorizar lo datos.


https://seaborn.pydata.org/generated/seaborn.jointplot.html#seaborn.jointplot

**Ejemplo:**

* La siguiente celda muestra el *dataframe* `dataset`.

````python
dataset
````

* La siguiente celda genera un gráfico conjunto (`jointplot`) de tipo histograma, diferenciando las especies por color.

````python
sns.jointplot(data=dataset, x="petal_length", y="petal_width", kind="hist", hue="species")
````

* La siguiente celda crea un `jointplot` simple de tipo histograma sin diferenciación por categorías.

````python
sns.jointplot(data=dataset, x="petal_length", y="petal_width", kind="hist")
````

## Configuración de la estética.

Por lo general, los gráficos son compartidos y publicados en diferentes contextos, por lo que es importante poder configurar la estética de las visualizaciones para adaptarlas a cada contexto.

*Seaborn* cuenta con varias opciones de configuración de la estética de las visualizaciones. Las funciones principales son:
* [`set_style`](https://seaborn.pydata.org/generated/seaborn.set_style.html) la cual establece el estilo de fondo de las visualizaciones.
* [`set_palette`](https://seaborn.pydata.org/generated/seaborn.set_palette.html) la cual establece la paleta de colores para las visualizaciones.
* [`set_context`](https://seaborn.pydata.org/generated/seaborn.set_context.html) la cual establece el contexto de las visualizaciones, ajustando el tamaño de los elementos gráficos.

https://seaborn.pydata.org/tutorial/aesthetics.html

### Paletas de colores.

Las paletas de colores son conjuntos de colores predefinidos que se pueden utilizar para mejorar la apariencia de las visualizaciones. *Seaborn* ofrece varias paletas de colores, como las paletas de colores secuenciales, divergentes y cualitativas.

La función ```seaborn.set_palette()``` permite establecer la paleta de colores para las visualizaciones.

```
sns.set_palette(<palette>)
```
Donde:
* ```<palette>``` es el nombre de la paleta de colores que se desea utilizar.

Las paletas de colores disponibles en *Seaborn* se pueden consultar en la siguiente página:

https://seaborn.pydata.org/tutorial/color_palettes.html

### Contexto.

El contexto de una visualización se refiere al entorno en el que se va a presentar la visualización, lo que puede influir en la elección de los tamaños y estilos de los elementos gráficos. *Seaborn* ofrece diferentes contextos predefinidos para adaptar las visualizaciones a diferentes situaciones.

La función ```seaborn.set_context()``` permite establecer el contexto de las visualizaciones.

```sns.set_context(<context>)
```
Donde:
* ```<context>``` es el nombre del contexto que se desea utilizar. Los contextos disponibles en *Seaborn* son los siguientes:
 * ```'notebook'```, usado para visualizaciones dentro de un cuaderno de Jupyter.
* ```'paper'```, usado para visualizaciones destinadas a ser impresas en papel.
* ```'talk'```, usado para presentaciones orales.
* ```'poster'```, usado para pósters y presentaciones grandes.

https://seaborn.pydata.org/generated/seaborn.set_context.html#seaborn.set_context

### Estilos.

El estilo de una visualización se refiere a la apariencia general de los elementos gráficos, como el fondo, las líneas y los puntos. *Seaborn* ofrece varios estilos predefinidos para mejorar la estética de las visualizaciones.

La función ```seaborn.set_style()``` permite establecer el estilo de las visualizaciones.

```
sns.set_style(<style>)
```
Donde:
* ```<style>``` es el nombre del estilo que se desea utilizar. Los estilos disponibles en *Seaborn* son los siguientes:

 * ```'darkgrid'``` 
 * ```'whitegrid'```
 * ```'dark'```
 * ```'white'```
 * ```'ticks'```
 
 https://seaborn.pydata.org/tutorial/aesthetics.html#seaborn-figure-styles


### La función ```seaborn.set_theme()```.

La función ```seaborn.set_theme()``` permite configurar la estética de las visualizaciones de forma global, estableciendo el contexto, el estilo, la paleta de colores y otros parámetros de forma simultánea.

```
seaborn.set_theme(context=<contexto>, style=<estilo>, palette=<paleta>, font=<letra>, font_scale=<escala letra>)
```
Donde:
* ```<contexto>``` es el nombre del contexto que se desea utilizar.
* ```<estilo>``` es el nombre del estilo que se desea utilizar.
* ```<paleta>``` es el nombre de la paleta de colores que se desea utilizar.
* ```<letra>``` es el nombre de la fuente tipográfica que se desea utilizar.
* ```<escala letra>``` es un número que ajusta el tamaño de la fuente tipográfica.

https://seaborn.pydata.org/generated/seaborn.set_theme.html#seaborn.set_theme

**Ejemplos:**

* La siguiente celda genera un histograma con la configuración estética predeterminada.

````python
(sns.histplot(data=dataset,
              x="sepal_length",
              hue="species",
              multiple="dodge"))
````

* La siguiente celda aplica el tema con estilo `"whitegrid"` y paleta `"mako"`, y luego regenera el histograma.

````python
sns.set_theme(style="whitegrid", palette="mako")
(sns.histplot(data=dataset,
              x="sepal_length",
              hue="species",
              multiple="dodge"))
````

* La siguiente celda cambia el contexto a `"notebook"`, estilo `"ticks"`, paleta `"pastel"` y fuente `"serif"` antes de graficar.

````python
sns.set_theme(context="notebook",
              style="ticks",
              palette="pastel",
              font="serif")
(sns.histplot(data=dataset, x="sepal_length", hue="species", multiple="dodge"))
````

* La siguiente celda configura un tema oscuro con contexto `"poster"` (para carteles) y regenera el gráfico.

````python
sns.set_theme(context='poster',
              style="dark",
              palette="colorblind",
              font="serif",
              font_scale=1)
(sns.histplot(data=dataset,
              x="sepal_length",
              hue="species",
              multiple="dodge"))
````

* La siguiente celda cambia el contexto a `"talk"` (para presentaciones), manteniendo el estilo oscuro.

````python
sns.set_theme(context='talk',
              style="dark",
              palette="colorblind",
              font="sans",
              font_scale=1)
(sns.histplot(data=dataset,
              x="sepal_length", 
              hue="species", 
              multiple="dodge"))
````

* La siguiente celda restaura el contexto a `"notebook"` con estilo oscuro y paleta `"colorblind"` para accesibilidad.

````python
sns.set_theme(context='notebook',
              style="dark",
              palette="colorblind",
              font="sans",
              font_scale=1)
(sns.histplot(data=dataset,
              x="sepal_length",
              hue="species", 
              multiple="dodge"))
````
