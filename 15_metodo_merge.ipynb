# El método ```df.merge()```.

````python
import pandas as pd
from datetime import datetime
````

El método ```df.merge()``` permite crear en un nuevo *dataframe* resultante de relacionar las columnas especificadas en el *dataframe* de origen (izquierdo) y las del que se ingresa como *argumento* (derecho), indicando las columnas en las que pueda encontrar elementos coincidentes.


```
df.merge(<dataframe>, left_on=[<coli_1>, <coli_2>, ..., <coli_n>], 
 right_on=[<cold_1>, <cold_2>, ..., <col_d n>], 
 on=[<col_1>, <col_2>,.. <col_n>],
 how=<modo>)
```

Donde:

* ```<dataframe>``` es un *dataframe* de *Pandas*.
* Cada ```<coli_i>``` es un objeto ```<str>``` que corresponde al identificador de una columna del *dataframe* que contiene al método.
* ```<cold_i>``` es un objeto ```<str>``` que corresponde al identificador de una columna del *dataframe* ```<dataframe>```.
* Cada ```<col_i>```es un objeto ```<str>``` que corresponde al identificador de una columna que comparte el mismo nombre en ambos *dataframes*.
* ```<modo>``` es el modo en el que se realizará la combinación y puede ser:

 * ```'inner'```, el cual es el valor por defecto y muestra solo las filas que tienen coincidencias en ambos *dataframes*.
 * ```'outer'```, muestra todas las combinaciones posibles entre ambos *dataframes*.
 * ```'left'```, muestra todas las filas del *dataframe* de origen (izquierdo) y solo las filas del *dataframe* ```<dataframe>``` (derecho) que tengan coincidencias con el primero.
 * ```'right'```, muestra todas las filas del *dataframe* ```<dataframe>``` (derecho) y solo las filas del *dataframe* de origen (izquierdo) que tengan coincidencias con el segundo.
 
https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.merge.html

**Ejemplos:**

* La siguiente celda creará al *dataframe* ```clientes```, el cual contiene a las columnas:

* ```'ident'```.
* ```'nombre'```.
* ```'primer apellido'```.
* ```'suc_origen'```.

````python
clientes = pd.DataFrame({'ident':(19232, 
                             19233, 
                             19234, 
                             19235, 
                             19236),
                       'nombre':('Adriana',
                               'Marcos',
                               'Rubén',
                               'Samuel',
                               'Martha'),
                       'primer apellido':('Sánchez',
                                        'García',
                                        'Rincón',
                                        'Oliva',
                                         'Martínez'),
                        'suc_origen':('CDMX01',
                                      'CDMX02',
                                      'CDMX02',
                                      'CDMX01',
                                      'CDMX03')
                        })
````

````python
clientes
````

* La siguiente celda creará al *dataframe* ```sucursales```, el cual contiene a las columnas:

* ```'clave'```.
* ```nombre_comercial```.

````python
sucursales = pd.DataFrame({'clave':('CDMX01', 
                                  'CDMX02', 
                                  'MTY01', 
                                  'GDL01'),
                         'nombre_comercial':('Galerías',
                                             'Centro',
                                             'Puerta de la Silla',
                                             'Minerva Plaza')})
````

````python
sucursales
````

* La siguiente celda creará al *dataframe* ```facturas```, el cual contiene a las columnas:

* ```'folio'```.
* ```'sucursal'```.
* ```'monto'```.
* ```'fecha'```.
* ```'cliente'```.

````python
facturas = pd.DataFrame({'folio':(15234, 
          
                                  15235, 
                      15236, 
                      15237, 
                      15238, 
                      15239, 
                      15240,
                      15241,
                      15242),
             'sucursal':('CDMX01',
                         'MTY01',
                         'CDMX02',
                         'CDMX02',
                         'MTY01',
                         'GDL01',
                         'CDMX02',
                         'MTY01',
                         'GDL01'),
             'monto':(1420.00,
                     1532.00,
                     890.00,
                     1300.00,
                     3121.47,
                     1100.5,
                     12230,
                     230.85,
                     1569),
             'fecha':(datetime(2019,3,11,17,24),
                     datetime(2019,3,24,14,46),
                     datetime(2019,3,25,17,58),
                     datetime(2019,3,27,13,11),
                     datetime(2019,3,31,10,25),
                     datetime(2019,4,1,18,32),
                     datetime(2019,4,3,11,43),
                     datetime(2019,4,4,16,55),
                     datetime(2019,4,5,12,59)),
            'cliente':(19234,
                       19232,
                       19235,
                       19233,
                       19236,
                       19237,
                       19232,
                       19233,
                       19232)
                        })
````

````python
facturas
````

* Cada una de las siguientes dos celdas regresarán un *dataframe* que relacionará los elementos de ```facturas["sucursal"]``` con ```sucursales["clave"]```.

````python
facturas.merge(sucursales, left_on="sucursal", right_on="clave")
````

````python
sucursales.merge(facturas, left_on="clave", right_on="sucursal")
````

* La siguiente celda regresará un *dataframe* que relacionará los elementos de ```facturas["cliente"]``` con ```clientes["ident"]```.

````python
facturas.merge(clientes, left_on="cliente", right_on="ident")
````

* La siguiente celda regresará un *dataframe* que relacionará los elementos de ```clientes["suc_origen"]``` con ```sucursales["clave"]```.

````python
clientes.merge(sucursales, left_on='suc_origen', right_on='clave')
````

* La siguiente celda regresará un *dataframe* que relacionará:
 * Los elementos de ```facturas["cliente"]``` con ```clientes["ident"]```.
 * Los elementos de ```facturas["sucursal"]``` con ```clientes["suc_origen"]```.

* El *dataframe* resultante contendrá exclusivamente aquellos elementos en los que exista coincidencia en ambas relaciones.

````python
facturas.merge(clientes, left_on=["cliente", "sucursal"],
               right_on=["ident", "suc_origen"])
````

* La siguiente celda regresará un *dataframe* que relacionará:
 * Los elementos de ```facturas["cliente"]``` con ```clientes["ident"]```.
 * Los elementos de ```facturas["sucursal"]``` con ```clientes["suc_origen"]```.
 * Se ingresará el argumento ```how="inner"```.

* El *dataframe* resultante contendrá exclusivamente aquellos elementos en los que exista coincidencia en ambas relaciones.

````python
facturas.merge(clientes, left_on=["cliente", "sucursal"],
               right_on=["ident", "suc_origen"], how="inner")
````

* La siguiente celda regresará un *dataframe* que relacionará:
 * Los elementos de ```facturas["cliente"]``` con ```clientes["ident"]```.
 * Los elementos de ```facturas["sucursal"]``` con ```clientes["suc_origen"]```.
 * Se ingresará el argumento ```how="outer"```.

* El *dataframe* resultante:
 * Contendrá a todas las posibles relaciones entre los *dataframes* ```facturas``` y ```clientes```.
 * Cuando no existan coincidencias entre *dataframes*, la información faltante será completada con valores ```pd.nan```. 

````python
facturas.merge(clientes, left_on=["cliente", "sucursal"],
               right_on=["ident", "suc_origen"], how="outer")
````

* La siguiente celda regresará un *dataframe* que relacionará:
 * Los elementos de ```facturas["cliente"]``` con ```clientes["ident"]```.
 * Los elementos de ```facturas["sucursal"]``` con ```clientes["suc_origen"]```.
 * Se ingresará el argumento ```how="left"```.

* El *dataframe* resultante:
 * Contendrá a las posibles relaciones del *dataframes* ```facturas```.
 * Cuando no existan coincidencias entre *dataframes*, la información faltante será completada con valores ```pd.nan```. 

````python
facturas.merge(clientes, left_on=["cliente", "sucursal"],
               right_on=["ident", "suc_origen"], how="left")
````

* La siguiente celda regresará un *dataframe* que relacionará:
 * Los elementos de ```facturas["cliente"]``` con ```clientes["ident"]```.
 * Los elementos de ```facturas["sucursal"]``` con ```clientes["suc_origen"]```.
 * Se ingresará el argumento ```how="right"```.

* El *dataframe* resultante:
 * Contendrá a las posibles relaciones del *dataframes* ```clientes```.
 * Cuando no existan coincidencias entre *dataframes*, la información faltante será completada con valores ```pd.nan```. 

````python
facturas.merge(clientes, left_on=["cliente", "sucursal"],
               right_on=["ident", "suc_origen"], how="right")
````
