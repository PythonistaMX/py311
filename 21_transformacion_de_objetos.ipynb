## El método ```df.combine()```.

El método ```combine()``` permite comparar dos *dataframes* por columnasy aplicar una función a cada par de elementos correspondientes para obtener un nuevo *dataframe*. 

```
df.combine(<dataframe>, <func>, axis=0, fill_value=<valor>)
```

Donde:
* ```<dataframe>``` es un *dataframe* cuyos índices y columnas coinciden con los índices y columnas del *dataframe* que contiene al método ```combine()```.
* ```<func>``` es una función que comparará cada par de elementos correspondientes de los dos *dataframes* y devolverá un nuevo valor.
* ```<valor>``` es un valor que se utilizará para reemplazar los valores faltantes antes de aplicar la función.

El primer argumento de la función ```<func>``` es el elemento del *dataframe* que contiene al método ```combine()```, mientras que el segundo argumento es el elemento del *dataframe* pasado como primer argumento del método.


https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.combine.html

**Ejemplos:**

* La siguiente celda define al *dataframe* `otra_poblacion`.

````python
otra_poblacion = pd.DataFrame({
        'Animal':(
            'lobo',
            'coyote',
            'jaguar',
            'cerdo salvaje',),
        'Norte_I':(
            1,
            1,
            1,
            1,),
        'Centro_II':(
            1,
            1, 
            1,
            1,)

        }).set_index('Animal')
````

````python
otra_poblacion
````

* Se utilizará el *dataframe* ```poblacion``` definido previamente.

````python
poblacion
````

* La funcion `promedio()` es definida a continuación.
   * La función recibe dos series como argumentos.
   * En caso de que una de las series contenga únicamente valores `np.nan` (lo que ocurre generalmente cuando el elementro corresondiente no existe en uno de los *dataframes*), la función devolverá la otra serie.
   * En cualquier otro caso, regresará una serie con el resultado de calcular el promedio de cada par de elementos correspondientes de ambas series.

````python
def promedio( derecha:pd.Series, izquierda:pd.Series) -> pd.Series:
    if derecha.isna().all():
        return izquierda
    if izquierda.isna().all():
        return derecha
    return (derecha + izquierda) / 2
````

* La siguiente celda regresará un *dataframe* resultante de combinar los *dataframes* `poblacion` y `otra_poblacion` utilizando la función `promedio()`.
* En este caso, una operación aritmética en la que uno de los operandos es `np.nan` devolverá `np.nan`.

````python
poblacion.combine(otra_poblacion, promedio)
````

* La siguiente celda regresará un *dataframe* resultante de combinar los *dataframes* `poblacion` y `otra_poblacion` utilizando la función `promedio()`.
* Los valores faltantes se reemplazarán por `0` antes de aplicar la función `promedio()`.

````python
poblacion.combine(otra_poblacion, promedio, fill_value=0)
````

## El método ```melt()```.

Al analizar un dataset, es común identificar elementos de este que infuyen en el resultado de otro.

El método ```melt()``` permite extraer datos a partir de un *dataframe*, identificando causas (```id_vars ```) y efectos (```value_vars```), permitiendo hacer más explícita esta relación.

Este método crea un nuevo *dataframe* en el que las columnas enumeradas como ```value_vars``` se presentan en función de las columnas enumeradas como ```id_vars```.

```
df.melt(id_vars=[<col i1>, <col i2>, ... <col in>], 
 value_vars=[<col v1>, <col v2>, ... <columnas vn>,
 var_name='<nombre de variable>',
 value_name='<nombre de valores>'])
```

Donde:

* ```<col ix>``` son las columnas que se designan como variables independientes.
* ```<col vx>``` son las columnas que se designan como variables dependientes de las columnas definidas en ```id_vars```.
* ```<nombre de variable>``` es el nombre que se le puede dar a la columna ```'variable'```.
* ```<nombre de valores>``` es el nombre que se le puede dar a la columna ```'value'```.

El *dataframe* resultante presentará:

* Las columnas listadas en el parámetro ```id_vars```. 
* Una columna llamada por defecto ```variable``` que contendrá en sus renglones los nombres de las columnas listadas en el parámetro ```value_vars```. El nombre de esta columna puede asignarse mediante el parámetro ``` var_name```.
* Una columna llamada por defecto ```value``` que contendrá los valores contenidos a su vez en las columnas listadas en el parámetro ```value_vars```. El nombre de esta columna puede asignarse mediante el parámetro ``` value_name```.

El *dataframe* mostrará las diversas combinaciones posibles entre las variables independientes que están relacionadas a las variables dependientes y los índices serán numéricos.

https://pandas.pydata.org/docs/reference/api/pandas.melt.html

**Ejemplos:**

* Se utilizará el *dataframe* ```poblacion_traspuesta``` definido previamente.

````python
poblacion_traspuesta
````

* La siguiente celda regresará un dataframe creado a partir de ```poblacion_traspuesta``` en el que la columna ```'lobo'``` se enlista en el parámetro ```id_vars``` y la columna ```venado``` se enlista en el parámetro ```value_vars```.

````python
poblacion_traspuesta.melt(id_vars=["lobo"], value_vars=["venado"])
````

* La siguiente celda regresará un dataframe creado a partir de ```poblacion_traspuesta``` en el que:
 * ```id_vars=["lobo", "coyote"]```
 * ```value_vars=["venado"]```.

````python
poblacion_traspuesta.melt(id_vars=["lobo", "coyote"], value_vars=["venado"])
````

* La siguiente celda regresará un dataframe creado a partir de ```poblacion_traspuesta``` en el que:
 * ```id_vars=["lobo", "coyote"]```
 * ```value_vars=["venado, "cerdo salvaje"]```.
 * ```var_name="presa"```, 
 * ```value_name="población"```

````python
poblacion_traspuesta.melt(id_vars=["lobo", "coyote"], value_vars=["venado", "cerdo salvaje"], 
                 var_name="presa", value_name="población")
````

## El método ```df.pivot()```.
El método ```df.pivot()``` permite crear estructuras de *dataframes* a partir de columnas de un *dataframe* base.

Las columnas del *dataframe* original se utilizarán para definir los índices, columnas y valores del nuevo *dataframe*.

```
df.pivot(columns=<id_col>, index=<id_index>, values=<id_val>)
```
Donde:

* ```<id_col>``` es las columna del *dataframe* que corresponde a las columnas del nuevo *dataframe*.
* ```<id_index>``` es las columna del *dataframe* que corresponde a los índices del nuevo *dataframe*.
* ```<id_val>``` es las columna del *dataframe* que corresponde a los valores del nuevo *dataframe*.

**NOTA:** Los valores que se utilizarán para crear un nuevo *dataframe* deben de ser del mismo tamaño.

https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.pivot.html

**Ejemplo:**

* La siguiente celda creará al *dataframe* `observaciones`, el cual representa:
   * El conteo de las variables: `'peatones'` y `'vehículos'`.
   * En diversas regiones geográficas: `'Norte'`, `'Sur'`, `'Este'`, `'Oeste'`.
   * A lo largo de distintos momentos del día: `'mañana'`, `'tarde'`, `'noche'`.

Este tipo de tablas se conocen como tablas de contingencia y son útiles para analizar la relación entre variables categóricas.

````python
observaciones = pd.DataFrame({'region':('Norte', 'Norte', 'Norte',
                                        'Sur', 'Sur', 'Sur',
                                        'Este', 'Este', 'Este',
                                        'Oeste', 'Oeste', 'Oeste'),
                             'horario':('mañana', 'tarde', 'noche',
                                        'mañana', 'tarde', 'noche',
                                        'mañana', 'tarde', 'noche',
                                        'mañana', 'tarde', 'noche'),
                              'peatones':(153, 589, 35,
                                          215, 702, 87,
                                          95, 806, 22,
                                          290, 1390, 150),
                             'vehículos': (np.nan, 1003, 304,
                                          506, 1421, 150,
                                          271, 1863, 198,
                                          948, 2279, 804)})
````

````python
observaciones
````

* La siguiente celda regresará un nuevo *dataframe* creado a partir del *dataframe* `observaciones` utilizando el método `pivot()`. En este caso:
 * ```columns='region'```
 * ```index='horario'```
 * ```values='peatones'```

````python
observaciones.pivot(columns='region', 
                    values='peatones', 
                    index='horario')
````

* La siguiente celda regresará un nuevo *dataframe* creado a partir del *dataframe* `observaciones` utilizando el método `pivot()`. En este caso:
 * ```columns='region'```
 * ```index='horario'```
 * ```values='vehiculos'```

````python
observaciones.pivot(columns='region', 
                    values='vehículos', 
                    index='horario')
````

## El método ```pd.pivot_table()```.

El método ```pivot_table()``` me permite crear tablas del mismo modo que ```pd.pivot()```, pero no es necesario que el tamaño de la nueva tabla sea igual al de la original. Aquellos valores faltantes que se utilicen para crear el nuevo *dataframe* pueden ser ponderados utilizando una función de agregación.

Las columnas del *dataframe* original se utilizarán para definir los índices, columnas y valores del nuevo *dataframe*.

```
df.pivot_table(columns=<id_col>, index=<id_index>, values=<id_val>, aggfunc=<func>)
```
Donde:

* ```<id_col>``` es las columna del *dataframe* que corresponde a las columnas del nuevo *dataframe*.
* ```<id_index>``` es las columna del *dataframe* que corresponde a los índices del nuevo *dataframe*.
* ```<id_val>``` es las columna del *dataframe* que corresponde a los valores del nuevo *dataframe*.
* ```<func>``` es la función de agregación que se aplicará a los valores ponderados del nuevo *dataframe*. La función por defecto es `np.mean()`.


https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.pivot_table.html



**Ejemplo:**

La siguiente celda creará al *dataframe* `observaciones_2` el cual tiene la misma estructura que el *dataframe* `observaciones` pero con algunos valores faltantes con la excepción de que existe una observación adicional que hace que haya cuatro observaciones para la región `'Sur'` en el horario `'mañana'` mientras que el resto tiene tres observaciones.

````python
observaciones_2 = pd.DataFrame({'region':('Norte', 'Norte', 'Norte',
                                        'Sur', 'Sur', 'Sur', 'Sur',
                                        'Este', 'Este', 'Este',
                                        'Oeste', 'Oeste', 'Oeste'),
                             'horario':('mañana', 'tarde', 'noche',
                                        'mañana', 'tarde', 'noche', 'mañana',
                                        'mañana', 'tarde', 'noche',
                                        'mañana', 'tarde', 'noche'),
                              'peatones':(153, 589, 35,
                                          215, 702, 87, 221,
                                          95, 806, 22,
                                          290, 1390, 150),
                             'vehiculos': (np.nan, 1003, 304,
                                          506, 1421, 150, np.nan,
                                          271, 1863, 198,
                                          948, 2279, 804)})
````

````python
observaciones_2
````

* Debido a esta diferencia en el número de observaciones, el método `pivot()` no podrá crear un nuevo *dataframe* a partir del *dataframe* `observaciones_2` y se generará una excepción de tipo `ValueError`.
    * La columna `'region'` sería designada como las columnas del nuevo *dataframe*.
    * La columna `'horario'` sería designada como los índices del nuevo *dataframe*.
    * La columna `'vehiculos'` sería designada como los valores del nuevo *dataframe*.

````python
observaciones_2.pivot(columns='region', 
                    values='vehiculos', 
                    index='horario')
````

* El método `pivot_table()` sí podrá crear un nuevo *dataframe* a partir del *dataframe* `observaciones_2` y se generará una tabla de contingencia en la que el valor de cada celda será el promedio de los valores correspondientes a esa región y horario.
    * La columna `'region'` sería designada como las columnas del nuevo *dataframe*.
    * La columna `'horario'` sería designada como los índices del nuevo *dataframe*.
    * La columna `'vehiculos'` sería designada como los valores del nuevo *dataframe*.

````python
observaciones_2.pivot_table(columns='region', 
                    values='vehiculos', 
                    index='horario')
````

* 

````python
observaciones_2.pivot_table(columns='region', 
                    values='peatones', 
                    index='horario')
````

* La siguiente celda regresará un nuevo *dataframe* creado a partir del *dataframe* `observaciones_2` utilizando el método `pivot_table()`. En este caso:
 * ```columns='region'```
 * ```index='horario'```
 * ```values='vehiculos'```
 * ```aggfunc=np.sum()```

````python
observaciones_2.pivot_table(columns='region', 
                    values='peatones', 
                    index='horario',
                    aggfunc=np.sum)
````

* La siguiente celda regresará un nuevo *dataframe* creado a partir del *dataframe* `observaciones_2` utilizando el método `pivot_table()`. En este caso:
 * ```columns='region'```
 * ```values='peatones'```

## El método pd.crosses().

El método `crosses()` permite crear un nuevo *dataframe* a partir de la combinación de dos o más columnas de un *dataframe* base. Este método es útil para analizar la relación entre variables categóricas.

```
df.crosses(columns=[<col1>, <col2>, ... <coln>], values=<col_val>, aggfunc=<func>)
```
Donde:
* ```<colx>``` son las columnas del *dataframe* que se utilizarán para crear el nuevo *dataframe*.
* ```<col_val>``` es la columna del *dataframe* que corresponde a los valores del nuevo *dataframe*.
* ```<func>``` es la función de agregación que se aplicará a los valores ponderados del nuevo *dataframe*. La función por defecto es `np.mean()`.
