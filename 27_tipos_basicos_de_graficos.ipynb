```python
import matplotlib as mpl
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

## Preliminares.

```python
np.random.seed(12314124)
```

```python
normal = np.random.normal(50, 20, 500)
```

```python
normal
```

```python
normal2 = np.random.normal(25, 12, 500)
```

```python
ancho = np.random.randint(5, 60, 500)
```

```python
plt.hist(normal)
```

```python
plt.hist(normal, orientation='horizontal')
```

```python
plt.hist(normal, color='red')
```

```python
plt.hist(normal, bins=13)
```

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist2d.html

```python
plt.hist2d(normal, normal2)
```

## Boxplot.

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.boxplot.html

```python
plt.boxplot(normal)
```

```python
plt.boxplot((normal, normal2))
```

## Scatterplot.

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html

```python
plt.scatter(normal, normal2)
```

```python
plt.scatter(normal, normal2, s=ancho, c=ancho)
```

## Gráfico de líneas.

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html


```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
```

```python
plt.plot(x, y, linestyle='--', marker='o')
```

## Gráfico de barras.

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html


```python
categorias = ['A', 'B', 'C', 'D', 'E']
valores = [23, 45, 56, 78, 32]
plt.bar(categorias, valores)
```

```python
plt.barh(categorias, valores)
```

## Gráfico de pastel (Pie chart).

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.pie.html


```python
plt.pie(valores, labels=categorias)
```

```python
plt.pie(valores, labels=categorias, autopct='%1.1f%%', startangle=90)
```

## Gráfico de área (área bajo curva).

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.fill_between.html


```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
plt.fill_between(x, y1, alpha=0.3)
plt.plot(x, y1)
```

```python
plt.fill_between(x, y1, y2, alpha=0.3)
plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.legend()
```

## Hexbin (Mapa de densidad hexagonal).

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hexbin.html


```python
plt.hexbin(normal, normal2, gridsize=30, cmap='YlOrRd')
plt.colorbar()
```

## Stem plot.

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.stem.html


```python
x = np.linspace(0, 10, 20)
y = np.sin(x)
plt.stem(x, y)
```

## Step plot (Gráfico por pasos).

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.step.html


```python
x = np.linspace(0, 10, 20)
y = np.sin(x)
plt.step(x, y)
```

```python
plt.step(x, y, where='mid')
```

## Violín plot (distribución de datos).

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.violinplot.html


```python
plt.violinplot([normal, normal2])
```

## Mapa de calor (Heatmap).

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.imshow.html


```python
data = np.random.rand(10, 10)
plt.imshow(data, cmap='viridis')
plt.colorbar()
```

## Gráfico de contorno (Contour plot).

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.contour.html


```python
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))
plt.contour(X, Y, Z, levels=15)
plt.colorbar()
```

```python
plt.contourf(X, Y, Z, levels=15, cmap='viridis')
plt.colorbar()
```
