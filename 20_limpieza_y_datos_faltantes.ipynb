# Limpieza y datos faltantes.

En este capítulo se explorarán diversos métodos enfocados a gestionar *dataframes* que no son homogéneos en sus contenidos.

````python
import pandas as pd
import numpy as np
````

## El *dataframe* ilustrativo.

El *dataframe* ```poblacion``` describe una serie de muestras poblacionales de animales en varias regiones geográficas.

````python
poblacion = pd.DataFrame({'Animal':('lobo',
                                   'coyote',
                                   'jaguar',
                                   'cerdo salvaje',
                                    'tapir',
                                    'venado',
                                    'ocelote',
                                    'puma'),
                         'Norte_I':(12,
                                   np.nan,
                                    None,
                                    2,
                                    4,
                                    2,
                                    14,
                                    5
                                   ),
                          'Norte_II':(23,
                                    4,
                                    25,
                                    21,
                                    9,
                                    121,
                                    1,
                                    2
                                   ),
                         'Centro_I':(15,
                                    23,
                                    2,
                                    None,
                                    40,
                                    121,
                                    0,
                                    5),
                         'Sur_I':(28,
                                  46,
                                  14,
                                  156,
                                  79,
                                  12,
                                  2,
                                  np.nan)}).set_index('Animal')
````

````python
poblacion
````

## Métodos de validación de *NaN*.

En muchos casos, los *dataframes* incluyen objetos de tipo ```np.nan```. Por lo general este tipo de dato denota datos incompletos cuyo verdadero valor es desconocido.

Poder transformar eficientemente ```np.nan``` en valores relevantes requiere de experiencia y conocimiento de los datos con los que se trabaja.

### El método `isna()` y el método `isnull()`.

Los métodos `isna()` e `isnull()` se utilizan para detectar valores faltantes en un *dataframe* o serie. Ambos métodos devuelven un *dataframe* o Series de tipo booleano, donde `True` indica la presencia de un valor faltante (`np.nan` o `None`) y `False` indica la ausencia de un valor faltante.

```
<obj>.isna()
```

Donde:
* `<obj>` es un objeto de *Pandas*.

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.isna.html

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.isnull.html

**Ejemplo:**

* Se utilizará al *dataframe* ```poblacion``` definido previamente.

````python
poblacion
````

* Las siguientes celdas evaluarán cada elemento de ```poblacion``` validando si existe algún valor igual a ```np.nan``` o `None`. En caso de que el elemento sea ```np.nan```, el valor dentro del *dataframe* resultante será ```True```.

````python
poblacion.isna()
````

````python
poblacion.isnull()
````

### El método ```notna()```.

El método `notna()` es el opuesto de `isna()`. Devuelve un *dataframe* o serie de tipo booleano, donde `True` indica la ausencia de un valor faltante y `False` indica la presencia de un valor faltante.

```
<obj>.notna()
```
Donde:

* `<obj>` es una serie o un *dataframe* de *pandas.

http://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.notna.html

**Ejemplo:**

* La siguiente celda evaluará a cada elemento de ```poblacion``` validando si existe algún valor igual a ```np.nan```. En caso de que el elemento sea ```np.nan```, el valor dentro del *dataframe* resultante será ```False```.

````python
poblacion.notna()
````

### El método ```notnull()```.

Este método de enmascaramiento detecta aquellos elementos que contienen valores distintos a ```np.nan``` o a ```None```.

```
df.notnull()
```
http://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.notnull.html

**Ejemplo:**

* La siguiente celda evaluará a cada elemento de ```poblacion``` validando si existen valore distintos a ```np.nan``` o a ```None```. En caso de que el elemento sea ```np.nan``` o ```None```, el valor dentro del *dataframe* resultante será ```False```.

````python
poblacion.notnull()
````

### El método ```fillna()```.

Este método sustituirá los valores ```np.nan``` con el valor designado como argumento.


```
df.fillna(<obj>)
```

Donde:

* ```<obj>``` es cualquier objeto de *Python*, *Numpy* o de *Pandas*.

http://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.fillna.html

**Ejemplo:**

* Se utilizará el *dataframe* ```poblacion``` definido previamente.

````python
poblacion
````

* Las siguientes celdas sustituirán a los elementos ```np.nan``` por el objeto ingresado como argumento.

````python
poblacion.fillna(0)
````

````python
poblacion.fillna(15)
````

````python
poblacion.fillna("inválido")
````

## El método ```interpolate()```.

Este método realiza cálculos de interpolación para sustituir a ```np.nan```.


```
<obj>.interpolate(method=<método>, axis=<eje>)
```

Donde:

* ```<obj>``` es un *dataframe* o una *serie* de *Pandas*.
* ```<método>``` es un método de interpolación. El valor por defecto es ```'linear'```.
* El parámetro ```axis``` define el eje desde el cual se tomarán los elementos de interpolación y su valor por defecto es ```0```.

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.interpolate.html

**Nota:** *Scipy* cuenta con diversos algoritmos de interpolación, los cuales pueden ser consultados en: 

* https://docs.scipy.org/doc/scipy/reference/interpolate.html
* https://docs.scipy.org/doc/scipy/reference/generated/scipy.interpolate.interp1d.html

**Ejemplos:**

* Se utilizará el *dataframe* ```poblacion``` definido previamente.

````python
poblacion
````

* La siguiente celda ejecutará el método ```poblacion.interpolate()``` con los argumentos por defecto.
* El *dataframe* resultante modificará aquellos elementos ```np.nan``` aplicando una interpolación lineal tomando como datos de referencia a los del renglón al que pertenece el elemento. 

````python
poblacion.interpolate()
````

* La siguiente celda ejecutará el método ```poblacion.interpolate()``` con el argumento ```axis=1```.
* El *dataframe* resultante modificará aquellos elementos ```np.nan``` aplicando una interpolación lineal tomando como datos de referencia a los de la columna a la que pertenece el elemento.

````python
poblacion.interpolate(axis=1)
````

* La siguiente celda aplicará el método `interpolate(0)` a la columna ```'Norte_I'``` del *dataframe* ```poblacion```.

````python
poblacion['Norte_I'].interpolate()
````

* La siguiente celda usará el argumento ```method="zero"```.
* El método ```"zero"``` requiere que exista un índice numérico para poder realizar la interpolación, por lo que se desencadenará una excepción de tipo ```ValueError```.

````python
poblacion.interpolate(method="zero")
````

* La siguiente celda creará un *dataframe* llamado ```poblacion_numerica```, basado en ```poblacion``` en el que los índices serán numéricos y se desechará la columna ```'Animal'```.

````python
poblacion_numerica = poblacion.reset_index().drop('Animal', axis=1)
````

````python
poblacion_numerica
````

* La siguiente celda aplicará el método ```poblacion_numerica.interpolate()```:
 * Ingresando el argumento ```axis=0```.
 * Ingresando el argumento ```method="zero"```.

````python
poblacion_numerica.interpolate(method="zero", axis=0)
````

## El método ```dropna()```.

Este método desechará los renglones o columnas que contengan valores ```np.nan```.


```
<obj>.dropna(axis=<eje>)
```

Donde:

* ```<obj>``` es un *dataframe* o una *serie* de *Pandas*.
* ```<eje>``` define el eje desde el cual se tomarán los elementos de referencia para eliminar renglones o columnas.

* El valor por defecto es ```0```, lo que indica que se eliminarán los renglones que contengan al menos un elemento ```np.nan```.
* Si el valor es ```1```, se eliminarán las columnas que contengan al menos un elemento ```np.nan```.



**Ejemplo:**

* Se utilizará el *dataframe* ```poblacion``` definido previamente.

````python
poblacion
````

* La siguiente celda eliminará los renglones que contengan al menos un elemento ```np.nan```.

````python
poblacion.dropna()
````

* La siguiente celda eliminará los renglones que contengan al menos un elemento ```np.nan```.

````python
poblacion.dropna(axis=1)
````

* La siguiente celda aplicará el método ```dropna()``` a la columna ```'Norte_I'``` del *dataframe* ```poblacion```.

````python
poblacion['Norte_I'].dropna()
````

## El método ```duplicated()```.

El método ```duplicated()``` devuelve un *dataframe* de tipo booleano en elemento (fila o columna, dependiendo del eje) que se repite dentro del objeto original es marcado con un valor ```True```. 

La primera aparición de un elemento no es marcada como duplicada, por lo que se marcará con un valor ```False```.


```
<obj>.duplicated(axis=<eje>)
```

Donde:
* ```<obj>``` es un *dataframe* o una *serie* de *Pandas*.
* ```<eje>``` puede ser 0 para filas o 1 para columnas.

http://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.duplicated.html


**Ejemplo:**

* Se utilizará el *dataframe* ```poblacion``` definido previamente. 

````python
poblacion
````

* La siguiente celda aplicará el método `duplicated()` al *dataframe* ```poblacion``` para identificar las filas duplicadas.
* En este caso, no existen filas duplicadas, por lo que el *dataframe* resultante estará compuesto únicamente por valores ```False```.

````python
poblacion.duplicated()
````

* La siguiente celda aplicará el método `duplicated()` a la columna `'Norte_I'` del *dataframe* ```poblacion``` para identificar los valores duplicados.

````python
poblacion['Norte_I'].duplicated()
````

* La siguiente celda definirá el dataframe `otra_poblacion`, en el que el rengón con índice `'coyote'` es idéntico al rengón con índice `'puma'`.

````python
otra_poblacion = pd.DataFrame({'Animal':('lobo',
                                   'coyote',
                                   'jaguar',
                                   'cerdo salvaje',
                                    'tapir',
                                    'venado',
                                    'ocelote',
                                    'puma'),
                         'Norte_I':(12,
                                    4,
                                    None,
                                    2,
                                    4,
                                    2,
                                    14,
                                    4
                                   ),
                          'Norte_II':(23,
                                    4,
                                    25,
                                    21,
                                    9,
                                    121,
                                    1,
                                    4
                                   ),
                         'Centro_I':(15,
                                    4,
                                    2,
                                    120,
                                    40,
                                    121,
                                    0,
                                    4),
                         'Sur_I':(28,
                                  4,
                                  14,
                                  156,
                                  79,
                                  12,
                                  2,
                                  4)}).set_index('Animal')
````

````python
otra_poblacion
````

* La siguiente celda aplicará el método `duplicated()` a `otra_poblacion` para identificar las filas duplicadas.

````python
otra_poblacion.duplicated()
````

## El método ```drop_duplicates()```.

El método ```drop_duplicates()``` elimina los renglones o columnas que se repiten dentro del objeto original, dejando únicamente la primera aparición de cada elemento.

```
<obj>.drop_duplicates(axis=<eje>)
```
Donde:
* ```<obj>``` es un *dataframe* o una *serie* de *Pandas*.
* ```<eje>``` puede ser 0 para filas o 1 para columnas.

http://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.drop_duplicates.html

**Ejemplo:**

* Se utilizará el *dataframe* ```otra_poblacion``` definido previamente.

````python
otra_poblacion
````

* Se utilizará el método `drop_duplicates()` para eliminar las filas duplicadas del *dataframe* `otra_poblacion`. La fila con índice `'puma'` será eliminada, ya que es idéntica a la fila con índice `'coyote'`.

````python
otra_poblacion.drop_duplicates()
````
